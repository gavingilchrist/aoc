```python
# Get raw data
with open('input/22.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
grid = np.array([list(i) for i in rawinput.split('\n')])
ypos, xpos = [i//2 for i in grid.shape]
n_infected = 0
for i in range(10000):
    if grid[ypos, xpos] == '#':
        ypos, xpos = grid.shape[1]-xpos-1, ypos
        grid = grid.T[::-1]
        grid[ypos, xpos] = '.'
    else:
        ypos, xpos = xpos, grid.shape[0]-ypos-1
        grid = grid[::-1].T
        grid[ypos, xpos] = '#'
        n_infected += 1
        
    if ypos == 0:
        grid = np.append(np.full_like(grid[:1], '.'), grid, axis=0)
    else:
        ypos -= 1
n_infected
```

```python
# Part 2
grid = np.array([list(i) for i in rawinput.split('\n')])
ypos, xpos = [i//2 for i in grid.shape]
n_infected = 0
for i in range(10_000_000):
    if grid[ypos, xpos] == '#':
        ypos, xpos = grid.shape[1]-xpos-1, ypos
        grid = grid.T[::-1]
        grid[ypos, xpos] = 'F'
    elif grid[ypos, xpos] == 'F':
        ypos, xpos = grid.shape[0]-ypos-1, grid.shape[1]-xpos-1
        grid = grid[::-1, ::-1]
        grid[ypos, xpos] = '.'
    elif grid[ypos, xpos] == '.':
        ypos, xpos = xpos, grid.shape[0]-ypos-1
        grid = grid[::-1].T
        grid[ypos, xpos] = 'W'
    else:
        grid[ypos, xpos] = '#'
        n_infected += 1
        
    if ypos == 0:
        grid = np.append(np.full_like(grid[:1], '.'), grid, axis=0)
    else:
        ypos -= 1
n_infected
```
