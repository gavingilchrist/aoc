```python
# Get raw data
with open('input/06.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
banks = [int(i) for i in rawinput.split('\t')]
nb = len(banks)
hist = []
while (z:=tuple(banks)) not in hist:
    hist += [z]
    banks[(ix:=banks.index(b:=max(banks)))] = 0
    banks = [j + b//nb + int((i-ix-1)%nb < b%nb) for i,j in enumerate(banks)]
len(hist)
```
