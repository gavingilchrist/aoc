```python
# Get raw data
with open('input/08.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
registers = dict()
for reg, op, amt, cond in [[(z:=i.split())[0], [-1,1][z[1]=='inc'], int(z[2]), 
                            f"registers.get('{z[4]}',0){z[5]}{z[6]}"] 
                           for i in rawinput.split('\n')]:
    if eval(cond):
        registers[reg] = registers.get(reg, 0) + op * amt
max(registers.values())
```

```python
# Part 2
registers = dict()
hival = 0
for reg, op, amt, cond in [[(z:=i.split())[0], [-1,1][z[1]=='inc'], int(z[2]), 
                            f"registers.get('{z[4]}',0){z[5]}{z[6]}"] 
                           for i in rawinput.split('\n')]:
    if eval(cond):
        registers[reg] = (z:=registers.get(reg, 0) + op * amt)
        if z>hival: hival = z
hival
```
