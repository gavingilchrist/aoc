```python
# Get raw data
with open('input/21.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
asarray = lambda x: np.array([list(i) for i in x.split('/')])
astuples = lambda x: tuple(tuple(j.item() for j in i) for i in x)
def allways(x):
    olist = {astuples(x)}
    for i in range(7):
        olist.add(astuples(x:=(x[::-1] if i%4==3 else x[::-1].T)))
    return olist

conv = {k: asarray(j[2])
        for i in rawinput.split('\n') 
        for j in [i.split()]
        for k in allways(asarray(j[0]))}

pixels = np.array([['.','#','.'],
                   ['.','.','#'],
                   ['#','#','#']])

for i in range(5):
    d = [i for i in [2,3] if (size:=pixels.shape[0]) % i == 0][0]
    pixels = np.concat([np.concat([conv[astuples(pixels[sr:sr+d, sc:sc+d])] 
                                   for sc in range(0, size, d)], 
                                  axis=1) 
                        for sr in range(0, size, d)], 
                       axis=0)
    
np.sum(pixels == '#').item()
```
