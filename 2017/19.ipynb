```python
# Get raw data
with open('input/19.txt', 'r') as f:
    rawinput = f.read()
```

```python
import numpy as np
```

```python
# Parts 1 and 2
routing = np.pad([list(i) for i in rawinput.split('\n') if i], 1, constant_values=' ')
posy, posx = 1, np.argmax(routing[1] == '|').item()
letters = ''
nsteps = 1
while True:
    if ord('A') <= ord(routing[posy, posx]) <= ord('Z'):
        letters += routing[posy, posx]
    if routing[posy+1, posx] != ' ':
        posy += 1
        nsteps += 1
    elif routing[posy, posx-1] != ' ':
        posy, posx = routing.shape[1]-posx-1, posy
        routing = routing.T[::-1]
    elif routing[posy, posx+1] != ' ':
        posy, posx = posx, routing.shape[0]-posy-1
        routing = routing[::-1].T
    else:
        break
letters, nsteps
```
