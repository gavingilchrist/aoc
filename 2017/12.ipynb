```python
# Get raw data
with open('input/12.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
pipes = {int(j[0]): [int(k.replace(',', '')) for k in j[2:]]
         for i in rawinput.split('\n') 
         for j in [i.split()]}

connected = {0}
while connected != (connected:=connected | {j for i in connected for j in pipes[i]}):
    pass

len(connected)
```

```python
# Part 2
programs = {*pipes}
groups = []
while programs:
    connected = {programs.pop()}
    while connected != (connected:=connected | {j for i in connected for j in pipes[i]}):
        pass
    groups += [connected]
    programs -= connected
len(groups)
```
