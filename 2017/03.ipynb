```python
# Get raw data
rawinput = 277678
```

```python
# Part 1
coords = [(sqn:=((z:=int((rawinput-1)**0.5)) + z%2) // 2)]*2
lsnum = rawinput - ((sqn-1)*2+1)**2
x, d = 0, -1
for i in range(4):
    coords[x] += d * (z:=min(lsnum, sqn*2))
    lsnum -= z
    if x: d = -d
    x = 1-x
sum([abs(i) for i in coords])
```

```python
# Part 2
import numpy as np

sq = np.pad([[1]], 1, constant_values=0)
layer = 0
coords = np.ones(2, dtype=int)
mvs = []
while np.max(sq) <= rawinput:
    if len(mvs)==0:
        sq = np.pad(sq, 1, constant_values=0)
        layer += 1
        coords += 1
        mvs += np.column_stack((np.append([1], np.repeat([0,1,0,1], layer*2)[1:]),
                                np.append([1], np.repeat([-1,1], layer*4)[1:]))).tolist()
    c,v = mvs.pop(0)
    coords[c] += v
    sq[*coords] = np.sum(sq[coords[0]-1:, coords[1]-1:][:3, :3])

sq[*coords].item()
```
