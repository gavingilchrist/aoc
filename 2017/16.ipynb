```python
# Get raw data
with open('input/16.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
def do_dance(programs):
    for i in rawinput.split(','):
        if i.startswith('s'):
            programs = programs[-int(i[1:]):]+programs[:-int(i[1:])]
        elif i.startswith('x'):
            x,y = [int(j) for j in i[1:].split('/')]
            programs[x], programs[y] = programs[y], programs[x]    
        elif i.startswith('p'):
            programs = [dict(zip((z:=i[1:].split('/')),z[::-1])).get(j, j) 
                        for j in programs]
    return programs

programs = do_dance([chr(i+ord('a')) for i in range(16)])
''.join(programs)
```

```python
# Part 2
histkey = lambda x: ''.join(x)
hist = [histkey(programs)]
dcount=0
while True:
    dcount += 1
    if (z:=histkey(programs:=do_dance(programs))) in hist:
        break
    else:
        hist += [z]

hist[(1_000_000_000 - 1 - dcount) % dcount]
```
