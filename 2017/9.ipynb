```python
# Get raw data
with open('input/9.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
rm_canc = lambda x: x[:(z:=x.index('!'))] + rm_canc(x[z+2:]) if '!' in x else x
rm_garbage = lambda x: x[:(s:=x.index('<'))] + rm_garbage(x[x.index('>',s)+1:]) if '<' in x else x
group = lambda x: x[x.index('{'):
                    [(z:={'{':1, '}':-1}.get(j,0)+z if i 
                      else {'{':1, '}':-1}.get(j,0)) 
                      for i,j in enumerate(x)].index(0, x.index('{'))+1]
def score(x,y=1):
    sctot = 0
    while '{' in x:
        sctot += y+score((z:=group(x))[1:-1], y+1)
        x = x[x.index('{') + len(z):]
    return sctot

score(rm_garbage(rm_canc(rawinput)))
```

```python
# Part 2
garbage_ct = lambda x: x.index('>', (s:=x.index('<')))-s-1 + garbage_ct(x[x.index('>',s)+1:]) if '<' in x else 0
garbage_ct(rm_canc(rawinput))
```
