```python
# Get raw data
with open('input/05.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
instr = [int(i) for i in rawinput.split('\n')]
ptr = 0
nsteps = 0
while 0 <= ptr < len(instr):
    instr[ptr] = (j:=instr[ptr])+1
    ptr += j
    nsteps += 1
nsteps
```

```python
# Part 2
instr = [int(i) for i in rawinput.split('\n')]
ptr = 0
nsteps = 0
while 0 <= ptr < len(instr):
    instr[ptr] = j-1 if (j:=instr[ptr]) >= 3 else j+1
    ptr += j
    nsteps += 1
nsteps
```
