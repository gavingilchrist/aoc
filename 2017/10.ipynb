```python
# Get raw data
with open('input/10.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
lengths = [int(i) for i in rawinput.split(',')]

n_nodes = 256
nodes = np.arange(n_nodes)
current_pos = 0
skipsize = 0
for length in lengths:
    ix = np.mod(np.arange(length) + current_pos, n_nodes)
    nodes[ix] = nodes[ix[::-1]]
    current_pos = (current_pos+length+skipsize) % n_nodes
    skipsize += 1
    
np.prod(nodes[:2]).item()
```

```python
# Part 2
lengths = [ord(i) for i in rawinput] + [17, 31, 73, 47, 23]

n_nodes = 256
nodes = np.arange(n_nodes)
current_pos = 0
skipsize = 0
for i in range(64):
    for length in lengths:
        ix = np.mod(np.arange(length) + current_pos, n_nodes)
        nodes[ix] = nodes[ix[::-1]]
        current_pos = (current_pos+length+skipsize) % n_nodes
        skipsize += 1

xor_all = lambda x: np.squeeze(x,1) if x.shape[1] == 1 else xor_all(np.column_stack((x[:, 0] ^ x[:, 1], x[:, 2:])))
''.join(np.vectorize(lambda x: f'00{hex(x)[2:]}'[-2:])(xor_all(nodes.reshape(16,16))))
```
