```python
# Get raw data
with open('input/07.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
supports = {j[0].split('(')[0].strip(): [k.strip() for k in j[1].split(',')]
            for i in rawinput.split('\n') 
            for j in[i.split('>')] if len(j)>1}
weights = {j[0]: int(j[1][1:-1]) 
           for i in rawinput.split('\n') 
           for j in[i.split()]}
(bottom:=[*{*weights}-{*sum(supports.values(), [])}][0])
```

```python
# Part 2
known = dict()
def tot_weight(disc):
    if disc not in known:
        if disc in supports:
            known[disc] = weights[disc] + sum([tot_weight(i) for i in supports[disc]])
        else:
            known[disc] = weights[disc]
    return known[disc]

disc = bottom
newwt = None
while (len(y:=[(x:=j).count(k) 
               for j in[[tot_weight(i) for i in supports[disc]]] 
               for k in j]) > 1 
       and 1 in y):
    disc = supports[disc][y.index(1)]
    newwt = weights[disc] + [z for i in x if (z:=i-x[y.index(1)]) != 0][0]    
newwt
```
