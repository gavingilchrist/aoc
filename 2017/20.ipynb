```python
# Get raw data
with open('input/20.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
pos, vel, acc = np.hsplit(np.array([[int(l) 
                                     for k in j 
                                     for l in k.split('<')[1].split('>')[0].split(',')]
                                    for i in rawinput.split('\n')
                                    for j in [i.split()]]), 
                          3)
t = int(1e8)
np.argmin(d:=np.sum(np.abs(pos + t*vel + t*(t+1)//2*acc), axis=1)).item()
```

```python
# Part 2
pt, vt, at = [np.copy(i) for i in [pos, vel, acc]]
t = 0
while True:
    t += 1
    pt = (ptprev:=pt) + (vt:=vt + at)
    ix = np.column_stack(((ix1:=np.repeat(np.arange(pt.shape[0]), 
                                          pt.shape[0])),
                          (ix2:=np.tile(np.arange(pt.shape[0]), 
                                        pt.shape[0]))))[ix1<ix2]
    if np.any(np.all(np.abs(np.sum(ptprev[ix] * [[[1], [-1]]], axis=1)) 
                     > np.abs(np.sum(pt[ix] * [[[1], [-1]]], axis=1)), 
                     axis=1)):
        coll_ix = np.unique(ix[np.all(np.ptp(pt[ix], axis=1) == 0, axis=1)])
        pt, vt, at = [np.delete(i, coll_ix, axis=0) for i in [pt, vt, at]]
    else:
        break
pt.shape[0]
```
