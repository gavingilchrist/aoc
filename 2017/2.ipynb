```python
# Get raw data
with open('input/2.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum([max(j)-min(j)
     for i in rawinput.split('\n')
     for j in [[int(k) for k in i.split('\t')]]])
```

```python
# Part 2
sum([num//denom 
     for i in rawinput.split('\n')
     for j in [[int(k) for k in i.split('\t')]]
     for k, num in enumerate(j) 
     for l, denom in enumerate(j) 
     if k!=l and num%denom == 0])
```
