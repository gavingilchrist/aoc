```python
# Get raw data
with open('input/24.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
components = {tuple(sorted([int(j) for j in i.split('/')])) for i in rawinput.split('\n')}

get_strongest = lambda x, y=0: max([(z:=sum(i)) + get_strongest(x-{i}, z-y)
                                    for i in x if any([j==y for j in i])]
                                   or [0])

get_strongest(components)
```

```python
# Part 2
get_longest_str = lambda x,y=0: (sorted([[sum(k) 
                                          for z in[sum(i)]
                                          for k in zip([1, z], 
                                                       get_longest_str(x-{i}, 
                                                                       z-y))]
                                         for i in x
                                         if any([j==y for j in i])])
                                 or [[0,0]])[-1]

get_longest_str(components)[1]
```
