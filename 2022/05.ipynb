```python
# Get raw data
with open('input/05.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
rawstacks, rawinstr = [i for i in rawinput.split('\n\n')]

stacks = [[j for j in i[-2::-1] 
           if j.strip()] 
          for i in zip(*rawstacks.split('\n')) 
          if i[-1].strip()]

for n,f,t in [[int(k)-(j>0) for j,k in enumerate(i.split()[1::2])] 
              for i in rawinstr.split('\n')]:
    stacks[t] += [stacks[f].pop() for i in range(n)]

print(''.join([i[-1] for i in stacks]))
```
