```python
# Get raw data
with open('input/14.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
walls = [[[int(k) for k in j.split(',')[::-1]] 
          for j in i.split(' -> ')] 
         for i in rawinput.split('\n')]
max_y = max([j[0] for i in walls for j in i])+3
cave = np.full((max_y, 1000), 0)
for i in walls:
    for j in range(1,len(i)):
        cave[min(y:=[i[j][0], i[j-1][0]]): max(y)+1,
             min(x:=[i[j][1], i[j-1][1]]): max(x)+1] = 2

def drop_sand():
    spos = [0,500]
    if cave[*spos] == 1:
        return True
    while True:
        if np.any(y:=cave[spos[0]:, spos[1]] != 0):
            ylow = np.argmax(y).item()
            if cave[*(z:=[spos[0]+ylow,spos[1]-1])] == 0:
                spos = z
            elif cave[*(z:=[spos[0]+ylow,spos[1]+1])] == 0:
                spos = z
            else:
                cave[spos[0]+ylow-1,spos[1]] = 1
                return False
        else:
            return True

while not drop_sand():  pass

np.sum(cave == 1).item()
```

```python
# Part 2
cave[max_y-1] = 2

while not drop_sand():  pass

np.sum(cave == 1).item()
```
