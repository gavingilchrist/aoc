```python
# Get raw data
with open('input/03.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum([(z:=ord([*{*k}&{*l}][0]))-[38,96][z>96] 
     for i in rawinput.split('\n')
     for j in [len(i)//2] 
     for k,l in [[i[:j],i[j:]]]])
```

```python
# Part 2
sum([(z:=ord([*{*k[0]}&{*k[1]}&{*k[2]}][0]))-[38,96][z>96]
     for i in [rawinput.split('\n')]
     for j in range(0,len(i),3)
     for k in [i[j:j+3]]])
```
