```python
# Part 1
def max_geodes(blueprint, timeleft=24, robots=None, resources=None):
    if robots is None:
        robots = {i: int(i=='ore') for i in matslist}
        resources = {i: 0 for i in matslist}
    dkey = '-'.join([str(blueprint),
                     str(timeleft), 
                     *[str(robots[i]) for i in matslist],
                     *[str(resources[i]) for i in matslist]])
    if dkey not in known:
        if timeleft:
            poss = [resources['geode'] + robots['geode']*timeleft]
            for nr_num in range(3,-1,-1):
                nr = matslist[nr_num]
                if robots[matslist[max(nr_num-1,0)]]:
                    ttm = max([-(min(resources[mat]-req, 0) // robots[mat])
                               for mat,req in blueprints[blueprint][nr].items()]) + 1
                    if (timeleft-ttm >= 4-nr_num 
                        and (nr in('geode','ore') or (resources[nr] < mats_max_usage[blueprint][nr]))):
                        poss += [max_geodes(blueprint,
                                            timeleft - ttm,
                                            {**robots, **{nr: robots[nr]+1}},
                                            {k: v + robots[k]*ttm - blueprints[blueprint][nr].get(k,0)
                                             for k,v in resources.items()})]
                        if nr=='geode' and ttm==1:
                            break
            known[dkey] = max(poss)
        else:
            known[dkey] = resources['geode']
    return known[dkey]


blueprints = {int(j[1][:-1]):
                 {'ore': 
                     {'ore': k[0]},
                  'clay': 
                      {'ore': k[1]},
                  'obsidian': 
                      {'ore': k[2], 
                       'clay': k[3]},
                  'geode': 
                      {'ore': k[4], 
                       'obsidian': k[5]}}
                 for i in rawinput.split('\n')
                 for j in [i.split()]
                 for k in [[int(j[l]) for l in [6, 12, 18, 21, 27, 30]]]}
matslist = ['ore', 'clay', 'obsidian', 'geode']
mats_max_usage = {b: {i: max([0]+[l for j in blueprints[b].values()
                                  for k,l in j.items()
                                  if k==i])
                      for i in matslist}
                  for b in blueprints}

known=dict()
sum([i*max_geodes(i) for i in blueprints])
```
