```python
# Get raw data
with open('input/09.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
motions = [[j[0],int(j[1])] for i in rawinput.split('\n') for j in [i.split()]]
directions = {'U': (-1,0), 'D': (1,0), 'L': (0,-1), 'R': (0,1)}

hpos = tpos = (0,0)
tpos_hist = {tpos}

for d,n in motions:
    for r in range(n):
        hpos = tuple([sum(i) for i in zip(hpos, directions[d])])
        htoff = [i-j for i,j in zip(hpos,tpos)]
        if max([abs(i) for i in htoff])>1:
            tpos = tuple([sum(j) 
                          for j in zip(tpos, 
                                       [min(max(i,-1),1) for i in htoff])])
            tpos_hist.add(tpos)

len(tpos_hist)
```

```python
# Part 2
pos = [(0,0) for i in range(10)]
tpos_hist = {pos[-1]}

for d,n in motions:
    for r in range(n):
        pos[0] = tuple([sum(i) for i in zip(pos[0], directions[d])])
        for s in range(1,10):
            htoff = [i-j for i,j in zip(pos[s-1],pos[s])]
            if max([abs(i) for i in htoff])>1:
                pos[s] = tuple([sum(j) 
                                for j in zip(pos[s], 
                                             [min(max(i,-1),1) for i in htoff])])
        tpos_hist.add(pos[-1])

len(tpos_hist)
```
