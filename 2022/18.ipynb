```python
# Get raw data
with open('input/18.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
lavadrop = {tuple([int(j) for j in i.split(',')]) 
            for i in rawinput.split('\n')}
sum([sum([tuple([[m,m+k][l==j] 
                 for l,m in enumerate(i)]) not in lavadrop 
          for j in [0,1,2] for k in [-1,1]])
     for i in lavadrop])
```

```python
# Part 2
limits = [(min(i)-1,max(i)+2) for i in zip(*lavadrop)]
air = {(i,j,k)
       for i in range(*limits[0])
       for j in range(*limits[1])
       for k in range(*limits[2])
       if (i,j,k) not in lavadrop}
air -= (water:={min(air)})

while (z:={tuple([[m,m+k][l==j] 
                  for l,m in enumerate(i)])
           for i in water
           for j in [0,1,2] for k in [-1,1]} & air):
    water.update(z)
    air -= z

sum([sum([tuple([[m,m+k][l==j] 
                 for l,m in enumerate(i)]) in water 
          for j in [0,1,2] for k in [-1,1]])
     for i in lavadrop])
```
