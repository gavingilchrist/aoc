```python
# Get raw data
with open('input/04.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum([p1f<=p2f<=p2t<=p1t or p2f<=p1f<=p1t<=p2t
     for i in rawinput.split('\n')
     for p1f,p1t,p2f,p2t in [[k+int(l) 
                              for j in i.split(',') 
                              for k,l in enumerate(j.split('-'))]]])
```

```python
# Part 2
sum([p1f<p2t and p2f<p1t
     for i in rawinput.split('\n')
     for p1f,p1t,p2f,p2t in [[k+int(l) 
                              for j in i.split(',') 
                              for k,l in enumerate(j.split('-'))]]])
```
