```python
# Get raw data
with open('input/15.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sensors, beacons = [[*{*l}]
                    for l in zip(*[[(sy,sx,abs(by-sy)+abs(bx-sx)),
                                    (by,bx)] 
                                   for sx,sy,bx,by in [[int(k.split('=')[1])
                                                        for j in i.split(':') 
                                                        for k in j.split(',')] 
                                                       for i in rawinput.split('\n')]])]

ytarget = 2000000
xranges = sorted([[ll,lh] 
                  for sy,sx,dist in sensors
                  for ll,lh in [[max(ytarget-(sy+dist-sx), (sy-dist+sx)-ytarget),
                                 min((sy+dist+sx)-ytarget, ytarget-(sy-dist-sx))+1]]
                  if ll < lh])

(sum([-max(j,z if i else j)+(z:=max(z,k) if i else k) 
      for i,(j,k) in enumerate(xranges)])
 -sum([b[0]==ytarget 
       and any([i<=b[1]<j for i,j in xranges]) 
       for b in beacons]))
```

```python
# Part 2
diag_p = [[sx-dist+sy,sx+dist+sy+1] for sy,sx,dist in sensors]
diag_n = [[sx-dist-sy,sx+dist-sy+1] for sy,sx,dist in sensors]
diag_p_conv = dict(zip((z:=sorted({*sum(diag_p,[])})), range(len(z))))
diag_n_conv = dict(zip((z:=sorted({*sum(diag_n,[])})), range(len(z))))
cell_flags = {(yy,xx): 0
              for yy in range(len(diag_p_conv)-1)
              for xx in range(len(diag_n_conv)-1)
              if 0<=(x:=((z:=sorted(diag_p_conv)[yy])+sorted(diag_n_conv)[xx])/2)<=4000000
              and 0<=z-x<=4000000}

for i in range(len(sensors)):
    for yy in range(*[diag_p_conv[j] for j in diag_p[i]]):
        for xx in range(*[diag_n_conv[j] for j in diag_n[i]]):
            if (yy,xx) in cell_flags:
                cell_flags[(yy,xx)] = 1
                
yy,xx = [k for k,v in cell_flags.items() if v==0][0]
(x:=sum([(z:=sorted(diag_p_conv)[yy]), sorted(diag_n_conv)[xx]])//2) * 4000000 + (z-x)
```
