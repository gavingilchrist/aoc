```python
# Get raw data
with open('input/20.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
class Number(object):
    def __init__(self, value):
        self.value = value
        self.prev = None
        self.next = None
        
    def __repr__(self):
        return str(self.value)
    
    def __int__(self):
        return self.value
        
    def set_prev(self, item):
        self.prev = item
        if item.next != self:
            item.set_next(self)
        
    def set_next(self, item):
        self.next = item
        if item.prev != self:
            item.set_prev(self)
            
    def move(self):
        nn = self.fwd_find(self.value % (len(sequence)-1) + 1)
        self.prev.set_next(self.next)
        self.set_prev(nn.prev)
        self.set_next(nn)
            
    def fwd_find(self, n):
        nfwd = n % len(sequence)
        item = self
        for i in range(nfwd):
            item = item.next
        return item

sequence = [Number(int(i)) for i in rawinput.split('\n')]
for i,j in zip(sequence, [*sequence[1:], sequence[0]]):
    i.set_next(j)
for i in sequence:
    i.move()
num0 = [i for i in sequence if i.value == 0][0]
sum([int(z:=(z if i else num0).fwd_find(1000)) for i in [0,1,2]])
```
