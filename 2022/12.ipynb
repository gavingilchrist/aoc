```python
# Get raw data
with open('input/12.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
def evolve(reachmap):
    hp = np.pad(heightmap,1,constant_values=np.max(heightmap+2))
    rp = np.pad(reachmap,1,constant_values=-1)
    r_adj = np.min(np.stack((np.where((hp[:-2,1:-1]-heightmap<2)&((z:=rp[:-2,1:-1])>=0), z+1, nc), 
                             np.where((hp[2:,1:-1]-heightmap<2)&((z:=rp[2:,1:-1])>=0), z+1, nc), 
                             np.where((hp[1:-1,:-2]-heightmap<2)&((z:=rp[1:-1,:-2])>=0), z+1, nc), 
                             np.where((hp[1:-1,2:]-heightmap<2)&((z:=rp[1:-1,2:])>=0), z+1, nc))),
                   axis=0)
    return np.where((reachmap==-1) & (r_adj != nc), r_adj, reachmap)

heightmap = np.array([list(i) for i in rawinput.split('\n')]).view(np.uint32).astype(int)
heightmap[*(startpos:=tuple([np.argmax(np.any(heightmap==ord('S'), axis=i)).item() 
                             for i in [1,0]]))] = ord('a')
heightmap[*(endpos:=tuple([np.argmax(np.any(heightmap==ord('E'), axis=i)).item() 
                           for i in [1,0]]))] = ord('z')
nc = heightmap.size

reachmap = -np.ones_like(heightmap)
reachmap[*endpos] = 0
while np.any(reachmap != (reachmap:=evolve(reachmap))):  pass

reachmap[*startpos].item()
```
