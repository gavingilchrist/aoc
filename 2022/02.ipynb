```python
# Get raw data
with open('input/02.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum([l+(l-k+1)%3*3 
     for i in rawinput.split('\n') 
     for k,l in[[(ord(j)-64)%23 for j in i.split()]]])
```

```python
# Part 2
sum([(k+l-2)%3+1+l*3
     for i in rawinput.split('\n')
     for j in [i.split()]
     for k,l in [[ord(j[0])-64, ord(j[1])-88]]])
```
