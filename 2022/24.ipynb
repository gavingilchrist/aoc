```python
# Get raw data
with open('input/24.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
blizzards = [[[(i,j)             
               for i,r in enumerate(rawinput.split('\n'))
               for j,c in enumerate(r)
               if c == d]
              for d in '^v<>']]
locmax = [max(i) for i in zip(*sum(blizzards[0],[]))]
while (z:=[[([locmax[0],i-1][i>1],j) for i,j in blizzards[-1][0]],
           [([1,i+1][i<locmax[0]],j) for i,j in blizzards[-1][1]],
           [(i,[locmax[1],j-1][j>1]) for i,j in blizzards[-1][2]],
           [(i,[1,j+1][j<locmax[1]]) for i,j in blizzards[-1][3]]]) != blizzards[0]:
    blizzards += [z]
    
startpos = (0,1)
endpos = (locmax[0]+1,locmax[1])

def get_min_steps(from_pos, to_pos, start_time=0):
    pos = {from_pos}
    step = start_time
    while True:
        pos = {p
               for i in pos
               for j in [[0,0],[-1,0],[1,0],[0,-1],[0,1]]
               for p in [tuple([sum(k) for k in zip(i,j)])]
               if (p in[from_pos,to_pos] or all([1<=l<=m for l,m in zip(p,locmax)]))
               and p not in sum(blizzards[step%len(blizzards)],[])}
        if to_pos in pos:
            break
        else:
            step += 1
    return step

(p1steps:=get_min_steps(startpos,endpos))
```

```python
# Part 2
get_min_steps(startpos, endpos, get_min_steps(endpos, startpos, p1steps))
```
