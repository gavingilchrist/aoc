```python
# Get raw data
with open('input/11.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
class Monkey(object):
    def __init__(self, **kwargs):
        for k,v in kwargs.items():
            setattr(self, k, [*v] if isinstance(v,list) else v)
        self.n_items_inspected = 0
        
    def catch(self, item):
        self.items += [item]
        
    def do_turn(self):
        while self.items:
            self.n_items_inspected += 1
            monkeys[self.throw_options[min((w:=self.worry_op(self.items.pop(0)) 
                                                // self.relief_factor) 
                                           % self.throw_test,
                                           1)]].catch(w)
            
monkey_params = [{'items': [int(k) for k in j[1].split(': ')[1].split(', ')],
                  'worry_op': eval(f"lambda old: {j[2].split(': ')[1].split(' = ')[1]}"),
                  'throw_test': int(j[3].split()[-1]),
                  'throw_options': [int(j[k].split()[-1]) for k in [4,5]],
                  'relief_factor': 3}
                 for i in rawinput.split('\n\n')
                 for j in [i.split('\n')]]
monkeys = [Monkey(**i) for i in monkey_params]

for i in range(20):
    for m in monkeys:
        m.do_turn()
        
(z:=sorted([m.n_items_inspected for m in monkeys]))[-2]*z[-1]
```

```python
# Part 2
from math import prod

for i in monkey_params:
    i['relief_factor'] = 1
monkeys = [Monkey(**i) for i in monkey_params]

relief_threshold = prod([m.throw_test for m in monkeys])

for i in range(10000):
    for m in monkeys:
        m.items = [i % relief_threshold for i in m.items]
        m.do_turn()

(z:=sorted([m.n_items_inspected for m in monkeys]))[-2]*z[-1]
```
