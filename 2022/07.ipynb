```python
# Get raw data
with open('input/07.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
class Folder(object):
    def __init__(self, name):
        self.name = name
        self.files = {}
        self.dirs = {}
        
    def __repr__(self):
        return self.name
        
    def run(self, instr):
        while instr:
            i = instr.pop(0)
            if i=='$ ls':
                instr = self.do_ls(instr)
            elif i != '$ cd ..':
                instr = self.dirs[i.split()[-1]].run(instr)
            else:
                return instr
            
    def do_ls(self, instr):
        while instr and not instr[0].startswith('$'):
            lsrow = instr.pop(0)
            if (z:=lsrow.split())[0]=='dir':
                path = [self.name, self.name[:-1]][self.name.endswith('/')]
                self.dirs[z[1]] = Folder('/'.join([path, z[1]]))
            else:
                self.files[z[1]] = int(z[0])
        return instr

    def get_dir_sizes(self):
        dir_sizes[self.name] = (sum(self.files.values())
                                + sum([i.get_dir_sizes()
                                       for i in self.dirs.values()]))
        return dir_sizes[self.name]

(files:=Folder('/')).run([i for i in rawinput.split('\n')[1:]])
dir_sizes = dict()
files.get_dir_sizes()
sum([i for i in dir_sizes.values() if i <= 100000])
```

```python
# Part 2
total_space = 70000000
req_space = 30000000
delete_target = dir_sizes['/'] + req_space - total_space
min([i for i in dir_sizes.values() if i>=delete_target])
```
