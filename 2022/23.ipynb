```python
# Get raw data
with open('input/23.txt', 'r') as f:
    rawinput = f.read()
```

```python
# Part 1
elf_locs = [(i,j)
             for i,r in enumerate(rawinput.split('\n'))
             for j,c in enumerate(r)
             if c == '#']
dir_order = [[(-1,-1),(-1,0),(-1,1)],
             [(1,-1),(1,0),(1,1)],
             [(-1,-1),(0,-1),(1,-1)],
             [(-1,1),(0,1),(1,1)]]

def do_iter(elf_locs, iter_num):
    elset = {*elf_locs}
    new_locs = {e: (None if len(z:=[i[1] for i in chk if len({*i}&elset)==0])==0 
                    or len({*sum(chk,[])}&elset)==0 
                    else z[0])
                for e in elf_locs
                for chk in [[[tuple([sum(k) for k in zip(e,j)]) for j in i] 
                             for i in (dir_order*2)[iter_num%4:][:4]]]}
    valid_new_locs = [j
                      for i in [[*new_locs.values()]]
                      for j in {*i}
                      if i.count(j)==1]
    new_locs = {k: v for k,v in new_locs.items() 
                if v and v in valid_new_locs}
    return [new_locs.get(e,e) for e in elf_locs] if new_locs else elf_locs

for h in range(10):
    elf_locs = do_iter(elf_locs, h)

[(z:=(z if i else 1)*(max(j)+1-min(j))) 
 for i,j in enumerate(zip(*elf_locs))][-1] - len(elf_locs)
```

```python
# Part 2
while elf_locs != (elf_locs:=do_iter(elf_locs, (h:=h+1))):  pass
h+1
```
