```python
# Get raw data
with open('input/10.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
program = [[int(k) if j else k 
            for j,k in enumerate(i.split())] 
           for i in rawinput.split('\n')]

x=1
cycle_states = [x]
for c in program:
    if c[0]=='addx':
        cycle_states += [x]*2
        x += c[1]
    else:
        cycle_states += [x]
        
sum([i*cycle_states[i] for i in range(20,len(cycle_states),40)])
```

```python
# Part 2
for i in range(1,len(cycle_states),40):
    print(''.join(['.#'[abs(k-j%40)<2] 
                   for j,k in enumerate(cycle_states[i:i+40])]))
```
