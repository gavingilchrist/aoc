```python
# Get raw data
with open('input/21.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
def parse(monkey):
    if str(mval:=monkeys[monkey]).isnumeric():
        return mval
    else:
        return eval(''.join([str(parse(i)) if i in monkeys else i 
                             for i in mval.split()]))

monkeys = dict([i.split(': ') for i in rawinput.split('\n')])
int(parse('root'))
```

```python
# Part 2
tryval = [1]
diffs = []
while True:
    monkeys['humn'] = tryval[-1]
    diffs += [parse((z:=monkeys['root'].split())[0])-parse(z[2])]
    if len(diffs)>1 and diffs[-1]*diffs[-2] < 0:
        break
    else:
        tryval += [tryval[-1]*10]
        
humn_rng = tryval[-2:]
diff_rng = diffs[-2:]

int(humn_rng[0] + (0-diff_rng[0]) / (diff_rng[1]-diff_rng[0]) * (humn_rng[1]-humn_rng[0]))
```
