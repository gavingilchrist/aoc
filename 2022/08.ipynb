```python
# Get raw data
with open('input/08.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
trees = np.array([[int(j) for j in i] for i in rawinput.split('\n')])
visible = np.pad(np.zeros_like(trees[1:-1,1:-1]), 1, constant_values=1)
for i in '1234':
    visible[1:] = np.where(trees[1:]>np.maximum.accumulate(trees[:-1], axis=0), 
                           1, 
                           visible[1:])
    trees = np.flip(trees.T, axis=1)
    visible = np.flip(visible.T, axis=1)
np.sum(visible).item()
```

```python
# Part 2
n_visible = np.ones_like(trees)
for i in '1234':
    n_visible *= np.stack([np.sum(np.cumsum(np.cumsum(trees[i+1:] >= trees[i:i+1], 
                                                      axis=0), 
                                            axis=0) < 2, 
                                  axis=0) 
                           for i in range(trees.shape[0])])
    trees = np.flip(trees.T, axis=1)
    n_visible = np.flip(n_visible.T, axis=1)
    
np.max(n_visible).item()
```
