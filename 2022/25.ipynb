```python
# Get raw data
with open('input/25.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
snafu_to_dec = lambda x: ('=-012'.index(x[-1])-2) + (5*snafu_to_dec(x[:-1])) if x else 0
dec_to_snafu = lambda x: dec_to_snafu((x-(z:=((x+2)%5))+2)//5) + ('=-012'[z]) if x else ''

dec_to_snafu(sum([snafu_to_dec(i) for i in rawinput.split('\n')]))
```
