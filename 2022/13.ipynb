```python
# Part 1
def is_inorder(pair):
    ptr=0
    while True:
        if (z:=[i for i,p in enumerate(pair) if len(p)==ptr]):
            return None if len(z)==2 else bool(1-z[0])
        itempair = [p[ptr] for p in pair]
        if any([isinstance(i,list) for i in itempair]):
            if (z:=is_inorder([i if isinstance(i,list) 
                               else [i] 
                               for i in itempair])) is not None:
                return z
        elif len({*itempair})==2:
            return itempair==sorted(itempair)
        ptr += 1

pairs = [[eval(j) for j in i.split('\n')] for i in rawinput.split('\n\n')]
sum([pn for pn,pair in enumerate(pairs,1) if is_inorder(pair)])
```
