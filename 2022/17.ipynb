```python
# Get raw data
with open('input/17.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
class Rock(object):
    rtypes = [{(0,0),(0,1),(0,2),(0,3)},
              {(0,1),(1,0),(1,1),(1,2),(2,1)},
              {(0,0),(0,1),(0,2),(1,2),(2,2)},
              {(0,0),(1,0),(2,0),(3,0)},
              {(0,0),(0,1),(1,0),(1,1)}]
    gasjets = list(rawinput)
    
    def __init__(self, rtype, chamber, jet_ptr=0):
        self.chamber = chamber
        pos = (max([*zip(*self.chamber)][0])+4, 2)
        self.pixels = {tuple([sum(j) for j in zip(i,pos)]) 
                       for i in self.rtypes[rtype]}
        self.jet_ptr = jet_ptr
        
    def drop(self):
        while True:
            self.apply_move((0,[-1,1][self.gasjets[self.jet_ptr]=='>']))
            self.jet_ptr = (self.jet_ptr+1) % len(self.gasjets)
            if self.apply_move((-1,0)):
                self.chamber.update(self.pixels)
                return self.jet_ptr
            
    def apply_move(self, dir_vec):
        pix_new = {tuple([sum(j) for j in zip(i, dir_vec)]) 
                   for i in self.pixels}
        if (all([0<=i[1]<7 for i in pix_new]) 
            and len(self.chamber & pix_new)==0):
            self.pixels = pix_new
            return False
        else:
            return True

chamber = {(0,i) for i in range(7)}
_ = [(p:=Rock(i%5,chamber,p if i else 0).drop()) for i in range(2022)]
max([*zip(*chamber)][0])
```
