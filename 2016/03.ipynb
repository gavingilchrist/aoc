```python
# Get raw data
with open('input/03.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum([sum(k[:2]) > k[2] for i in rawinput.split('\n') for k in[sorted([int(j) for j in i.split()])]])
```

```python
# Part 2
sum([sum(m[:2]) > m[2]
     for i in range(0, len(rawinput.split('\n')), 3)
     for j in[rawinput.split('\n')[i:i+3]]
     for k in range(3)
     for m in[sorted([int(l.split()[k]) for l in j])]])
```
