```python
# Get raw data
with open('input/10.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
class Bot(object):
    def __init__(self, low_to, high_to):
        self.low_to = low_to
        self.high_to = high_to
        self.values = list()
        self.n_values = 0
        
    def assign(self, value):
        if value not in self.values:
            self.values = sorted(self.values + [value])
            self.n_values += 1
        
    def maybe_pass(self):
        if self.n_values == 2:
            if self.low_to[0] == 'bot':
                bots[self.low_to[1]].assign(self.values[0])
            else:
                outputs[self.low_to[1]] = self.values[0]
            if self.high_to[0] == 'bot':
                bots[self.high_to[1]].assign(self.values[1])
            else:
                outputs[self.high_to[1]] = self.values[1]

bots = dict()
outputs = dict()
for i in rawinput.split('\n'):
    if i.startswith('bot'):
        j = i.split()
        bots[int(j[1])] = Bot((j[5], int(j[6])), (j[10], int(j[11])))
for i in rawinput.split('\n'):
    if i.startswith('value'):
        j = i.split()
        bots[int(j[5])].assign(int(j[1]))

b = 0
while bots[b].values != [17, 61]:
    bots[b].maybe_pass()
    b = (b+1) % len(bots)
b
```

```python
# Part 2
while not {0,1,2}.issubset({*outputs.keys()}):
    bots[b].maybe_pass()
    b = (b+1) % len(bots)
outputs[0] * outputs[1] * outputs[2]
```
