```python
# Get raw data
with open('input/08.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
instr = [['rect']+[int(k) for k in j[1].split('x')] 
         if j[0] == 'rect' 
         else [''.join(j[:2]), int(j[2][2:]), int(j[4])]
         for i in rawinput.split('\n') for j in[i.split()]]

screen = np.zeros((6, 50), dtype=int)
for i,j,k in instr:
    if i == 'rect':
        screen[:k, :j] = 1
    elif i == 'rotaterow':
        screen[j] = np.roll(screen[j], k)
    elif i == 'rotatecolumn':
        screen[:, j] = np.roll(screen[:, j], k)
np.sum(screen).item()
```

```python
# Part 2
import matplotlib.pyplot as plt
plt.imshow(screen, cmap='gray_r')
plt.show()
```
