```python
# Get raw data
with open('input/4.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
instr = [['-'.join(k[:-1]), int(k[-1]), j[1][:-1]] 
         for i in rawinput.split('\n') 
         for j in [i.split('[')] 
         for k in [j[0].split('-')]]
sum([sec_id
     for name, sec_id, checksum in instr
     if ''.join([*zip(*sorted([[-name.count(i), i] for i in {*name}-{'-'}]))][1][:5]) == checksum])
```

```python
# Part 2
for name, sec_id, checksum in instr:
    if ''.join([[chr(97 + (ord(i)-97+sec_id) % 26), ' '][i=='-'] for i in name]) == 'northpole object storage':
        print(sec_id)
```
