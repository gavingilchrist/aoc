```python
# Get raw data
with open('input/12.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
conv = lambda x: int(x) if x.isnumeric() else -int(x[1:]) if x[0]=='-' and x[1:].isnumeric() else x
instr = [(i.split()[0], 
          conv(i.split()[1]),
          None if len(i.split())<3 else conv(i.split()[2]))  
         for i in rawinput.split('\n')]

def run(registers):
    ptr = 0
    while ptr < len(instr):
        i, p1, p2 = instr[ptr]
        if i=='cpy':
            registers[p2] = registers[p1] if isinstance(p1, str) else p1
            ptr += 1
        elif i=='inc':  
            registers[p1] += 1
            ptr += 1
        elif i=='dec':  
            registers[p1] -= 1
            ptr += 1
        elif i=='jnz':
            if (registers[p1] if isinstance(p1, str) else p1) != 0:
                ptr += p2
            else:
                ptr += 1
    return registers['a']

run({i:0 for i in 'abcd'})
```

```python
# Part 2
run({i:int(i=='c') for i in 'abcd'})
```
