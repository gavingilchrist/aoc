```python
# Get raw data
with open('input/6.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
''.join([(z:={j.count(k):k for k in {*j}})[max(z)] for j in zip(*[list(i) for i in rawinput.split('\n')])])
```

```python
# Part 2
''.join([(z:={j.count(k):k for k in {*j}})[min(z)] for j in zip(*[list(i) for i in rawinput.split('\n')])])
```
