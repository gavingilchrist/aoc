```python
# Get raw data
with open('input/18.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
next_row = lambda x: [['.','^'][len({*i})-1] for i in zip(['.', *x[:-1]], [*x[1:], '.'])]
for i in range(40):
    if i==0:
        n_safe = (row:=rawinput).count('.')
    else:
        n_safe += (row:=next_row(row)).count('.')
n_safe
```
