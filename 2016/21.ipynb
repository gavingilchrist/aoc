```python
# Get raw data
with open('input/21.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
instr = [[j[0], 
          (z:=j[1+int(j[0]!='rotate')]), 
          j[{'rotate': 2, 'reverse': 4, 'swap': 5, 'move': 5}[j[0]]+(4*(z=='based'))]] 
         for i in rawinput.split('\n') for j in[i.split()]]

password = 'abcdefgh'

for i,j,k in instr:
    if i=='swap':
        j,k = sorted([int(j) if j.isnumeric() else password.index(j),
                    int(k) if k.isnumeric() else password.index(k)])
        password = password[:j]+password[k]+password[j+1:k]+password[j]+password[k+1:]
    elif i=='rotate':
        if j=='left':
            password = (password*3)[int(k):][:8]
        elif j=='right':
            password = (password*3)[16-int(k):][:8]
        else:
            password = (password*3)[16-(z:=password.index(k))-1-(z>=4):][:8]
    elif i=='move':
        password = (z:=password[:(j:=int(j))]+password[j+1:])[:(k:=int(k))]+password[j]+z[k:]
    elif i=='reverse':
        password = password[:(j:=int(j))]+password[j:(k:=int(k)+1)][::-1]+password[k:]

print(password)
```

```python
# Part 2
password = 'fbgdceah'

for i,j,k in instr[::-1]:
    if i=='swap':
        j,k = sorted([int(j) if j.isnumeric() else password.index(j),
                    int(k) if k.isnumeric() else password.index(k)])
        password = password[:j]+password[k]+password[j+1:k]+password[j]+password[k+1:]
    elif i=='rotate':
        if j=='right':
            password = (password*3)[int(k):][:8]
        elif j=='left':
            password = (password*3)[16-int(k):][:8]
        else:
            password = [z for x,y in enumerate([1,2,3,4,6,7,8,9]) if (z:=(password*3)[int(y):][:8])[x]==k][0]
    elif i=='move':
        password = (z:=password[:(k:=int(k))]+password[k+1:])[:(j:=int(j))]+password[k]+z[j:]
    elif i=='reverse':
        password = password[:(j:=int(j))]+password[j:(k:=int(k)+1)][::-1]+password[k:]

print(password)
```
