```python
# Get raw data
rawinput = 'cxdnnyjw'
```

```python
import hashlib
```

```python
# Part 1
passwd = ''
ix = 0
while len(passwd) < 8:
    md5_hash = hashlib.md5()
    md5_hash.update(f'{rawinput}{ix}'.encode('utf-8'))
    if (z:=md5_hash.hexdigest())[:5] == '00000':
        passwd += z[5]
    ix += 1
passwd
```

```python
# Part 2
passwd = '--------'
ix = 0
while '-' in passwd:
    md5_hash = hashlib.md5()
    md5_hash.update(f'{rawinput}{ix}'.encode('utf-8'))
    if (z:=md5_hash.hexdigest())[:5] == '00000' and z[5] in list('01234567') and passwd[int(z[5])] == '-':
        passwd = passwd[:int(z[5])] + z[6] + passwd[int(z[5])+1:]
    ix += 1
passwd
```
