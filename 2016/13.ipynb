```python
import numpy as np
```

```python
# Part 1
target = (39,31)
nsteps = -np.ones(target, dtype=int)
nsteps[1, 1] = 0

steps_to_target = -1

mapwalls = np.vectorize(lambda x: ['.', '#'][sum([i=='1' for i in bin(x)]) % 2])

def do_step(nsteps):
    return np.where((office == '.') 
                    & (nsteps == -1) 
                    & ((z:=np.max(np.stack(((y:=np.pad(nsteps, 
                                                       1, 
                                                       constant_values=-1))[:-2,1:-1], 
                                            y[2:,1:-1], 
                                            y[1:-1,:-2], 
                                            y[1:-1,2:])), 
                                  axis=0)) >= 0),
                    z+1,
                    nsteps)

while steps_to_target < 0:
    nsteps = np.pad(nsteps, ((0,1),(0,1)), constant_values=-1)
    office = mapwalls((x:=np.expand_dims(np.arange(nsteps.shape[1]), 0))**2 
                    + (y:=np.expand_dims(np.arange(nsteps.shape[0]), 1))**2 
                    + 3*x + 2*x*y + y + rawinput)
    while np.any(nsteps != (nsteps:=do_step(nsteps))):
        if (steps_to_target:=nsteps[*target]) >= 0:  break

steps_to_target.item()
```
