```python
# Get raw data
with open('input/9.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
def decompress(x):
    if '(' in x:
        lb, rb = x.index('('), x.index(')')
        nc, rept = [int(i) for i in x[lb+1:rb].split('x')]
        return x[:lb] + (x[rb+1:rb+1+nc] * rept) + decompress(x[rb+1+nc:])
    else:
        return x

len(decompress(rawinput))
```

```python
# Part 2
def decomp_length(x):
    if '(' in x:
        lb, rb = x.index('('), x.index(')')
        nc, rept = [int(i) for i in x[lb+1:rb].split('x')]
        return sum([decomp_length(x[:lb]),
                    decomp_length(x[rb+1:rb+1+nc]) * rept,
                    decomp_length(x[rb+1+nc:])])
    else:
        return len(x)

decomp_length(rawinput)
```
