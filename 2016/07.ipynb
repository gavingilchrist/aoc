```python
# Get raw data
with open('input/07.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
bracketed = lambda x,y=0: str(z:=[y,y+1][x[0]=='[']) + bracketed(x[1:], [z,z-1][x[0]==']']) if x else ''
sum([{b for t,u,v,w,b in zip(*[list(i[j:][:len(i)-3]) for j in range(4)], bracketed(i)) if t==w!=u==v} == {'0'}
     for i in rawinput.split('\n')])
```

```python
# Part 2
sum([any([z.count(k)==2 
          for z in [[w[0] 
                     for w in {([t+u+t, u+t+u][int(b)], b) 
                               for t,u,v,b in zip(*[list(i[j:][:len(i)-2]) for j in range(3)], 
                                                  bracketed(i)) 
                               if t==v!=u}]]
          for k in {*z}])
     for i in rawinput.split('\n')])
```
