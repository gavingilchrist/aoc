```python
# Get raw data
with open('input/24.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
ductmap = np.array([list(i) for i in rawinput.split('\n')])
n_pts = np.unique(ductmap).shape[0]-2

def do_step(stepmap):
    return np.pad(np.where((ductmap[1:-1,1:-1] != '#')
                           & (stepmap[1:-1,1:-1] == -1)
                           & ((stepmap[:-2,1:-1] != -1) 
                              | (stepmap[2:,1:-1] != -1) 
                              | (stepmap[1:-1,:-2] != -1) 
                              | (stepmap[1:-1,2:] != -1)),
                           np.max(stepmap)+1,
                           stepmap[1:-1,1:-1]), 1, constant_values=-1)
    
def get_nsteps(p1, p2):
    stepmap = np.where(ductmap == str(p1), 0, -1)
    while (nsteps:=np.max(np.where(ductmap == str(p2), stepmap, -1)).item()) < 0:
        stepmap = do_step(stepmap)
    return nsteps

distances = {k:s 
             for p1 in range(n_pts-1) 
             for p2 in range(p1+1, n_pts) 
             for s in [get_nsteps(p1, p2)] 
             for k in [(p1,p2),(p2,p1)]}

def min_tot_dist(pts, fr=0):
    if pts:
        return min([v+min_tot_dist(pts-{k[1]}, k[1]) 
                    for k,v in distances.items() 
                    if k[0]==fr and k[1] in pts])
    else:
        return 0

min_tot_dist({*range(n_pts)}-{0}, 0)
```

```python
# Part 2
def min_tot_dist(pts, fr=0):
    if pts:
        return min([v+min_tot_dist(pts-{k[1]}, k[1]) 
                    for k,v in distances.items() 
                    if k[0]==fr and k[1] in pts])
    else:
        return distances[(fr, 0)]

min_tot_dist({*range(n_pts)}-{0}, 0)
```
