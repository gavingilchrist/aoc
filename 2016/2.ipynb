```python
# Get raw data
with open('input/2.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
dirs = {'U': (0, -1), 'R': (1, 1), 'D': (0, 1), 'L': (1, -1)}
coords = [1, 1]
code = ''
for line in rawinput.split('\n'):
    for step in line:
        if 0 <= coords[(z:=dirs[step])[0]] + z[1] < 3:
            coords[z[0]] += z[1]
    code += str(coords[0]*3 + coords[1] + 1)
code
```

```python
# Part 2
keypad = """xx1xx
x234x
56789
xABCx
xxDxx"""
keys = {(i,j): c for i,r in enumerate(keypad.split('\n')) for j,c in enumerate(r) if c != 'x'}

coords = (2, 0)
code = ''
for line in rawinput.split('\n'):
    for step in line:
        if (z:=tuple([[j,j+(y:=dirs[step])[1]][i==y[0]] for i,j in enumerate(coords)])) in keys:
            coords = z
    code += keys[coords]
code
```
