```python
# Get raw data
rawinput = 'zpqevtbw'
```

```python
import hashlib
```

```python
# Part 1
get_md5 = lambda x: hashlib.md5(x.encode('utf-8')).hexdigest()
find_rept = lambda x,n,l=1: {*[[*y][0] for i in range(len(x)-n+1) if len(y:={*x[i:i+n]})==1][:l]}

hashes = [get_md5(f'{rawinput}{i}') for ix in[range(1000)] for i in ix]
padkeys = []
b=0

while len(padkeys) < 64:
    s3ix = [[j,l] for j,k in enumerate(hashes,b) for l in find_rept(k,3)]
    hashes += [get_md5(f'{rawinput}{i}') for ix in[range(b+1000,b+2000)] for i in ix]
    s5ix = [[j,l] for j,k in enumerate(hashes,b) for l in find_rept(k,5,None)]
    padkeys += [i for i,j in s3ix for k,l in s5ix if l==j and i<k<=i+1000]
    b += 1000
    hashes = hashes[1000:]
    
padkeys[63]
```

```python
# Part 2
get_md5_rec = lambda x,n=2017: get_md5_rec(get_md5(x),n-1) if n else x

hashes = [get_md5_rec(f'{rawinput}{i}') for ix in[range(1000)] for i in ix]
padkeys = []
b=0

while len(padkeys) < 64:
    s3ix = [[j,l] for j,k in enumerate(hashes,b) for l in find_rept(k,3)]
    hashes += [get_md5_rec(f'{rawinput}{i}') for ix in[range(b+1000,b+2000)] for i in ix]
    s5ix = [[j,l] for j,k in enumerate(hashes,b) for l in find_rept(k,5,None)]
    padkeys += [i for i,j in s3ix for k,l in s5ix if l==j and i<k<=i+1000]
    b += 1000
    hashes = hashes[1000:]
    
padkeys[63]
```
