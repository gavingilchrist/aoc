```python
# Get raw data
with open('input/01.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
dirs = [(0, -1), (1, 1), (0, 1), (1, -1)]
dir_ix = 0
coords = [0, 0]
for step in rawinput.split(','):
    turn, go = (step:=step.strip())[0], int(step[1:])
    axis, fb = dirs[dir_ix:=(dir_ix + {'L': -1, 'R': 1}[turn]) % 4]
    coords[axis] += fb * go
sum([abs(i) for i in coords])
```

```python
# Part 2
dir_ix = 0
coords = (0, 0)
visited_coords = {coords}
brk = False
for step in rawinput.split(','):
    turn, go = (step:=step.strip())[0], int(step[1:])
    axis, fb = dirs[dir_ix:=(dir_ix + {'L': -1, 'R': 1}[turn]) % 4]
    for i in range(go):
        coords = tuple([k, k+fb][j==axis] for j,k in enumerate(coords))
        if coords in visited_coords:
            brk = True
            break
        else:
            visited_coords = visited_coords | {coords}
    if brk:
        break
sum([abs(i) for i in coords])
```
