```python
# Get raw data
with open('input/20.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
blocked_ips = [[int(j) for j in i.split('-')] for i in rawinput.split('\n')]
valid_ips = [[0, 4294967295]]
for l,h in blocked_ips:
    if (ix:=[[i,ll,hh] for i,(ll,hh) in enumerate(valid_ips) if hh>=l and ll<=h]):
        valid_ips = sorted([j for i,j in enumerate(valid_ips) 
                            if i not in [*zip(*ix)][0]]
                           + ([[z, l-1]] if l>(z:=[*zip(*ix)][1][0]) else [])
                           + ([[h+1, z]] if h<(z:=[*zip(*ix)][2][-1]) else []))
valid_ips[0][0]
```

```python
# Part 2
sum([j-i+1 for i,j in valid_ips])
```
