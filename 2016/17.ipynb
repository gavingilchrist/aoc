```python
rawinput = 'gdjjyniy'
```

```python
import hashlib
```

```python
# Part 1
get_md5 = lambda x: hashlib.md5(x.encode('utf-8')).hexdigest()
def get_min_n_steps(passcode, r=0, c=0, hist=None):
    hist = hist or ''
    if r == c == 3:
        return [[len(hist), hist]]
    else:
        return sorted([m
                       for i,(j,k,l) in zip([*get_md5(f'{passcode}{hist}')[:4]], 
                                            [(-1,0,'U'),(1,0,'D'),(0,-1,'L'),(0,1,'R')])
                       if int(i,16)>10 and 0<=r+j<4 and 0<=c+k<4
                       for m in get_min_n_steps(passcode, r+j, c+k, hist+l)])[:1]

get_min_n_steps(rawinput)[0][1]
```

```python
# Part 2
def get_max_n_steps(passcode, r=0, c=0, hist=None):
    hist = hist or ''
    if r == c == 3:
        return [[len(hist), hist]]
    else:
        return sorted([m
                       for i,(j,k,l) in zip([*get_md5(f'{passcode}{hist}')[:4]], 
                                            [(-1,0,'U'),(1,0,'D'),(0,-1,'L'),(0,1,'R')])
                       if int(i,16)>10 and 0<=r+j<4 and 0<=c+k<4
                       for m in get_max_n_steps(passcode, r+j, c+k, hist+l)])[::-1][:1]

get_max_n_steps(rawinput)[0][0]
```
