```python
# Get raw data
with open('input/15.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
discs = np.array([[int(j[3]),int(j[11][:-1])] for i in rawinput.split('\n') for j in [i.split()]])
mult, add = 1, 0
for i in range(discs.shape[0]):
    ix = np.arange(100)*mult+add
    add, mult = ix[np.sum(np.mod(np.expand_dims(ix, 1) 
                                 + np.expand_dims(discs[:i+1, 1] + np.arange(1, i+2), 0), 
                                np.expand_dims(discs[:i+1, 0], 0)), axis=1) == 0][:2]
    mult -= add
add.item()
```

```python
# Part 2
discs = np.array([[int(j[3]),int(j[11][:-1])] for i in rawinput.split('\n') for j in [i.split()]]+[[11, 0]])
mult, add = 1, 0
for i in range(discs.shape[0]):
    ix = np.arange(100)*mult+add
    add, mult = ix[np.sum(np.mod(np.expand_dims(ix, 1) 
                                 + np.expand_dims(discs[:i+1, 1] + np.arange(1, i+2), 0), 
                                np.expand_dims(discs[:i+1, 0], 0)), axis=1) == 0][:2]
    mult -= add
add.item()
```
