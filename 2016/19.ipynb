```python
# Part 1
elves = np.arange(1,rawinput+1)
eo = 0
while elves.shape[0] > 1:
    elves, eo = elves[eo::2], np.mod(eo+elves.shape[0], 2)
elves[0].item()
```

```python
# Part 2
elves = np.arange(1,rawinput+1)
while elves.shape[0] > 1:
    ix = 2*np.arange(ne:=elves.shape[0]) + np.arange(ne, 0, -1) // 2
    elves = np.append(elves[(z:=np.arange(ne)[ix >= ne])[~np.isin(z, ix)]], elves[ix < ne])
elves[0].item()
```
