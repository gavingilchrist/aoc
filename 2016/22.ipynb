```python
# Get raw data
with open('input/22.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
df = [[int(k[2][1:]), int(k[1][1:]), int(j[1][:-1]), int(j[2][:-1]), int(j[3][:-1])] 
      for i in rawinput.split('\n')[2:] for j in[i.split()] for k in[j[0].split('-')]]
sum([i!=j and df[i][3] and df[i][3] <= df[j][4] for i in range(len(df)) for j in range(len(df))])
```

```python
# Part 2
pt0 = [i for i in df if i[3]==0][0][:2]
pt1 = [j for j in df if j[0] == {i[0] for i in df if i[3]>100}.pop() and j[3]<100][-1][:2]
pt2 = [0,29]
pt3 = [0,0]
sum([abs(i-j) for i,j in zip(pt0,pt1)])+sum([abs(i-j) for i,j in zip(pt1,pt2)])+5*sum([abs(i-j) for i,j in zip(pt2,pt3)])+1
```
