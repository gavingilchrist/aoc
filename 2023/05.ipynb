```python
# Get raw data
with open('input/05.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
conv = dict()
for i,j in enumerate(rawinput.split('\n\n')):
    if i:
        fr,to = (z:=j.split('\n'))[0].split()[0].split('-to-')
        conv[fr] = {to: sorted([[l[1],l[1]+l[2],l[0]-l[1]] 
                                for k in z[1:] 
                                for l in [[int(m) for m in k.split()]]])}
    else:
        seeds = [int(k) for k in j.split(':')[1].split()]

def get_location(num, numtype='seed'):
    if numtype == 'location':
        return num
    else:
        nexttype, convtab = [[k,v] for k,v in conv[numtype].items()][0]
        return get_location(num+(z[0] if (z:=[p for l,h,p in convtab 
                                              if l <= num < h]) 
                                 else 0), 
                            nexttype)

min([get_location(i) for i in seeds])
```

```python
# Part 2
def get_min_location(num_l, num_h, numtype='seed'):
    if numtype == 'location':
        return num_l
    else:
        nexttype, convtab = [[k,v] for k,v in conv[numtype].items()][0]
        return (min(z) 
                if (z:=[y
                        for j in [[i for i in convtab if i[1]>num_l and i[0]<=num_h]]
                        for t,u,v in sum([[*([[j[k-1][1],l[0],0]] if k else []), l] 
                                          for k,l in enumerate(j)],[])
                        if u>t
                        and (y:=get_min_location(max(t,num_l)+v,
                                                 min(u,num_h)+v,
                                                 nexttype)) is not None])
                else None)
        
min([get_min_location((z:=seeds[i]),z+seeds[i+1]) 
     for i in range(0,len(seeds),2)])
```
