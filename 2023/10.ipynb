```python
# Get raw data
with open('input/10.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
pipe_conn = {'|': [[-1,0],[1,0]],
             '-': [[0,-1],[0,1]],
             'J': [[-1,0],[0,-1]],
             'L': [[-1,0],[0,1]],
             '7': [[1,0],[0,-1]],
             'F': [[1,0],[0,1]]}
nswe = [[-1,0],[1,0],[0,-1],[0,1]]
nswe_conn = [[k 
              for k,v in pipe_conn.items() 
              for j in v if all([sum(l)==0 for l in zip(i,j)])] 
             for i in nswe]

pipes = {(i,j): c
         for i,r in enumerate(rawinput.split('\n'))
         for j,c in enumerate(r)
         if c != '.'}

pos_hist = [k for k,v in pipes.items() if v=='S']
start_conn = ''.join(['01'[pipes[tuple([sum(l) for l in zip(pos_hist[0],[i,j])])] in k] 
                      for (i,j),k in zip(nswe,nswe_conn)])
pipes[pos_hist[0]] = {'1100':'|','1010':'J','0110':'7','1001':'L','0101':'F','0011':'-'}[start_conn]

pos_hist += [tuple([sum(i) for i in zip(pos_hist[0],nswe[start_conn.index('1')])])]
while pos_hist[-1] != pos_hist[0]:
    pos_hist += [z
                 for i in pipe_conn[pipes[pos_hist[-1]]]
                 if (z:=tuple([sum(j) for j in zip(pos_hist[-1],i)])) != pos_hist[-2]]

(len(pos_hist)-1)//2
```

```python
# Part 2
pipe_pos_sp = {tuple(n)
               for i,j in enumerate(pos_hist[1:])
               for z in [[[l*2 for l in k] for k in [pos_hist[i],j]]]
               for n in [[sum(m)//2 for m in zip(*z)], z[-1]]}
rngs = [[min(i)-1,max(i)+2] for i in zip(*pipe_pos_sp)]

water = {(i,j) for i in range(*rngs[0]) for j in range(*rngs[1])}
air = {(i,j) for i in range(*rngs[0])[1:-1] for j in range(*rngs[1])[1:-1]}
water -= air
air -= pipe_pos_sp

while (z:={y
           for i in water
           for j in [[-1,0],[1,0],[0,-1],[0,1]]
           if (y:=tuple([sum(k) for k in zip(i,j)])) in air}):
    water.update(z)
    air -= z
sum([all([j%2==0 for j in i]) for i in air])
```
