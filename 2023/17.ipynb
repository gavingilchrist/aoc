```python
# Get raw data
with open('input/17.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
heatmap = np.array([[int(j) for j in i] for i in rawinput.split('\n')])

def get_min_heatloss(minsteps,maxsteps):
    directions = np.array([[-1,0],[0,1],[1,0],[0,-1]])
    
    pos = np.array([[0,0]])
    lastdir = np.array([[0,0]])
    heatloss = np.array([0])
    min_hl = dict()
    
    while pos.shape[0]:
        dtile = np.tile(directions, (pos.shape[0],1))
        f1 = np.all(dtile * np.repeat(lastdir, 4, axis=0) == 0, axis=1)
        
        sit = np.column_stack(((posf:=np.repeat(pos, 4, axis=0)[f1]), 
                               (dtf:=dtile[f1]), 
                               (hlf:=np.repeat(heatloss,4)[f1])))
        sortix = np.lexsort([sit[:,i] for i in range(4,-1,-1)])
        f2 = np.any(sit[sortix,:-1] 
                    != np.append(-np.ones_like(sit[:1,:-1]),
                                 sit[sortix[:-1],:-1],
                                 axis=0), 
                    axis=1)[np.argsort(sortix)]
        
        for i in sit[f2].tolist():
            if (y:=(*i[:2],int(i[2]==0))) not in min_hl or i[-1] < min_hl[y]:
                min_hl[y] = i[-1]
        f3 = np.array([min_hl[(*i[:2],int(i[2]==0))]==i[-1] for i in sit[f2]])
        
        pos = (z:=np.expand_dims(posf[f2][f3],1) 
               + (np.expand_dims(dtf[f2][f3],1) 
                  * np.expand_dims(np.arange(1,maxsteps+1),(0,2))))[:, minsteps-1:].reshape(-1,2)
        heatloss = (np.expand_dims(hlf[f2][f3],1)
                    + np.cumsum(heatmap[*[np.mod(z, heatmap.shape)[:,:,i] 
                                          for i in[0,1]]],
                                axis=1)[:, minsteps-1:]).reshape(-1)        
        
        f4 = np.all((pos >= 0) & (pos < heatmap.shape), axis=1)
        pos, lastdir, heatloss = [i[f4] 
                                  for i in [pos, 
                                            np.repeat(dtf[f2][f3], 
                                                      maxsteps-minsteps+1, 
                                                      axis=0), 
                                            heatloss]]
        
    return min([min_hl[(*[j-1 for j in heatmap.shape], i)] 
                for i in [0,1]])
    
get_min_heatloss(1,3)
```

```python
# Part 2
get_min_heatloss(4,10)
```
