```python
# Get raw data
with open('input/07.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
get_handtype = lambda x: (['11111', '2111', '221', '311', '32', '41', '5']
                          .index(''.join(sorted([str(x.count(i)) for i in {*x}])[::-1])))
ranked_hands = sorted([[get_handtype(j[0]),
                        sum([(13**i)*('23456789TJQKA'.index(j)) 
                             for i,j in enumerate(j[0][::-1])]),
                        int(j[1])] 
                       for i in rawinput.split('\n') 
                       for j in [i.split()]])
sum([(i+1)*j[2] for i,j in enumerate(ranked_hands)])
```

```python
# Part 2
get_cardposs = lambda x: [i+j 
                          for i in ([*'23456789TQKA'] if x[0]=='J' else [x[0]]) 
                          for j in get_cardposs(x[1:])] if x else ['']
ranked_hands = sorted([[max([get_handtype(i) for i in get_cardposs(j[0])]),
                        sum([(13**i)*('J23456789TQKA'.index(j)) 
                             for i,j in enumerate(j[0][::-1])]),
                        int(j[1])] 
                       for i in rawinput.split('\n') 
                       for j in [i.split()]])
sum([(i+1)*j[2] for i,j in enumerate(ranked_hands)])
```
