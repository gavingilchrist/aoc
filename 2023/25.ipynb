```python
# Part 1
def get_shortest_paths(start):
    shortest = {}
    sh_add = {(start,start): [start]}
    while sh_add:
        shortest.update(sh_add)
        sh_add = {(start,j): i+[j]
                  for i in sh_add.values()
                  for j in connections[i[-1]]
                  if (start,j) not in shortest}
    del(shortest[(start,start)])
    return shortest

# Get full list of connections for each node
connections = {j[0]: {*j[1].split()}
               for i in rawinput.split('\n') 
               for j in [i.split(': ')]}
for i in rawinput.split('\n'):
    j = i.split(': ')
    for k in j[1].split():
        if k in connections:
            connections[k].add(j[0])
        else:
            connections[k] = {j[0]}
            
# Get shortest path between each pair of nodes
shortest_paths = {k:v 
                  for i in connections 
                  for k,v in get_shortest_paths(i).items()}

# For each wire (connected pair), assign a "bottleneck score" - number of shortest 
# paths each wire appears in, weighted by the length of each path
bn_score = {tuple(sorted([k,l])):0 for k,v in connections.items() for l in v}
for i in shortest_paths.values():
    for j in zip(i[:-1],i[1:]):
        bn_score[tuple(sorted(j))] += len(i)
        
# Identify wires to cut - top 3 by bottleneck score        
cut_wires = {*[*zip(*sorted([[-v,k] for k,v in bn_score.items()])[:3])][1]}

# Amalgamate pairs to make groups
groups = {j:i for i,j in enumerate(connections)}
for i in {*bn_score} - cut_wires:
    gt,gf = [groups[j] for j in i]
    for k in groups:
        if groups[k]==gf:
            groups[k]=gt
            
# Check there are 2 groups, output product of nodes per group
assert len(gps:={*groups.values()}) == 2
prod([[*groups.values()].count(i) for i in {*groups.values()}])
```
