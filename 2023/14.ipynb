```python
# Get raw data
with open('input/14.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
rotate_r = lambda x: [''.join([*i][::-1]) for i in zip(*x)]
rotate_l = lambda x: [''.join([*i]) for i in zip(*x)][::-1]
rollup = lambda x: rotate_l(['#'.join([''.join(sorted(k)) for k in j.split('#')]) 
                             for j in rotate_r(x)])
get_load = lambda x: sum([sum([j for j,k in enumerate(i,1) if k=='O'])
                          for i in rotate_r(x)])

get_load(rollup(layout:=[i for i in rawinput.split('\n')]))
```

```python
# Part 2
cycle = lambda x,y=4: cycle(rotate_r(rollup(x)),y-1) if y else x

hist = ['|'.join(layout)]
while (z:='|'.join(layout:=cycle(rollup(layout)))) not in hist:
    hist += [z]

get_load(hist[(x:=hist.index(z)) + (1000000000-(y:=len(hist))) % (y-x)].split('|'))
```
