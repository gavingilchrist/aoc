```python
# Get raw data
with open('input/15.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
do_hash = lambda x: ((do_hash(x[:-1])+ord(x[-1]))*17)%256 if x else 0
sum([do_hash(i) for i in rawinput.split(',')])
```

```python
# Part 2
boxes = [[] for i in range(256)]
for i in rawinput.split(','):
    t,d = [[j,i.index(j)] for j in '-=' if j in i][0]
    box = do_hash(i[:d])
    if t=='-':
        boxes[box] = [j for j in boxes[box] if j[0]!=i[:d]]
    elif t=='=':
        q = [[i[:d],int(i[d+1:])]]
        boxes[box] = [q.pop() if j[0]==i[:d] else j for j in boxes[box]]+q

sum([sum([i*k*l[1] for k,l in enumerate(j,1)]) 
     for i,j in enumerate(boxes,1)])
```
