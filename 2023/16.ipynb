```python
# Get raw data
with open('input/16.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
nr,nc = len(z:=rawinput.split('\n')), len(z[0])
yx = [[[(i,j) for j in range(nc)] for i in range(nr)]]
for i in '012':
    yx += [[[*j[::-1]] for j in zip(*yx[-1])]]
yxconv = [dict(zip(sum(yx[0],[]),sum(yx[i],[]))) for i in range(4)]

mirrors = [{(i,j): c
            for i,r in enumerate(z) for j,c in enumerate(r) 
            if c!='.'}]
for i in '012':
    mirrors += [{yxconv[3][k]: {'/':'\\','\\':'/','|':'-','-':'|'}[v] 
                 for k,v in mirrors[-1].items()}]
mirrors = [{k:v for k,v in i.items() if v!='-'} for i in mirrors]


def do_beam(loc=(0,-1),orient=0):
    if (loc,orient) in hist:
        return
    else:
        hist.add((loc,orient))
    if (path:=[i for i in mirrors[orient] if i[0]==loc[0] and i[1]>loc[1]]):
        turn = min(path)
        energized.update({yxconv[orient][(loc[0],i)] 
                          for i in range(loc[1]+1,turn[1]+1)})
        if mirrors[orient][turn] in '|/':
            do_beam(yxconv[3][turn], (orient+1)%4)
        if mirrors[orient][turn] in '|\\':
            do_beam(yxconv[1][turn], (orient-1)%4)
    else:
        energized.update({yxconv[orient][(loc[0],i)] 
                          for i in range(loc[1]+1,[nc,nr][orient%2])})
        
energized = set()
hist = set()
do_beam()
len(energized)
```

```python
# Part 2
max_energized = 0
for orient in range(4):
    for row in range([nr,nc][orient%2]):
        energized = set()
        hist = set()
        do_beam((row,-1),orient)
        if len(energized) > max_energized:
            max_energized = len(energized)
max_energized
```
