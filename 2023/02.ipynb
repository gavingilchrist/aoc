```python
# Get raw data
with open('input/02.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum([i+1
     for i,j in enumerate(rawinput.split('\n'))
     for dl in [[{l.split()[1]: int(l.split()[0]) for l in k.split(', ')} 
                 for k in j.split(': ')[1].split('; ')]]
     if all([max([d.get(c,0) for d in dl])<=x 
             for c,x in [['red',12],['green',13],['blue',14]]])])
```

```python
# Part 2
sum([[(z:=(z if cn else 1) * (max(y) if (y:=[d[c] for d in dl if c in d]) else 0))
      for cn,c in enumerate(['red','green','blue'])][-1]
     for i,j in enumerate(rawinput.split('\n'))
     for dl in [[{l.split()[1]: int(l.split()[0]) for l in k.split(', ')} 
                 for k in j.split(': ')[1].split('; ')]]])
```
