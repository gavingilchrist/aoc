```python
# Get raw data
with open('input/01.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum([int(f'{j[0]}{j[-1]}')
     for i in rawinput.split('\n')
     for j in [[k for k in i if ord('0')<=ord(k)<=ord('9')]]])
```

```python
# Part 2
nums = ['zero', 'one', 'two', 'three', 'four', 'five', 'six', 'seven', 'eight', 'nine']
get_nums = lambda x: ((str(nums.index(z[0])) if (z:=[k for k in nums if x.startswith(k)]) 
                       else ['',x[0]][x[0].isnumeric()])
                      +get_nums(x[1:]) if x else '')
sum([int(f'{j[0]}{j[-1]}') 
     for i in rawinput.split('\n') 
     for j in [get_nums(i)]])
```
