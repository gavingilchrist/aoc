```python
# Get raw data
with open('input/21.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
farmmap = np.array([list(i) for i in rawinput.split('\n')])
reach = (farmmap == 'S').astype(int)
for i in range(64):
    reach = np.where((farmmap != '#')
                     & ((z:=np.max(np.stack(((y:=np.pad(reach,1,constant_values=-1))[:-2,1:-1], 
                                             y[2:,1:-1], y[1:-1,:-2], y[1:-1,2:])), 
                                   axis=0)) == 1),
                     z, 0)
np.sum(reach).item()
```

```python
# Part 2
rf = (rp:=farmmap.shape[0])//2
reach = np.pad((farmmap == 'S').astype(int), rp)
farmmap_exp = np.tile(farmmap, (3,3))
poss_cts = {(i,j): [0] for i in range(3) for j in range(3)}
for i in range(500):
    reach = np.where((farmmap_exp != '#')
                     & ((z:=np.max(np.stack(((y:=np.pad(reach,1,constant_values=-1))[:-2,1:-1], 
                                             y[2:,1:-1], y[1:-1,:-2], y[1:-1,2:])), 
                                   axis=0)) == 1),
                     z, 0)
    for i in [0,1,2]:
        for j in [0,1,2]:
            poss_cts[(i,j)] += [np.sum(reach[i*rp:(i+1)*rp,j*rp:(j+1)*rp]).item()]
    
partial_cts = {
    'N': poss_cts[(0,1)][rf:][:rp*3],
    'S': poss_cts[(2,1)][rf:][:rp*3],
    'W': poss_cts[(1,0)][rf:][:rp*3],
    'E': poss_cts[(1,2)][rf:][:rp*3],
    'NW': poss_cts[(0,0)][rf:][:rp*3],
    'NE': poss_cts[(0,2)][rf:][:rp*3],
    'SW': poss_cts[(2,0)][rf:][:rp*3],
    'SE': poss_cts[(2,2)][rf:][:rp*3],
}
full_cts = poss_cts[(1,1)][499:]

nsteps = 26501365
sum([((z:=-((rf-nsteps-1)//rp))-1)**2 * full_cts[nsteps%2],
     (z-2)**2 * full_cts[1-nsteps%2],
     *[partial_cts[i][j*rp+(nsteps-rf)%rp]
       for i in 'NSWE'
       for j in [0,1]],
     *[partial_cts[i][j*rp+(nsteps-rf)%rp]*(z-j)
       for i in ['NW','NE','SW','SE']
       for j in [0,1,2]]])
```
