```python
# Get raw data
with open('input/04.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum([int(2**(len({int(k) for k in j[0].split(':')[1].split()} 
                 & {int(k) for k in j[1].split()})-1))
     for i in rawinput.split('\n')
     for j in [i.split('|')]])
```

```python
# Part 2
def get_n_cards(cardnum):
    if cardnum not in known:
        if (z:=nmatches[cardnum]):
            known[cardnum] = sum([1+get_n_cards(c)
                                  for c in range(cardnum+1, cardnum+z+1)])            
        else:
            known[cardnum] = 0
    return known[cardnum]

nmatches = [len({int(k) for k in j[0].split(':')[1].split()} 
                & {int(k) for k in j[1].split()})
            for i in rawinput.split('\n')
            for j in [i.split('|')]]

known=dict()
sum([1+get_n_cards(i) for i in range(len(nmatches))])
```
