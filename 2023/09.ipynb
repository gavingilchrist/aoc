```python
# Get raw data
with open('input/09.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
get_next = lambda x: (x[0] if len({*x})==1 
                      else x[-1]+get_next([i-j for i,j in zip(x[1:],x[:-1])]))
sum([get_next([int(j) for j in i.split()]) for i in rawinput.split('\n')])
```

```python
# Part 2
get_prev = lambda x: (x[0] if len({*x})==1 
                      else x[0]-get_prev([i-j for i,j in zip(x[1:],x[:-1])]))
sum([get_prev([int(j) for j in i.split()]) for i in rawinput.split('\n')])
```
