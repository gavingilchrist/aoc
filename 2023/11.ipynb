```python
# Get raw data
with open('input/11.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
galaxies = [(i,j)
            for i,r in enumerate(rawinput.split('\n'))
            for j,c in enumerate(r)
            if c=='#']
conv = [dict(zip((z:=sorted((y:=[*range(min(i),max(i)+1)])+[*{*y}-{*i}])), 
                 range(len(z))))
        for i in zip(*galaxies)]
galaxies_exp = [tuple([conv[j][i[j]] for j in [0,1]]) for i in galaxies]

sum([sum([abs(l-m) for l,m in zip(*[galaxies_exp[k] for k in [i,j]])])
 for i in range(len(galaxies_exp)-1)
 for j in range(i+1,len(galaxies_exp))])
```

```python
# Part 2
conv = [dict(zip(y, [(z:=(z if j else 0)+[1,1000000][j in {*y}-{*i}])
                     for j in y]))
        for i in zip(*galaxies)
        for y in [[*range(min(i),max(i)+1)]]]
galaxies_exp = [tuple([conv[j][i[j]] for j in [0,1]]) for i in galaxies]

sum([sum([abs(l-m) for l,m in zip(*[galaxies_exp[k] for k in [i,j]])])
 for i in range(len(galaxies_exp)-1)
 for j in range(i+1,len(galaxies_exp))])
```
