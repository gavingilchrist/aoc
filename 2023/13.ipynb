```python
# Get raw data
with open('input/13.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
reflix = lambda x,n_err=0: [k for k in range(1,len(x))
                            if sum([sum([len({*n})>1 for n in zip([*l],[*m])])
                                    for l,m in zip(x[:k][::-1],x[k:])])==n_err]
flip_rc = lambda x: [''.join(i) for i in zip(*x)]

sum([(reflix(flip_rc(z:=i.split('\n'))) 
      + [j*100 for j in reflix(z)])[0]
     for i in rawinput.split('\n\n')])
```

```python
# Part 2
sum([(reflix(flip_rc(z:=i.split('\n')),1) 
      + [j*100 for j in reflix(z,1)])[0]
     for i in rawinput.split('\n\n')])
```
