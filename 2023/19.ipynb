```python
# Get raw data
with open('input/19.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
from math import prod
```

```python
# Part 1
def get_accepted(x,m,a,s,rule='in',cond=0):
    if eval((z:=workflow[rule][cond])[0]):
        return (z[1]=='A' if z[1] in ['A','R'] 
                else get_accepted(x,m,a,s,z[1],0))
    else:
        return get_accepted(x,m,a,s,rule,cond+1)

raw_wf, raw_parts = [i for i in rawinput.split('\n\n')]
workflow = {j[0]: [(['True']+l.split(':'))[-2:] for l in k]
            for i in raw_wf.split('\n') 
            for j in [i.split('{')] 
            for k in [j[1][:-1].split(',')]}
parts = [[int(j[2:]) for j in i[1:-1].split(',')] 
         for i in raw_parts.split('\n')]

sum([sum(i) for i in parts if get_accepted(*i)])
```

```python
# Part 2
def get_accepted_combos(xmas=None, rule='in', cond=0):
    if xmas is None:
        xmas = {i:(1,4001) for i in 'xmas'}
    if rule=='A':
        return prod([j-i for i,j in xmas.values()])
    elif rule=='R':
        return 0
    else:
        expr,ift = workflow[rule][cond]
        if expr=='True':
            return get_accepted_combos(xmas,ift,0)
        else:
            limits = [*xmas[expr[0]]]
            if (z:=int(expr[2:])+(expr[1]=='>')) in range(*xmas[expr[0]]):
                limits.insert(1,z)
            params = [[(nx:={k:[v,i][k==expr[0]] 
                             for k,v in xmas.items()}),
                       *((ift,0) if eval(f"{nx[expr[0]][0]}{expr[1:]}") 
                         else (rule,cond+1))]
                      for i in zip(limits[:-1],limits[1:])]
            return sum([get_accepted_combos(*p) for p in params])

get_accepted_combos()
```
