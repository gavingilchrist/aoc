```python
# Get raw data
with open('input/03.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
next_to_symbol = {(i+y,j+x)
                  for i,r in enumerate(rawinput.split('\n'))
                  for j,c in enumerate(r)
                  if c in '#$%&*+-/=@'
                  for y in [-1,0,1]
                  for x in [-1,0,1]}
sum([l 
     for l,m in zip([int(k)
                     for k in ''.join([i if i in '0123456789' else ',' 
                                       for i in rawinput.replace('\n','.')]).split(',')
                     if k],
                    ['1' in k 
                     for k in ''.join(['01'[(i,j) in next_to_symbol] if c in '0123456789' else ','
                                       for i,r in enumerate(rawinput.split('\n'))
                                       for j,c in enumerate(r)]).split(',')
                     if k])
     if m])
```

```python
# Part 2
w = len(rawinput.split('\n')[0])
ntg_cells = dict()
for i,r in enumerate(rawinput.split('\n')):
    for j,c in enumerate(r):
        if c == '*':
            for y in [-1,0,1]:
                for x in [-1,0,1]:
                    if (z:=(i+y,j+x)) not in ntg_cells:
                        ntg_cells[z] = []
                    ntg_cells[z] += [i*w+j]

gears = dict()
for l,m in zip([int(k)
                for k in ''.join([i if i in '0123456789' else ',' 
                                  for i in rawinput.replace('\n','.')]).split(',')
                if k],
               [{*sum(eval(f'[{k}]'),[])}
                for k in ''.join([f"{ntg_cells.get((i,j),[])}," if c in '0123456789' else '|'
                                  for i,r in enumerate(rawinput.split('\n'))
                                  for j,c in enumerate(r)]).split('|')
                if k]):
    for n in m:
        if n not in gears:
            gears[n] = []
        gears[n] += [l]
        
sum([i[0]*i[1] for i in gears.values() if len(i)==2])
```
