```python
# Get raw data
with open('input/08.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
directions, rawmap = [i for i in rawinput.split('\n\n')]
turns = {j[0]: tuple(k) 
         for i in rawmap.split('\n') 
         for j in [i.split(' = ')] 
         for k in [j[1][1:-1].split(', ')]}

loc = 'AAA'
ptr = 0
nsteps = 0
while loc != 'ZZZ':
    loc = turns[loc][directions[ptr]=='R']
    ptr = (ptr+1)%len(directions)
    nsteps += 1
nsteps
```

```python
# Part 2
from math import lcm

def get_nsteps(loc):
    ptr = 0
    nsteps = 0
    while loc[-1] != 'Z':
        loc = turns[loc][directions[ptr]=='R']
        ptr = (ptr+1)%len(directions)
        nsteps += 1
    return nsteps

lcm(*[get_nsteps(i) for i in turns if i.endswith('A')])
```
