```python
# Get raw data
with open('input/12.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
def get_poss(springs,groups,x=None,y=None):
    x = x or len(springs)
    y = y or len(groups)
    if (dkey:=(x,y)) not in known:
        rlim = x-sum(groups[-y:])-y+1
        if rlim < 0:
            known[dkey] = 0
        elif y==1:
            known[dkey] = len([q
                               for i in range(rlim+1)
                               for q in [('.'*i)+('#'*groups[-y])+('.'*(rlim-i))]
                               if all([{*l}!={*'.#'} for l in zip([*q],springs[-x:])])])
        else:
            known[dkey] = sum([get_poss(springs,groups,x-len(q),y-1)
                               for i in range(rlim+1)
                               for q in [('.'*i)+('#'*groups[-y])+'.']
                               if all([{*l}!={*'.#'} for l in zip(q, springs[-x:])])])
    return known[dkey]

tot_n_poss = 0
for i in rawinput.split('\n'):
    for j in [i.split()]:
        known = dict()
        tot_n_poss += get_poss(j[0], [int(k) for k in j[1].split(',')])
tot_n_poss
```
