```python
# Get raw data
with open('input/24.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
dir_ix = {'nw': (-1,0), 'ne': (-1,1), 'w': (0,-1), 'e': (0,1), 'sw': (1,-1), 'se': (1,0)}
get_directions = lambda x: [dir_ix[(z:=[i for i in dir_ix 
                                        if x.startswith(i)][0])],
                            *get_directions(x[len(z):])] if x else []
directions = [get_directions(i) for i in rawinput.split('\n')]

black_tiles = set()
for i in rawinput.split('\n'):
    t = [*zip(*[[(z:=(z if k else 0)+l) for k,l in enumerate(j)]
                  for j in zip(*get_directions(i))])][-1]
    black_tiles.remove(t) if t in black_tiles else black_tiles.add(t)

len(black_tiles)
```

```python
# Part 2
for h in range(100):
    black_tiles = {i 
                   for i in {tuple([sum(k) for k in zip(i,dir_ix[j])])
                             for i in black_tiles
                             for j in dir_ix}
                   for z in [sum([tuple([sum(k) for k in zip(i,dir_ix[j])]) in black_tiles 
                                  for j in dir_ix])]
                   if [z==2,z in [1,2]][i in black_tiles]}
len(black_tiles)
```
