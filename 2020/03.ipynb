```python
# Get raw data
with open('input/03.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
slope = {i: [j for j,c in enumerate(r) if c=='#'] for i,r in enumerate(rawinput.split('\n'))}
w = rawinput.index('\n')
sum([i*3%w in slope[i] for i in range(max(slope))])
```

```python
# Part 2
treecount = lambda x,y: sum([i*x//y%w in slope[i] for i in range(0,max(slope)+1,y)])
[(z:=z*l if i else l)
 for i,(j,k) in enumerate([[1,1],[3,1],[5,1],[7,1],[1,2]]) 
 for l in [treecount(j,k)]][-1]
```
