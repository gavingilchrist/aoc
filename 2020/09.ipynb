```python
# Get raw data
with open('input/09.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
from collections import deque
```

```python
# Part 1
numbers = [int(i) for i in rawinput.split('\n')]
dq = deque([], maxlen=25)
while len(dq)<25 or any([numbers[0]-i in {*dq}-{i} for i in dq]):
    dq.append(numbers.pop(0))
(p1:=numbers[0])
```

```python
# Part 2
numbers = [int(i) for i in rawinput.split('\n')]
for ptr in range(len(numbers)):
    if p1 in (z:=[(y:=y+j if i else j) 
                  for i,j in enumerate(numbers[ptr:])]):
        break
sum(sorted(numbers[ptr:][:(x:=z.index(p1))+1])[::x])
```
