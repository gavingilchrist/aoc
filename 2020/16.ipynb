```python
# Get raw data
with open('input/16.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
rawinput_split = rawinput.split('\n\n')
rules = [[tuple([l+int(m) 
                 for l,m in enumerate(k.split('-'))]) 
          for k in j] 
         for i in rawinput_split[0].split('\n')
         for j in [i.split(': ')[1].split(' or ')]]
nearby_tickets = [[int(j) for j in i.split(',')] 
                  for i in rawinput_split[2].split('\n')[1:]]

sum([f 
     for t in nearby_tickets 
     for f in t
     if not any([any([j<=f<k for j,k in r]) 
                 for r in rules])])
```

```python
# Part 2
valid_tickets = [t 
                 for t in nearby_tickets 
                 if all([any([any([j<=f<k for j,k in r]) 
                              for r in rules]) 
                         for f in t])]

poss = [{*range(len(rules))} for i in rules]
for t in valid_tickets:
    for fn,f in enumerate(t):
        for rn,r in enumerate(rules):
            if not any([j<=f<k for j,k in r]):
                poss[fn] -= {rn}
while max(z:=[len(i) for i in poss]) > 1:
    for k in [i for i,j in enumerate(poss) if len(j)==1]:
        poss = [[j-poss[k], j][i==k] for i,j in enumerate(poss)]
        
[(z:=z*k if j else k)
 for j,k in sorted([*zip([[*i][0] for i in poss],
                         [int(i) for i in rawinput_split[1].split('\n')[1].split(',')])])[:6]][-1]
```
