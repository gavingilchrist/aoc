```python
# Get raw data
with open('input/25.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
public_keys = [int(i) for i in rawinput.split('\n')]
loop_sizes = [None for i in public_keys]

value = 1
ls = 0
while not all(loop_sizes):
    value = (value * 7) % 20201227
    ls += 1
    for i in [0,1]:
        if loop_sizes[i] is None and value == public_keys[i]:
            loop_sizes[i] = ls

value = 1
for i in range(loop_sizes[0]):
    value = (value * public_keys[1]) % 20201227
    
value
```
