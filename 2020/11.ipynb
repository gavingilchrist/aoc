```python
# Get raw data
with open('input/11.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
seats = {(i,j): 0 
         for i,r in enumerate(rawinput.split('\n')) 
         for j,c in enumerate(r) if c=='L'}

def evolve(seats):
    return {(i,j): (1 if (z:=sum([seats.get((i+k,j+l), 0) 
                                  for k in [-1,0,1] for l in [-1,0,1] 
                                  if k or l])) == 0 
                         and seats[(i,j)] == 0 
                    else 0 if z >= 4 and seats[(i,j)] == 1
                    else seats[(i,j)])
            for i,j in seats}
    
while seats != (seats:=evolve(seats)):  pass

sum(seats.values())
```

```python
# Part 2
seats = {(i,j): 0 
         for i,r in enumerate(rawinput.split('\n')) 
         for j,c in enumerate(r) if c=='L'}
nr,nc = [max(i)+1 for i in zip(*seats)]
nextseat = {(i,j): [z[0]
                    for k in [0,1,2] for l in [0,1,2]
                    if (k*l)%2==0 
                    and (z:=[m for m in zip(yd[k], xd[l]) if m in seats])]
            for i,j in seats
            for yd,xd in [[[range(i-1,-1,-1), [i]*nr, range(i+1,nr)],
                           [range(j-1,-1,-1), [j]*nc, range(j+1,nc)]]]}

def evolve(seats):
    return {(i): (1 if (z:=sum([seats[j] for j in nextseat[i]]))==0 and seats[i]==0
                  else 0 if z >= 5 and seats[i]==1
                  else seats[i])
            for i in seats}
    
while seats != (seats:=evolve(seats)):  pass

sum(seats.values())
```
