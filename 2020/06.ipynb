```python
# Get raw data
with open('input/06.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum([len({*i.replace('\n','')}) for i in rawinput.split('\n\n')])
```

```python
# Part 2
sum([len([(z:=z&{*k} if j else {*k}) 
          for j,k in enumerate(i.split('\n'))][-1])
     for i in rawinput.split('\n\n')])
```
