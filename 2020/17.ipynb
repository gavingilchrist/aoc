```python
# Get raw data
with open('input/17.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
get_startmap = lambda ndims: {tuple([*[0]*(ndims-2),i,j]): 1 
                              for i,r in enumerate(rawinput.split('\n')) 
                              for j,c in enumerate(r) 
                              if c=='#'}
    
get_combos = lambda x: [tuple([i,*j]) 
                        for i in range(*x[0]) 
                        for j in get_combos(x[1:])] if x else [tuple([])]

def evolve(cubes, steps):
    for h in range(steps):
        cubes.update({i: int(z)
                      for i in get_combos([(min(i)-1,max(i)+2) 
                                           for i in zip(*cubes)])
                      for j in [sum([cubes.get(tuple([sum(l) 
                                                      for l in zip(i,k)]),0) 
                                     for k in get_combos([(-1,2)]*len([*cubes][0]))
                                     if {*k}-{0}])]                  
                      if any([(y:=cubes.get(i,0))==1,(z:=[j==3,j in (2,3)][y])])})
    return cubes
    
sum(evolve(get_startmap(3), 6).values())
```
