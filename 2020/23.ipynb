```python
# Part 1
class Cup(object):
    def __init__(self, label):
        self.label = label
        self.next_cup = None
            
    def __repr__(self):
        return f'Cup {self.label}'
    
    def setnext(self, next_cup):
        self.next_cup = next_cup
    
    def getnext(self, n):
        return [(z:=(z if i else self).next_cup) 
                for i in range(n)]
    
def play(cup):
    pickup = cup.getnext(3)
    cup.next_cup = pickup[-1].next_cup
    dest = cups[cup.label-1 or len(cups)]
    while dest in pickup:
        dest = cups[dest.label-1 or len(cups)]
    pickup[-1].next_cup = dest.next_cup
    dest.next_cup = pickup[0]
    return cup.next_cup
            
labels = [int(i) for i in rawinput]
cups = {i: Cup(i) for i in labels}
for i,j in enumerate(labels):
    cups[j].setnext(cups[labels[(i+1)%len(labels)]])
    
_ = [(z:=play(z if i else cups[labels[0]])) 
     for i in range(100)]

print(''.join([str(i.label) for i in cups[1].getnext(len(cups)-1)]))
```
