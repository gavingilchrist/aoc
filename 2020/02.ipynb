```python
# Get raw data
with open('input/02.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum([int(l[0]) <= j[1].count(k[1]) <= int(l[1])
     for i in rawinput.split('\n') 
     for j in [i.split(':')] 
     for k in [j[0].split()] 
     for l in [k[0].split('-')]])
```

```python
# Part 2
sum([sum([j[1][int(m)]==k[1] for m in l]) == 1
     for i in rawinput.split('\n') 
     for j in [i.split(':')] 
     for k in [j[0].split()] 
     for l in [k[0].split('-')]])
```
