```python
# Get raw data
with open('input/04.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum([{j.split(':')[0] for j in i.replace('\n',' ').split()}
     .issuperset({'byr', 'ecl', 'eyr', 'hcl', 'hgt', 'iyr', 'pid'})
     for i in rawinput.split('\n\n')])
```

```python
# Part 2
check = {
    'byr': lambda x: len(x)==4 and 1920 <= int(x) <= 2002,
    'iyr': lambda x: len(x)==4 and 2010 <= int(x) <= 2020,
    'eyr': lambda x: len(x)==4 and 2020 <= int(x) <= 2030,
    'hgt': lambda x: x[:-2].isnumeric() and ((x[-2:]=='cm' and 150 <= int(x[:-2]) <= 193) 
                                             or (x[-2:]=='in' and 59 <= int(x[:-2]) <= 76)),
    'hcl': lambda x: len(x)==7 and x.startswith('#') and {*x[1:]}.issubset({*'0123456789abcdef'}),
    'ecl': lambda x: x in ['amb','blu','brn','gry','grn','hzl','oth'],
    'pid': lambda x: len(x)==9 and x.isnumeric(),
}

sum([all([check[k](d.get(k,'')) for k in check])
     for i in rawinput.split('\n\n')
     for d in [{k[0]:k[1] 
                for j in i.replace('\n',' ').split() 
                for k in [j.split(':')]}]])
```
