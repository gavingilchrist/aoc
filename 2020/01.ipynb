```python
# Get raw data
with open('input/01.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
expenses = [int(i) for i in rawinput.split('\n')]
[(p:=k*p if j else k) 
 for j,k in enumerate([i for i in expenses 
                       if 2020-i in expenses])][-1]
```

```python
# Part 2
[(p:=k*p if j else k) 
 for j,k in enumerate([2020-expenses[h]-expenses[i] 
                       for h in range(len(expenses)-1)
                       for i in range(h+1,len(expenses))
                       if 2020-expenses[h]-expenses[i] in expenses])][-1]
```
