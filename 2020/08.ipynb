```python
# Get raw data
with open('input/08.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
instr = [[j[0], int(j[1])] for i in rawinput.split('\n') for j in [i.split()]]

def testcode(instr):
    ptr = 0
    ptr_hist = set()
    accumulator = 0
    while 0 <= ptr < len(instr) and (f:=ptr not in ptr_hist):
        ptr_hist.add(ptr)
        i, p = instr[ptr]
        if i=='nop':
            ptr += 1
        elif i=='acc':  
            accumulator += p
            ptr += 1
        elif i=='jmp':  
            ptr += p
            
    return f, accumulator

testcode(instr)[1]
```

```python
# Part 2
for i in range(len(instr)):
    if (instr[i][0] in['jmp','nop']
        and (r:=testcode([[{'jmp':'nop', 'nop':'jmp'}[k],l] if j==i else [k,l] 
                          for j,(k,l) in enumerate(instr)]))[0]):
        break
r[1]
```
