```python
# Get raw data
rawinput = '0,3,1,6,7,5'
```

```python
# Part 1
starting = [int(i) for i in rawinput.split(',')]
hist = {}
for i in range(2020):
    if i<len(starting):
        num = starting[i]
        hist[num] = i
    else:
        pnum = num
        num= i-1-hist.get(num,i-1)
        hist[pnum] = i-1
num
```
