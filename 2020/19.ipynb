```python
# Get raw data
with open('input/19.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
str_combos = lambda x: [i+j
                        for i in get_strlist(x[0])
                        for j in str_combos(x[1:])] if x else ['']

known = dict()
def get_strlist(rulenum):
    if rulenum not in known:
        known[rulenum] = ([r[0]] if isinstance((r:=rules[rulenum])[0], str)
                          else [j
                                for i in r
                                for j in str_combos(i)])
    return known[rulenum]

raw_rules, raw_messages = [i for i in rawinput.split('\n\n')]
rules = {int(j[0]):
            [k[1:-1] if k.startswith('"')
             else [int(l) for l in k.split()] 
             for k in j[1].split(' | ')]
         for i in raw_rules.split('\n') 
         for j in [i.split(': ')]}
messages = raw_messages.split('\n')

matching1 = len({*get_strlist(0)} & {*messages})
matching1
```

```python
# Part 2
def get_rule_poss(m):
    for _,i in sorted([[-len(k),k] for k in known_r if m.startswith(k)]):
        return tuple([known_r[i], *([] if len(i)==len(m) else get_rule_poss(m[len(i):]))])
    
def test_rule_poss(rp):
    while rp:
        rp = {tuple([*i[:l], (top_rules[k],), *i[l+len(k):]])
              for i in rp
              for k in top_rules
              for l in range(len(i)-len(k)+1)
              if all([t in u for t,u in zip(k,i[l:][:len(k)])])}
        if ((0,),) in rp:
            return True
    return False

known_r = dict()
for k,v in known.items():
    if k != 0:
        for i in v:
            known_r[i] = tuple([*known_r.get(i,[]), k])
            
top_rules = {**{tuple(i):k 
                for k,v in rules.items() 
                if k in [8,11,0] for i in v},
             **{(42,8):8, (42,11,31):11}}

matching2 = sum([test_rule_poss({get_rule_poss(m)})
                 for m in sorted({*messages}-{*get_strlist(0)})])
matching1+matching2
```
