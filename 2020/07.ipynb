```python
# Get raw data
with open('input/07.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
rules = {j[0][:-5]:
            [(int(l[0]), ' '.join(l[1:]))
             for k in j[1][:-1].replace(' bags','').replace(' bag','').split(', ')
             for l in [k.split()]
             if l[0] != 'no']
         for i in rawinput.split('\n') 
         for j in [i.split(' contain ')]}

contains_shiny_gold = lambda x: any([i[1]=='shiny gold' or contains_shiny_gold(i[1])
                                     for i in rules[x]])

sum([contains_shiny_gold(i) for i in rules])
```

```python
# Part 2
total_bags = lambda x: sum([i*(1+total_bags(j)) for i,j in rules[x]])

total_bags('shiny gold')
```
