```python
# Get raw data
with open('input/10.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
joltages = sorted([int(i) for i in rawinput.split('\n')])
(z:={k:y.count(k) 
     for y in [[i-j for i,j in zip(joltages+[max(joltages)+3], [0]+joltages)]]
     for k in {*y}})[1] * z[3]
```

```python
# Part 2
known = dict()
def get_n_combos(x,y=0):
    dkey = f'{len(x)}-{y}'
    if dkey not in known:
        if x:
            known[dkey] = sum([get_n_combos(x[i+1:],x[i]) 
                               for i in range(min(3,len(x))) 
                               if x[i]-y<=3])
        else:
            known[dkey] = 1
    return known[dkey]

get_n_combos(joltages)
```
