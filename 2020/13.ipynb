```python
# Get raw data
with open('input/13.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
ts, buses = int((z:=rawinput.split('\n'))[0]), [int(i) for i in z[1].split(',') if i!='x']
sorted([[(z:=-ts%i),z*i] for i in buses])[0][1]
```

```python
# Part 2
get_ts = lambda x,y=1,z=0: get_ts(x[1:], 
                                  y*x[0][0],
                                  [i for i in range(z,y*x[0][0],y) 
                                   if (i+x[0][1])%x[0][0]==0][0]) if x else z
get_ts([[int(j),i] 
        for i,j in enumerate(rawinput.split('\n')[1].split(',')) 
        if j!='x'])
```
