```python
# Get raw data
with open('input/21.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
whittle = lambda w,x,y=None: (whittle(w, x[1:], 
                                      (y or all_ingred) & {*x[0][0]} if w in x[0][1] 
                                      else (y or all_ingred)) if x 
                              else (y or all_ingred))

food_data = [(i.split('(')[0].split(),
              i.split('(')[1][9:-1].split(', '))
             for i in rawinput.split('\n')]
all_ingred = {*sum((z:=[*zip(*food_data)])[0],[])}

allergens = {i: whittle(i,food_data) for i in {*sum(z[1],[])}}

while max([len(i) for i in allergens.values()])>1:
    for k,v in allergens.items():
        if len(v)>1:
            v -= {i for v in allergens.values() if len(v)==1 for i in v}
allergens = {k:[*v][0] for k,v in allergens.items()}

sum([i not in allergens.values() for i in sum(z[0],[])])
```

```python
# Part 2
print(','.join([*zip(*sorted([[k,v] for k,v in allergens.items()]))][1]))
```
