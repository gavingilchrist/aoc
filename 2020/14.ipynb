```python
# Get raw data
with open('input/14.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
memory = {}
mask = None
for i in rawinput.split('\n'):
    if i.startswith('mask'):
        mask = i[7:]
    else:
        k,v = int((z:=[j.strip() for j in i.split('=')])[0][4:-1]), int(z[1])
        memory[k] = int(''.join([[i,j][i=='X'] 
                                 for i,j in zip([*mask],
                                                [*f'{'0'*35}{bin(v)[2:]}'[-36:]])])
                        ,2)     

sum(memory.values())
```

```python
# Part 2
get_addresses = lambda x: [i+j
                           for i in {'0': [x[0][1]], 
                                     '1': ['1'], 
                                     'X': ['0', '1']}[x[0][0]]
                           for j in get_addresses(x[1:])] if x else ['']

memory = {}
mask = None
for i in rawinput.split('\n'):
    if i.startswith('mask'):
        mask = i[7:]
    else:
        k,v = int((z:=[j.strip() for j in i.split('=')])[0][4:-1]), int(z[1])
        kbits = [*zip([*mask],[*f'{'0'*35}{bin(k)[2:]}'[-36:]])]
        for j in get_addresses(kbits):
            memory[int(j,2)] = v

sum(memory.values())
```
