```python
# Get raw data
with open('input/22.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
def do_combat(cards):
    while all([i for i in cards]):
        cr = [i.pop(0) for i in cards]
        w = cr.index(max(cr))
        cards[w] += cr[::1-2*w]
    return w, cards[w]

cards = [[int(j) for j in i.split(':\n')[1].split('\n')] for i in rawinput.split('\n\n')]

sum([j*k for j,k in enumerate(do_combat(cards)[1][::-1],1)])
```

```python
# Part 2
def do_combat(cards):
    cardhist,w = set(),None
    while all([i for i in cards]):
        if (ch:='|'.join([','.join([str(j) for j in i]) for i in cards])) in cardhist:
            w = 0
            break
        cardhist.add(ch)
        cr = [i.pop(0) for i in cards]
        if all([len(cards[i])>=j for i,j in enumerate(cr)]):
            w = do_combat([cards[i][:j] for i,j in enumerate(cr)])[0]
        else:
            w = cr.index(max(cr))
        cards[w] += cr[::1-2*w]
    return w, cards[w]

cards = [[int(j) for j in i.split(':\n')[1].split('\n')] for i in rawinput.split('\n\n')]

sum([j*k for j,k in enumerate(do_combat(cards)[1][::-1],1)])
```
