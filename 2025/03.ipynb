```python
# Get raw data
with open('input/03.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
bbanks = [i for i in rawinput.split('\n')]
sum([max([int(i[j]+i[k]) for j in range(len(i)-1) for k in range(j+1, len(i))]) for i in bbanks])
```

```python
# Part 2
get_max = lambda batt, n=12: (y:=max({*batt[:-z]})) + get_max(batt[batt.index(y)+1:], z) if (z:=n-1) else max({*batt})
sum([int(get_max(i)) for i in bbanks])
```
