```python
# Get raw data
with open('input/01.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
np.sum(np.mod(np.cumsum([50]+[[-1,1][i[0]=='R']*int(i[1:]) for i in rawinput.split('\n')]), 100) == 0).item()
```

```python
# Part 2
np.sum(np.mod(np.cumsum([50]+[[-1,1][i[0]=='R'] for i in rawinput.split('\n') for j in range(int(i[1:]))]), 100) == 0).item()
```
