```python
# Get raw data
with open('input/08.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
boxes = np.array([[int(j) for j in i.split(',')] for i in rawinput.split('\n')])

conn_ix = np.array([[i,j] for i in range(999) for j in range(i+1,1000)])
conn_distsq = np.sum((boxes[conn_ix[:, 0]]-boxes[conn_ix[:, 1]])**2, axis=1)
closest_pairs = np.argsort(conn_distsq)[:1000]

circ_id = np.arange(1000)

for i in conn_ix[closest_pairs]:
    t,f = np.sort(circ_id[i])
    circ_id[circ_id == f] = t

np.prod(np.sort(np.unique(circ_id, return_counts=True)[1])[-3:]).item()
```

```python
# Part 2
closest_pairs_nxt = np.argsort(conn_distsq)[1000:]
ptr = 0
while np.unique(circ_id).shape[0] > 1:
    t,f = circ_id[conn_ix[closest_pairs_nxt[ptr]]]
    circ_id[circ_id == f] = t
    ptr += 1
    
np.prod(boxes[conn_ix[closest_pairs_nxt[ptr-1]], 0]).item()
```
