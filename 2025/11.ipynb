```python
# Get raw data
with open('input/11.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
dev_op = {j: k.split() for i in rawinput.split('\n') for j,k in [i.split(':')]}

known = dict()
def get_n_paths(loc):
    if loc not in known:
        if loc == 'out':
            known[loc] = 1
        else:
            known[loc] = sum([get_n_paths(i) for i in dev_op[loc]])
    return known[loc]

get_n_paths('you')
```

```python
# Part 2
known = dict()
def get_n_paths(loc, dac=False, fft=False):
    dkey = f'{loc}-{int(dac)}-{int(fft)}'
    if dkey not in known:
        if loc == 'out':
            known[dkey] = int(dac & fft)
        else:
            known[dkey] = sum([get_n_paths(i, dac or loc=='dac', fft or loc=='fft') 
                               for i in dev_op[loc]])
    return known[dkey]

get_n_paths('svr')
```
