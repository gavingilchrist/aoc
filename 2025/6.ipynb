```python
# Get raw data
with open('input/6.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum([eval(j[-1].join(j[:-1])) for j in zip(*[i.split() for i in rawinput.split('\n')])])
```

```python
# Part 2
inlines = rawinput.split('\n')

ix = [i for i,j in enumerate(inlines[-1]) if j.strip()]+[max([len(i) for i in inlines])+1]
ixrg = [[i,j] for i,j in zip(ix[:-1], ix[1:])]

sum([eval(l[-1].strip().join([''.join(n) for n in zip(*[[*m] for m in l[:-1]])]))
     for l in zip(*[[i[j:k-1] for j,k in ixrg] for i in inlines])])
```
