```python
# Get raw data
with open('input/04.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
papermap = np.pad(np.array([list(i) for i in rawinput.split('\n')]), 1, constant_values='.')
np.sum((papermap[1:-1, 1:-1] == '@') 
       & (np.sum(np.stack([papermap[i:i+papermap.shape[0]-2, j:j+papermap.shape[1]-2] == '@'
                           for i in [0,1,2] for j in [0,1,2] if i!=1 or j!=1]), 
                 axis=0) < 4)).item()
```

```python
# Part 2
def removeable(papermap):
    return np.pad(((papermap[1:-1, 1:-1] == '@') 
                   & (np.sum(np.stack([papermap[i:i+papermap.shape[0]-2, j:j+papermap.shape[1]-2] == '@'
                                       for i in [0,1,2] for j in [0,1,2] if i!=1 or j!=1]), 
                             axis=0) < 4)),
                  1,
                  constant_values=False)
    
def removeable_total(papermap):
    if np.sum(z:=removeable(papermap)):
        return (np.sum(z) + removeable_total(np.where(z, '.', papermap))).item()
    else:
        return 0

removeable_total(papermap)
```
