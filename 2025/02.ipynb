```python
# Get raw data
with open('input/02.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
checkids = [i.split('-') for i in rawinput.split(',')]
sum([z
     for pid_l, pid_h in checkids
     for i in range(len(pid_l), len(pid_h)+1) if i%2==0
     for j in range(10**(i//2-1), 10**(i//2))
     if int(pid_l) <= (z:=int(str(j)*2)) <= int(pid_h)])
```

```python
# Part 2
sum({z
     for pid_l, pid_h in checkids
     for i in range(len(pid_l), len(pid_h)+1)
     for r in range(2, i+1) if i%r==0
     for j in range(10**(i//r-1), 10**(i//r))
     if int(pid_l) <= (z:=int(str(j)*r)) <= int(pid_h)})
```
