```python
# Get raw data
with open('input/5.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sp = rawinput.index('\n\n')
fresh = [[int(j) for j in i.split('-')] for i in rawinput[:sp].split('\n')]
avail = [int(i) for i in rawinput[sp:].split('\n') if i]
sum([any([l<=i<=h for l,h in fresh]) for i in avail])
```

```python
# Part 2
fresh_ranges = []
for l,h in fresh:
    add_row = [min((z:=[*zip(*[[l,h]]
                             +[[ll,hh]
                               for ll,hh in fresh_ranges 
                               if ll<=l<=hh or ll<=h<=hh])])[0]), 
               max(z[1])]
    fresh_ranges = ([i for i in fresh_ranges if i[1]<add_row[0]] 
                    + [add_row] 
                    + [i for i in fresh_ranges if i[0]>add_row[1]])

sum([hh-ll+1 for ll,hh in fresh_ranges])
```
