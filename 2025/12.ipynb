```python
# Get raw data
with open('input/12.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
shapes = [[[y,x] for y,r in enumerate(i[j*5+1:j*5+4]) for x,c in enumerate(r) if c=='#']
          for i in [rawinput.split('\n')] for j in range(6)]
region_dim, region_shapes = [[*l] 
                             for l in zip(*[[[int(k) for k in j[0].split('x')][::-1], 
                                             [int(k) for k in j[1].split()]]
                                            for i in rawinput.split('\n') if 'x' in i
                                            for j in [i.split(':')]])]

def quick_eval(dim, req):
    region_area = dim[0]*dim[1]
    req_tot_area = sum([s*len(shapes[sn]) for sn, s in enumerate(req)])
    region_n3x3 = dim[0]//3 * dim[1]//3
    req_n3x3 = sum(req)
    return -1 if req_tot_area > region_area else 1 if region_n3x3 >= req_n3x3 else 0

quick_eval_results = [quick_eval(region_dim[i], region_shapes[i]) for i in range(len(region_dim))]
assert quick_eval_results.count(0) == 0

quick_eval_results.count(1)
```
