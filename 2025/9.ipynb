```python
# Get raw data
with open('input/9.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
redtiles = np.array([[int(j) for j in i.split(',')] for i in rawinput.split('\n')])
np.max(np.prod(np.abs(np.expand_dims(redtiles, 1) - np.expand_dims(redtiles, 0)) + 1, axis=2)).item()
```

```python
# Part 2
tile_r = np.ones((y:=np.unique(redtiles[:, 0])).shape[0]*2+1, dtype=int)
tile_r[::2] = (z:=np.concatenate(([0], y, [99999])))[1:]-z[:-1]-1
redtiles[:, 0] = np.vectorize({j.item(): i*2+1 for i,j in enumerate(y)}.get)(redtiles[:, 0])
tile_c = np.ones((y:=np.unique(redtiles[:, 1])).shape[0]*2+1, dtype=int)
tile_c[::2] = (z:=np.concatenate(([0], y, [99999])))[1:]-z[:-1]-1
redtiles[:, 1] = np.vectorize({j.item(): i*2+1 for i,j in enumerate(y)}.get)(redtiles[:, 1])

tile_area_map = np.expand_dims(tile_r, 1) * np.expand_dims(tile_c, 0)
tile_color_map = np.full(tile_area_map.shape, '-')
tile_color_map[redtiles[:, 0], redtiles[:, 1]] = '#'

for i in range(z:=redtiles.shape[0]):
    t,u,v,w=(np.sort(np.column_stack((redtiles[i], redtiles[(i+1)%z])), axis=1)+[[0,1]]).reshape(-1)
    tile_color_map[t:u,v:w] = np.where((y:=tile_color_map[t:u,v:w])=='-', 'X', y)

tile_color_map = np.pad(tile_color_map[1:-1,1:-1], 1, constant_values='.')
fill_greys = lambda x: np.pad(np.where((x[1:-1,1:-1] == '-')
                                       & ((x[:-2,1:-1] == '.') | (x[2:,1:-1] == '.') | (x[1:-1,:-2] == '.') | (x[1:-1,2:] == '.')),
                                       '.', x[1:-1,1:-1]),
                              1, constant_values='.')
while np.any(tile_color_map != (tile_color_map:=fill_greys(tile_color_map))):  pass

tile_color_map = np.where(tile_color_map == '-', 'X', tile_color_map)

max([np.sum(tile_area_map[t:u,v:w]).item()
     for i in range(redtiles.shape[0]-1)
     for j in range(i+1, redtiles.shape[0])
     for t,u,v,w in[(np.sort(redtiles[[i,j]].T, axis=1)+[[0,1]]).reshape(-1)]
     if ~np.any(tile_color_map[t:u,v:w] == '.')])
```
