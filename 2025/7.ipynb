```python
# Get raw data
with open('input/7.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
splitters = [i for i in rawinput.split('\n') if len({*i})>1]
beams = {splitters.pop(0).index('S')}
n_splits = 0
for i in splitters:
    n_splits += len(splits:=beams & {j for j,k in enumerate(i) if k=='^'})
    beams = (beams - splits) | {i+j for i in splits for j in [-1,1]}
n_splits
```

```python
# Part 2
splitters = [i for i in rawinput.split('\n') if len({*i})>1]
beams = [(splitters.pop(0).index('S'), 1)]
for i in splitters:
    spos = [j for j,k in enumerate(i) if k=='^']
    beams = [(l,k) for j,k in beams for l in ([j-1,j+1] if j in spos else [j])]
    beams = [(j, sum(z[y.index(j):][:y.count(j)])) 
             for y,z in [[*zip(*sorted(beams))]] 
             for j in sorted({*y})]
sum([*zip(*beams)][1])
```
