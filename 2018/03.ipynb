```python
# Get raw data
with open('input/03.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
claims = [[int(j[0][1:]),
           *[int(k) for k in j[2][:-1].split(',')],
           *[int(k) for k in j[3].split('x')]]
          for i in rawinput.split('\n') 
          for j in [i.split()]]

fabric = np.zeros((1000, 1000), dtype=int)
for cid, xp, yp, xd, yd in claims:
    fabric[xp:xp+xd, yp:yp+yd] += 1
    
np.sum(fabric > 1).item()
```

```python
# Part 2
[cid for cid, xp, yp, xd, yd in claims if np.all(fabric[xp:xp+xd, yp:yp+yd] == 1)][0]
```
