```python
# Get raw data
with open('input/17.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
veins = np.array([[int(o)+n
                   for l in k
                   for z in [l[2:].replace('..',',').split(',')]
                   for m in [z*(3-len(z))]
                   for n,o in enumerate(m)] 
                  for i in rawinput.split('\n') 
                  for k in [sorted([j.strip() for j in i.split(',')])[::-1]]])

min_y, min_x = np.min(veins.T.reshape(2,-1), axis=1)
veins -= [[min_y, min_y, min_x-2, min_x-2]]
spring_col = (502 - min_x).item()

max_y, max_x = np.max(veins.T.reshape(2,-1), axis=1)
claymap = np.full((max_y, max_x), '.')
for yf,yt,xf,xt in veins:
    claymap[yf:yt,xf:xt] = '#'

water = {(0, spring_col)}
settled_water = set()
claymap[*zip(*water)] = '|'

while True:
    lenpre = (len(water), len(settled_water))
    for i,j in sorted([*water]):
        if (nbr:=i+1 < claymap.shape[0]) and claymap[i+1,j] == '.':
            f = i+1
            t = claymap.shape[0] if np.all(claymap[i+1:,j] == '.') else i+np.argmax(claymap[i+1:,j] != '.')+1
            claymap[f:t, j] = '|'
            water.update({(k, j) for k in range(f,t)})
        elif nbr and claymap[i+1,j] in['#','~']:
            f = j-np.argmax(((claymap[i,:j]=='#') | (claymap[i+1,:j]=='.') | (claymap[i+1,:j]=='|'))[::-1])
            t = j+np.argmax((claymap[i,j+1:]=='#') | (claymap[i+1,j+1:]=='.') | (claymap[i+1,j+1:]=='|'))+1
            if (fc:=claymap[i,f-1]) == (tc:=claymap[i,t]) == '#':
                claymap[i, f:t] = '~'
                settled_water.update({(i, k) for k in range(f,t)})
                water -= settled_water
            else:
                f,t = f-int(fc!='#'), t+int(tc!='#')
                claymap[i, f:t] = '|'
                water.update({(i, k) for k in range(f,t)})
    if (len(water), len(settled_water)) == lenpre:
        break

len(water)+len(settled_water)
```
