```python
# Get raw data
with open('input/12.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
pots_init = list(rawinput.split('\n')[0].split()[2])
conv = {j[0]:j[2] for i in rawinput.split('\n')[2:] for j in [i.split()]}

pots = [*pots_init]
leftid = 0
for i in range(20):
    pots = ['.']*(z:=max(4-pots.index('#'),0))+pots+['.']*(4-pots[::-1].index('#'))
    pots = [conv.get(''.join(i), '.') for i in zip(*[pots[i:] for i in range(5)])]
    leftid += 2-z

sum([i for i,j in enumerate(pots, leftid) if j=='#'])
```

```python
# Part 2
pots = [*pots_init]
leftid = 0
t = 0
sumplants = lambda x,y: sum([i for i,j in enumerate(x, y) if j=='#'])
hist = [[0, sumplants(pots, leftid)]]
while True:
    t += 1
    pots_prev = [*pots]
    pots = ['.']*(z:=max(4-pots.index('#'),0))+pots+['.']*(4-pots[::-1].index('#'))
    pots = [conv.get(''.join(i), '.') for i in zip(*[pots[i:] for i in range(5)])]
    leftid += 2-z
    hist += [[t, sumplants(pots, leftid)]]
    if pots == pots_prev:
        break
    
(x1,y1),(x2,y2) = hist[-2:]
m = y2 - y1
c = y1 - m * x1

m*int(5e10) + c
```
