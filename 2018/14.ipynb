```python
rawinput = 409551
```

```python
# Part 1
scoreboard = '37'
ptrs = [0,1]
while len(scoreboard) < rawinput+10:
    scoreboard += str(sum([int(scoreboard[i]) for i in ptrs]))
    ptrs = [(i+1+int(scoreboard[i]))%len(scoreboard) for i in ptrs]
scoreboard[rawinput:][:10]
```

```python
# Part 2
plus0 = {str(i):i for i in range(10)}
plus1 = {str(i):(i+1) for i in range(10)}

scoreboard = ['37']
sblen = 2
ptr0, ptr1 = 0,1
seekstr = str(rawinput)
chunksize = 100000

while True:
    scoreboard[-1] += (z:=str(plus0[(s0:=scoreboard[ptr0//chunksize][ptr0%chunksize])]
                              +plus0[(s1:=scoreboard[ptr1//chunksize][ptr1%chunksize])]))
    if seekstr in scoreboard[-1][-10:]:
        break
    if ((sblen-1) % chunksize) > (((sblen:=sblen+len(z))-1) % chunksize):
        scoreboard = scoreboard[:-1]+[scoreboard[-1][:chunksize], scoreboard[-1][chunksize:]]
    ptr0 = (ptr0+plus1[s0]) % sblen
    ptr1 = (ptr1+plus1[s1]) % sblen

chunksize*(len(scoreboard)-1) + scoreboard[-1].index(seekstr)
```
