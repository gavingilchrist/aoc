```python
# Get raw data
with open('input/25.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
points = np.array([[int(j) for j in i.split(',')] for i in rawinput.split('\n')])
pair_ix = np.column_stack(((p1:=np.repeat(np.arange(points.shape[0]), points.shape[0])),
                           (p2:=np.tile(np.arange(points.shape[0]), points.shape[0]))))[p1<p2]
match_ix = pair_ix[np.sum(np.ptp(points[pair_ix], axis=1), axis=1) <= 3].tolist()

const = np.arange(points.shape[0])
for i in match_ix:
    t,f = np.sort(const[i])
    const[const == f] = t

np.unique(const).shape[0]
```
