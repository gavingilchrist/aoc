```python
# Get raw data
with open('input/05.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
get_type = lambda x: (ord(x) - ord('A')) % 32
get_polarity = lambda x: (ord(x) - ord('A')) // 32
destroy = lambda x: len({get_type(i) for i in x}) == 1 and len({get_polarity(i) for i in x}) == 2

def react(polymer):
    ptr = len(polymer)-2
    while ptr >= 0:
        if destroy(polymer[ptr:ptr+2]):
            polymer = polymer[:ptr]+polymer[ptr+2:]
        ptr -= 1
    return polymer
    
len(react(rawinput))
```

```python
# Part 2
min([len(react(''.join([i for i in rawinput if get_type(i) != t]))) for t in range(26)])
```
