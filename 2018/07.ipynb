```python
# Get raw data
with open('input/07.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
prereqs = dict()
for i in rawinput.split('\n'):
    prereqs[i.split()[7]] = prereqs.get(i.split()[7], set()) | {i.split()[1]}

avail = sorted({j for i in prereqs.values() for j in i if j not in prereqs})
order = []
while avail:
    order += [avail[0]]
    avail = sorted({*avail[1:]} | {k for k,v in prereqs.items() if len(v-{*order}) == 0} - {*order})
    
''.join(order)
```

```python
# Part 2
worker_task = [None]*5
worker_timeleft = [None]*5
avail = sorted({j for i in prereqs.values() for j in i if j not in prereqs})
order = []
t = 0
while avail or any(worker_timeleft):
    for i in range(5):
        if worker_task[i] is None and avail:
            worker_task[i] = avail.pop(0)
            worker_timeleft[i] = ord(worker_task[i])-4

    tsk = min([i for i in worker_timeleft if i])
    worker_timeleft = [None if i is None or i==tsk else i-tsk
                       for i in worker_timeleft]
    order += [j for i,j in zip(worker_timeleft, worker_task) if j and not i]
    worker_task = [[i,None][i in order] for i in worker_task]
    avail = sorted(({*avail} | {i for i in prereqs if len(prereqs[i]-{*order}) == 0}) 
                   - {*order} - {*worker_task})
    t += tsk
    
t
```
