```python
# Get raw data
with open('input/18.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
lca = np.array([list(i) for i in rawinput.split('\n')])
for t in range(10):
    lca = np.where((lca == '.') & 
                   (np.sum((adj:=np.stack([np.pad(lca, 1, constant_values=' ')
                                              [i:i+lca.shape[0], j:j+lca.shape[0]]
                                           for i in [0,1,2] 
                                           for j in [0,1,2] 
                                           if {i,j}!={1}]))=='|',
                           axis=0) >= 3),
                   '|',
                   np.where((lca == '|') 
                            & (np.sum(adj=='#', axis=0) >= 3),
                            '#',
                            np.where((lca == '#')
                                     & ((np.sum(adj=='#', axis=0) == 0) 
                                        | (np.sum(adj=='|', axis=0) == 0)),
                                     '.',
                                     lca)))
np.prod(np.unique(lca, return_counts=True)[1][::2]).item()
```
