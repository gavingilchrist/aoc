```python
# Get raw data
with open('input/04.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
dt = np.array([i[1:17] for i in rawinput.split('\n')], dtype=np.datetime64)
guard = np.array([int(z[1][1:]) if (z:=i[19:].split())[0]=='Guard' else -1 for i in rawinput.split('\n')])
stchg = np.array([{'wakes': -1, 'falls': 1}.get(i[19:].split()[0],0) for i in rawinput.split('\n')])

sortix = np.argsort(dt)
dt, guard, stchg = [i[sortix] for i in [dt, guard, stchg]]

guard[:] = [(z:=i if i > 0 else z) for i in guard]
minute = np.where(np.strings.slice(dt.astype(str), 11, 13) == '00',
                  np.strings.slice(dt.astype(str), 14, 16).astype(int),
                  -1)

glist = np.unique(guard)
guard_id = np.vectorize({j.item():i for i,j in enumerate(glist)}.get)
grid = np.zeros((glist.shape[0], 60), dtype=int)
for g, f, t in np.column_stack((guard_id(guard[stchg==1]), 
                                minute[stchg==1], 
                                minute[stchg==-1])):
    grid[g, f:t] += 1
    
(glist[(gix:=np.argmax(np.sum(grid, axis=1)))] * np.argmax(grid[gix])).item()
```

```python
# Part 2
(glist[np.argmax(np.any((z:=grid == np.max(grid)), axis=1))] * np.argmax(np.any(z, axis=0))).item()
```
