```python
serial_num = 9435
```

```python
import numpy as np
```

```python
# Part 1
x = np.expand_dims(np.arange(300)+1, 0)
y = np.expand_dims(np.arange(300)+1, 1)
power = np.mod(((r:=x+10)*y+serial_num)*r // 100, 10) - 5
power_sq = np.sum(np.stack([power[i:i+298,j:j+298] for i in [0,1,2] for j in [0,1,2]]), axis=0)
f'{(z:=np.argmax(power_sq.reshape(-1)).item()) % 298 + 1},{z // 298 + 1}'
```

```python
# Part 2
max_power = 0
max_power_sq = None
for n in range(300):
    power_sq = np.sum(np.stack(([power_sq[:-1, :-1]] if n else [])
                               +[power[i:i+300-n,j:j+300-n] 
                                 for i in range(n+1) 
                                 for j in range(n+1) 
                                 if i==n or j==n]), 
                      axis=0)
    if (z:=np.max(power_sq).item()) > max_power:
        max_power = z
        max_ix = np.argmax(power_sq.reshape(-1)).item()
        max_power_sq = f'{max_ix % (300-n) + 1},{max_ix // (300-n) + 1},{n+1}'
max_power, max_power_sq
```
