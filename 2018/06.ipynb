```python
# Get raw data
with open('input/06.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
def get_closest_cts(ymin, ymax, xmin, xmax):
    y = np.expand_dims(np.arange(ymin, ymax), (1,2))
    x = np.expand_dims(np.arange(xmin, xmax), (0,2))
    mdist = abs(y - np.expand_dims(locs[:, 0], (0,1))) + abs(x - np.expand_dims(locs[:, 1], (0,1)))
    closest = np.where(np.sum(mdist == np.min(mdist, axis=2, keepdims=True), axis=2) > 1, -1, np.argmin(mdist, axis=2))

    return np.column_stack(np.unique(closest, return_counts=True))

locs = np.array([[int(j) for j in i.split(',')] for i in rawinput.split('\n')])
ymin, xmin = np.min(locs, axis=0)
ymax, xmax = np.max(locs, axis=0)+1

closest_cts1 = get_closest_cts(ymin, ymax, xmin, xmax)
closest_cts2 = get_closest_cts(ymin-5, ymax+5, xmin-5, xmax+5)

np.max(closest_cts1[closest_cts1[:, 1] == closest_cts2[:, 1], 1]).item()
```

```python
# Part 2
y = np.expand_dims(np.arange(ymin, ymax), (1,2))
x = np.expand_dims(np.arange(xmin, xmax), (0,2))
mdist = abs(y - np.expand_dims(locs[:, 0], (0,1))) + abs(x - np.expand_dims(locs[:, 1], (0,1)))
np.sum(np.sum(mdist, axis=2) < 10000).item()
```
