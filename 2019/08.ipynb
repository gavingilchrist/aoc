```python
# Get raw data
with open('input/08.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Part 1
image_data = np.array([int(i) for i in rawinput]).reshape(-1,6,25)
targ_layer = np.argmin(np.sum(image_data == 0, axis=(1,2)))
np.prod(np.unique(image_data[targ_layer], return_counts=True)[1][1:]).item()
```

```python
# Part 2
img = image_data[np.sum(np.cumprod(image_data == 2, axis=0), axis=0),
                 np.expand_dims(np.arange(image_data.shape[1]), 1),
                 np.expand_dims(np.arange(image_data.shape[2]), 0)]
plt.imshow(img, cmap='gray_r')
```
