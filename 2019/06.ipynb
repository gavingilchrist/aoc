```python
# Get raw data
with open('input/06.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sat = {}
for i in rawinput.split('\n'):
    if (k:=(z:=i.split(')'))[0]) in sat:
        sat[k] += [z[1]]
    else:
        sat[k] = [z[1]]
        
orbit_level = {}
def get_orbit_levels(body, l=0):
    orbit_level[body] = l
    if body in sat:
        for i in sat[body]:
            get_orbit_levels(i, l+1)
get_orbit_levels('COM')

sum(orbit_level.values())
```

```python
# Part 2
orbit_hier = {}
def get_orbit_hier(body, l=None):
    orbit_hier[body] = [body] if l is None else l+[body]
    if body in sat:
        for i in sat[body]:
            get_orbit_hier(i, orbit_hier[body])
get_orbit_hier('COM')

ix = [i!=j for i,j in zip(orbit_hier['YOU'], orbit_hier['SAN'])].index(True)
len(orbit_hier['YOU']) + len(orbit_hier['SAN']) - 2*(ix+1)
```
