```python
# Get raw data
with open('input/14.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
def get_reactions(rawinput):
    batch = {'ORE': 1}
    inputs = {'ORE': []}
    for i in rawinput.split('\n'):
        rowsplit = [j.split() for j in i.replace('=>', ',').split(',')]
        batch[z[1]] = int((z:=rowsplit.pop())[0])
        inputs[z[1]] = [(int(k[0]), k[1]) for k in rowsplit]
    return inputs, batch

def get_chem_order(inputs):
    chem_order = ['ORE']
    queue = {*inputs} - {'ORE'}
    while queue:
        chem_order += (z:=sorted([i for i in queue 
                                  if {*[*zip(*inputs[i])][1]}.issubset({*chem_order})]))
        queue -= {*z}
    return chem_order[-2::-1]

def get_ore_req(inputs, batch, fuel_req):
    tot_req = {'FUEL': fuel_req}
    for chem in get_chem_order(inputs):
        tot_req[chem] = ((z:=sum([tot_req[k]*i 
                                for k,v in inputs.items() 
                                for i,j in v if j==chem])) 
                        + (-z)%batch[chem]) // batch[chem]
    return tot_req['ORE']

get_ore_req(*(reactions:=get_reactions(rawinput)), 1)
```

```python
# Part 2
tot_ore = 1_000_000_000_000
bs_limits = [0,1]
while get_ore_req(*reactions, (z:=bs_limits[-1])) < tot_ore:
    bs_limits = [z, z*10]
while bs_limits[1]-bs_limits[0] > 1:
    bs_limits[get_ore_req(*reactions, z) > tot_ore] = (z:=sum(bs_limits)//2)

bs_limits[0]
```
