```python
# Get raw data
with open('input/01.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 2
fuel_req = lambda x: sum(z:=[y for i in x if (y:=i//3-2) > 0])+fuel_req(z) if x else 0
fuel_req([int(i) for i in rawinput.split('\n')])
```
