```python
# Get raw data
with open('input/13.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
from intcode import Program  # From Day 9
```

```python
# Part 1
class Game(Program):
    def __init__(self, instr, verbose=False):
        super().__init__(instr, [], verbose)
        self.tiles = {}
        self.score = 0
        self.ball_x = None
        self.bat_x = None
        
    def do_output(self, addr, val):
        if self.verbose:  print(f"Ptr={self.ptr:2d}:  OUTPUT value {val[0]}")
        if len(self.output) == 2:
            output = (self.output.pop(0), self.output.pop(0), val[0])
            if output[:2] == (-1,0):
                self.score = output[2]
            else:
                self.tiles[output[:2]] = output[2]
                if self.verbose:  print(f"NEW TILE {output[:2]}: {output[2]}")
                if output[2] == 3:  self.bat_x = output[0]
                if output[2] == 4:  self.ball_x = output[0]
                if output[2] > 2 and self.ball_x is not None and self.bat_x is not None:
                    self.input_val = [min(max(self.ball_x - self.bat_x, -1), 1)]
                    if self.verbose:  print(f"JOYSTICK INPUT =  {self.input_val[0]}")
        else:
            self.output += [val[0]]
        self.ptr += 2
        
    def add_quarters(self, quarters):
        self.instr_orig[0] = quarters
        self.reset()

game = Game([int(i) for i in rawinput.split(',')])
game.do_exec()
sum([game.tiles[i]==2 for i in game.tiles])
```

```python
# Part 2
game = Game([int(i) for i in rawinput.split(',')])
game.add_quarters(2)
game.do_exec()
game.score
```
