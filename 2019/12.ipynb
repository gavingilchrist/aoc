```python
# Get raw data
with open('input/12.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
pos = np.array([[int(j.split('=')[1]) for j in i[1:-1].split(',')] 
                for i in rawinput.split('\n')])
vel = np.zeros_like(pos)

for i in range(1000):
    vel += np.sum(np.sign(np.expand_dims(pos, 0)-np.expand_dims(pos, 1)), axis=1)
    pos += vel
    
np.sum(np.sum(np.abs(pos), axis=1) * np.sum(np.abs(vel), axis=1)).item()
```

```python
# Part 2
startpos = [*zip(*[[int(j.split('=')[1]) for j in i[1:-1].split(',')]
                   for i in rawinput.split('\n')])]
rpt_after = []
for d in startpos:
    z = [*[0]*4, *d]
    hist = [tuple(z)]
    while True:
        newvals = [tuple(z:=(y:=[sum(k) 
                                 for k in zip(z[:4], 
                                              [sorted(z[4:])[::-1].index(j)
                                               -sorted(z[4:]).index(j) 
                                               for j in z[4:]])]) 
                         + [sum(k) for k in zip(z[4:], y)]) 
                   for i in range(10000)]
        if len(hist:=hist + newvals) > (ddct:=len({*hist})):
            break
    rpt_after += [ddct]

np.lcm.reduce(rpt_after).item()
```
