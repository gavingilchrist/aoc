```python
# Get raw data
with open('input/02.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
opcodes = {
    1: lambda x,y: x+y,
    2: lambda x,y: x*y,
}

base_program = [int(i) for i in rawinput.split(',')]

def get_result(noun,verb):
    program = [{1:noun,2:verb}.get(i,j) for i,j in enumerate(base_program)]
    ptr = 0
    while True:
        op, in1, in2, out = program[ptr:ptr+4]
        if op == 99:
            break
        else:
            program[out] = opcodes[op](program[in1],program[in2])
            ptr += 4
    return program[0]

get_result(12,2)
```

```python
# Part 2
for noun,verb in [[noun,verb] for noun in range(100) for verb in range(100)]:
    if get_result(noun,verb) == 19690720:
        break
100 * noun + verb
```
