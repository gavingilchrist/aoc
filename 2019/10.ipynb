```python
# Get raw data
with open('input/10.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
asteroids = [(x,y) 
             for y,r in enumerate(rawinput.split('\n')) 
             for x,c in enumerate(r) if c=='#']

def is_blocking(x,y,z):
    assert x != y
    if x[0]==y[0]:
        x,y,z = [i[::-1] for i in [x,y,z]]
    sl = (y[1]-x[1])/(y[0]-x[0])
    ic = x[1]-sl*x[0]
    return (z[1] == round(z[0]*sl+ic,6)) and (z[0] == sorted([x[0],y[0],z[0]])[1])

max(z:=[sum([sum([is_blocking(i,j,k) 
                  for k in {*asteroids}-{i}-{j}]) == 0 
             for j in {*asteroids}-{i}])
        for i in asteroids])
```

```python
# Part 2
from math import atan2, pi

laser_loc = asteroids[z.index(max(z))]
target_order = dict()
n_destroyed = 0
for i in {*asteroids}-{laser_loc}:
    bearing = round(atan2(*[(i-j)*k for i,j,k in zip(i,laser_loc,[1,-1])]) % (2*pi), 8)
    target_order[bearing] = [*[*zip(*sorted([[sum([(k-l)**2 
                                                   for k,l in zip(j,laser_loc)]) ** 0.5, 
                                              j] 
                                             for j in target_order.get(bearing, []) + [i]]))][1]]

while True:
    firstinline = [target_order[k].pop(0) for k in sorted(target_order) if target_order[k]]
    if n_destroyed + len(firstinline) > 200:
        v200 = firstinline[200-n_destroyed-1]
        break
    else:
        n_destroyed += len(firstinline)

sum([i*j for i,j in zip(v200, [100,1])])
```
