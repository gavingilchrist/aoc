```python
# Get raw data
with open('input/21.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
from intcode import Program  # From Day 9
import numpy as np
```

```python
# Part 1
class Droid(Program):
    instr = [int(i) for i in rawinput.split(',')]
    def __init__(self, inputs, verbose=False):
        super().__init__(self.instr, inputs, verbose)
        
droid = Droid([ord(i) 
               for i in 'NOT A J|NOT B T|OR T J|NOT C T|OR T J|AND D J|WALK|'
                            .replace('|','\n')])
droid.do_exec()
droid.output[-1]
```
