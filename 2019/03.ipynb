```python
# Get raw data
with open('input/03.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
wire_pts = []
for i in rawinput.split('\n'):
    wire_end = (0,0)
    wire_pts += [{wire_end}]
    for j in i.split(','):
        offset = {'U': (-1, 0), 'D': (1, 0), 'L': (0, -1), 'R': (0, 1)}[j[0]]
        wire_pts[-1].update({*(z:=[*zip(*[[wire_end[k]+offset[k]*(l+1) 
                                           for l in range(int(j[1:]))] 
                                          for k in [0,1]])])})
        wire_end = z[-1]

min([sum([abs(j) for j in i]) for i in (wire_pts[0] & wire_pts[1]) - {(0,0)}])
```

```python
# Part 2
wire_pts = []
for i in rawinput.split('\n'):
    wire_end = (0,0)
    nsteps = 0
    wire_pts += [{wire_end: 0}]
    for j in i.split(','):
        offset = {'U': (-1, 0), 'D': (1, 0), 'L': (0, -1), 'R': (0, 1)}[j[0]]
        wire_pts[-1].update([[k,v] 
                             for z in [[*zip(*[[wire_end[k]+offset[k]*(l+1) 
                                                for l in range(int(j[1:]))] 
                                               for k in [0,1]])]]
                             for k,v in zip(z, range(nsteps+1, nsteps+len(z)+1))
                             if k not in wire_pts[-1]])
        wire_end = tuple([wire_end[i]+offset[i]*int(j[1:]) for i in [0,1]])
        nsteps += int(j[1:])
        
min([sum([wire_pts[j][i] for j in [0,1]]) 
     for i in {*wire_pts[0]} & {*wire_pts[1]} - {(0,0)}])
```
