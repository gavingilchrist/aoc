```python
# Get raw data
with open('input/25.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
from intcode import Program  # From Day 9
import numpy as np
```

```python
# Part 1
class Droid(Program):
    instr = [int(i) for i in rawinput.split(',')]
    def __init__(self, command_list=None, verbose=False):
        super().__init__(self.instr, [], verbose)
        self.command_list = command_list or []
        
    def show_output(self):
        while self.output:
            spl = self.output.index(10)
            line = ''.join([chr(i) for i in self.output[:spl]])
            self.output = self.output[spl+1:]
            print(line)
        print()
        
    def do_input(self, addr, val):
        if len(self.input_val) == 0:
            self.show_output()
            if self.command_list:
                command = self.command_list.pop(0)
            else:
                command = input('>')
            if command == 'quit':
                self.ptr = self.instr.index(99)
                return
            self.input_val = [ord(i) for i in command]+[10]
            
        if self.verbose:  print(f"Ptr={self.ptr:2d}:  Set address {addr[0]} from INPUT to {self.input_val[0]}")
        self.maybe_extend_memory(addr[0])
        self.instr[addr[0]] = self.input_val.pop(0)
        self.ptr += 2
        
    def do_exec(self, reset=True):
        if reset:  self.reset()
        while True:
            opmode = self.instr[self.ptr]
            opcode = opmode % 100
            if opcode == 99:
                self.show_output()
                if self.verbose:  print("STOP")
                break
            self.ops[opcode](*self.process_params(opmode))
```

```python
command_list = ['west', 'west', 'west', 'take coin', 'east', 'east', 'east', 'north', 'north', 
                'take mutex', 'east', 'take antenna', 'west', 'south', 'east', 'take cake', 'east', 
                'north', 'take pointer', 'south', 'west', 'west', 'south', 'east', 'east', 
                'take tambourine', 'east', 'take fuel cell', 'east', 'take boulder', 'north', 'east']

chk_items = ['fuel cell', 'cake', 'pointer', 'boulder', 'mutex', 'antenna', 'tambourine', 'coin']
prev_inv = {*chk_items}
for i in range(1,256):
    inv = {j for i,j in zip(list(f'0000000{bin(i)[2:]}'[-8:]), chk_items) if i=='1'}
    command_list += [f'take {i}' for i in inv-prev_inv] + [f'drop {i}' for i in prev_inv-inv] + ['inv', 'east']
    prev_inv = {*inv}

droid = Droid(command_list)
droid.do_exec()
```
