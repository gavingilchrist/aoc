```python
# Get raw data
with open('input/20.txt', 'r') as f:
    rawinput = f.read()
```

```python
import numpy as np
```

```python
# Part 1
w = rawinput.index('\n')
x4letters = lambda x: ''.join([[i,'X'][64<ord(i)<91] for i in x])

portals = dict()
for i in range(len(ri:=rawinput.replace('\n',''))-2):
    if (z:=x4letters(y:=ri[i:i+3])) == '.XX':
        portals[l] = portals.get((l:=y[1:]),[])+[(i//w,i%w)]
    elif z == 'XX.':
        portals[l] = portals.get((l:=y[:-1]),[])+[(i//w,i%w+2)]
for i in range(len(ri:=rawinput.replace('\n',''))-(w*2)):
    if (z:=x4letters(y:=ri[i:i+3*w:w])) == '.XX':
        portals[l] = portals.get((l:=y[1:]),[])+[(i//w,i%w)]
    elif z == 'XX.':
        portals[l] = portals.get((l:=y[:-1]),[])+[(i//w+2,i%w)]
        
tp = {j[0]:j[1] for i in portals.values() if len(i)==2 for j in [i, i[::-1]]}
path_pts = {(i//w,i%w) for i,j in enumerate(rawinput.replace('\n','')) if j=='.'}

path = set()
pathlast = {portals['AA'][0]}
nsteps = 0
while portals['ZZ'][0] not in pathlast:
    path, pathlast = (path | pathlast,
                      (({l
                         for i in pathlast
                         for l in [tuple([sum(k) for k in zip(i,j)]) 
                                   for j in [[-1,0],[1,0],[0,-1],[0,1]]]
                                            +[tp.get(i)]
                         if l})
                       & path_pts) - path)
    nsteps += 1
nsteps
```

```python
# Part 2
ty, by, tx, bx = [sorted({*k})[l] 
                  for k in zip(*[j for i in portals.values() for j in i]) 
                  for l in[0,-1]]
on_outside = lambda x: x[0] in [ty,by] or x[1] in [tx,bx]

path = set()
pathlast = {tuple([*portals['AA'][0], 0])}
nsteps = 0
while tuple([*portals['ZZ'][0], 0]) not in pathlast:
    path, pathlast = (path | pathlast,
                      {l
                       for i in pathlast
                       for l in [tuple([sum(k) for k in zip(i,j)])
                                 for j in [[-1,0,0],[1,0,0],[0,-1,0],[0,1,0]]]           
                                           + [tuple([*tp.get(i[:2], (-1,-1)), i[2]+1-on_outside(i[:2])*2])]
                       if l[:2] in path_pts
                       and l not in path
                       and (l[:2] not in tp 
                            or l[2]>0 
                            or not on_outside(l[:2]) 
                            or l[:2]==portals['ZZ'][0])})
    nsteps += 1
nsteps
```
