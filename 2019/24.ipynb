```python
# Get raw data
with open('input/24.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
def evolve(scan):
    adjct = np.sum(np.stack([pscan[y:y+5, x:x+5]
                             for pscan in [np.pad(scan,1,constant_values='.')]
                             for y,x in [[0,1],[2,1],[1,0],[1,2]]]) == '#',
                   axis=0)
    return np.where(((scan=='#') & (adjct == 1))
                    | ((scan=='.') & ((adjct == 1)|(adjct == 2))),
                    '#', '.')

scan = np.array([list(i) for i in rawinput.split('\n')])
scanhist = scan.reshape(-1).view('<U25').tolist()
while True:
    scan = evolve(scan)
    if (z:=scan.reshape(-1).view('<U25').tolist())[0] in scanhist:
        break
    else:
        scanhist += z
np.sum((2**np.arange(25))[scan.reshape(-1) == '#']).item()
```

```python
# Part 2
adj_n = lambda l,y,x: [(l-1,1,2)] if y==0 else [(l+1,4,i) for i in range(5)] if (y,x)==(3,2) else [(l,y-1,x)]
adj_s = lambda l,y,x: [(l-1,3,2)] if y==4 else [(l+1,0,i) for i in range(5)] if (y,x)==(1,2) else [(l,y+1,x)]
adj_w = lambda l,y,x: [(l-1,2,1)] if x==0 else [(l+1,i,4) for i in range(5)] if (y,x)==(2,3) else [(l,y,x-1)]
adj_e = lambda l,y,x: [(l-1,2,3)] if x==4 else [(l+1,i,0) for i in range(5)] if (y,x)==(2,1) else [(l,y,x+1)]
get_adj = lambda x: adj_n(*x)+adj_s(*x)+adj_w(*x)+adj_e(*x)

bugs = [(0,i,j) for i,r in enumerate(rawinput.split('\n')) for j,c in enumerate(r) if c=='#']
for i in range(200):
    bugs = sorted({(l,i,j)
                   for z in [[*zip(*bugs)][0]]
                   for l in range(min(z)-1, max(z)+2)
                   for i in range(5)
                   for j in range(5)
                   if i!=2 or j!=2
                   for b,c in[[(l,i,j) in bugs, sum([k in bugs for k in get_adj((l,i,j))])]]
                   if (b and c==1) or (not b and c in(1,2))})

len(bugs)
```
