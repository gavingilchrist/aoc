```python
rawinput = '134792-675810'
```

```python
import numpy as np
```

```python
# Part 1
l,h = [int(i) for i in rawinput.split('-')]
poss = np.arange(l,h+1)
digits = poss.astype('<U6').view(np.uint32).reshape(poss.shape[0], -1).astype(int)-48
np.sum(np.all(digits[:, 1:] >= digits[:, :-1], axis=1) 
       & np.any(digits[:, 1:] == digits[:, :-1], axis=1)).item()
```

```python
# Part 2
np.sum(np.all(digits[:, 1:] >= digits[:, :-1], axis=1) 
       & np.any(((digits[:, :-1] 
                  != np.append(-np.ones_like(digits[:, :1]), 
                               digits[:, :-2], axis=1))
                 & (digits[:, 1:] == digits[:, :-1])
                 & (digits[:, 1:] 
                    != np.append(digits[:, 2:], 
                                 -np.ones_like(digits[:, :1]), axis=1))),
                axis=1)).item()
```
