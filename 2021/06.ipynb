```python
# Get raw data
with open('input/06.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
fish_counts = {j: z.count(j)
               for z in[[int(i) for i in rawinput.split(',')]]
               for j in range(9)}

for i in range(80):
    fish_counts = {j:(fish_counts[j+1] if j<8 else 0)
                     +(fish_counts[0] if j in[6,8] else 0)
                   for j in range(9)}

sum(fish_counts.values())
```

```python
# Part 2
for i in range(80,256):
    fish_counts = {j:(fish_counts[j+1] if j<8 else 0)
                     +(fish_counts[0] if j in[6,8] else 0)
                   for j in range(9)}

sum(fish_counts.values())
```
