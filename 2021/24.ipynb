```python
# Get raw data
with open('input/24.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
def run(instr, registers, digit):
    registers = {**registers}
    ptr = 0
    while ptr < len(instr):
        i, p1, p2 = instr[ptr]
        p2 = registers[p2] if isinstance(p2, str) else p2
        if i=='inp':
            registers[p1] = digit
        elif i=='add':
            registers[p1] += p2
        elif i=='mul':
            registers[p1] *= p2
        elif i=='div':
            registers[p1] = int(registers[p1]/p2)
        elif i=='mod':
            registers[p1] = registers[p1] % p2
        elif i=='eql':
            registers[p1] = int(registers[p1]==p2)
        ptr += 1
    return registers

def get_zero_combos(instr, registers):
    if instr:
        return sorted([[str(i)+j,k]
                       for i in range(1,10)
                       for j,k in get_zero_combos(instr[1:],
                                                  run(instr[0], registers, i))])
    elif registers['y']==0:
        return [['', registers]]
    else:
        return []
    
conv = lambda x: int(x) if x.isnumeric() else -int(x[1:]) if x[0]=='-' and x[1:].isnumeric() else x
instr_full = [[i.split()[0], 
               conv(i.split()[1]),
               None if len(i.split())<3 else conv(i.split()[2])] 
              for i in rawinput.split('\n')]

instr_split = [instr_full[i:j] 
               for z in [[i for i,j in enumerate(instr_full) 
                          if j[0]=='inp']+[len(instr_full)]]
               for i,j in zip(z[:-1],z[1:])]
instr_gps = [[i,j] 
             for z in [[0]+[i+1 for i,j in enumerate(instr_split)
                            if min([k for k in [*zip(*j)][2] 
                                    if isinstance(k,int)])<0]]
             for i,j in zip(z[:-1],z[1:])]

digits = ''
registers = {i:0 for i in 'wxyz'}
for f,t in instr_gps:
    d_add, registers = get_zero_combos(instr_split[f:t], registers)[-1]
    digits += d_add
int(digits)
```

```python
# Part 2
def get_zero_combos_outer(instr_gps, registers=None):
    if registers is None:
        registers = {i:0 for i in 'wxyz'}
    if instr_gps:
        for i,j in get_zero_combos(instr_split[instr_gps[0][0]:instr_gps[0][1]], 
                                   registers):
            if (k:=get_zero_combos_outer(instr_gps[1:], j)) is not None:
                return i+k
    else:
        return '' if registers['z']==0 else None

int(get_zero_combos_outer(instr_gps))
```
