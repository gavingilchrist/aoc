```python
# Get raw data
with open('input/07.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
crab_pos = [int(i) for i in rawinput.split(',')]

min([sum([abs(j-i) for j in crab_pos])
     for i in range(min(crab_pos), max(crab_pos))])
```

```python
# Part 1
min([sum([(z:=abs(j-i))*(z+1)//2 for j in crab_pos])
     for i in range(min(crab_pos), max(crab_pos))])
```
