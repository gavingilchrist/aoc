```python
# Get raw data
with open('input/25.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
def evolve(cucumbers):
    filt_f = ((cucumbers == '>') 
              & (np.append(cucumbers[:, 1:], cucumbers[:, :1], axis=1) == '.'))
    filt_t = np.append(filt_f[:,-1:], filt_f[:,:-1], axis=1)
    cucumbers = np.where(filt_f, '.', np.where(filt_t, '>', cucumbers))
    filt_f = ((cucumbers == 'v') 
              & (np.append(cucumbers[1:], cucumbers[:1], axis=0) == '.'))
    filt_t = np.append(filt_f[-1:], filt_f[:-1], axis=0)
    return np.where(filt_f, '.', np.where(filt_t, 'v', cucumbers))

cucumbers = np.array([list(i) for i in rawinput.split('\n')])
stepnum = 1
while np.any(cucumbers != (cucumbers:= evolve(cucumbers))):
    stepnum += 1
stepnum
```
