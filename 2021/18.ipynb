```python
# Get raw data
with open('input/18.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
nest_depth = lambda x: [(z:=(z if i else 0)-(i and y==']')+((y:=j)=='[')) 
                        for i,j in enumerate(x)]
num_mask = lambda x: ''.join([' X'[48<=ord(i)<=57] for i in x])

def get_numix(x,last=False):
    mask = num_mask(x[::1-2*last])
    if 'X' in mask:
        ix = (s:=mask.index('X')), mask.index(' ',s)
        return tuple([len(x)-i for i in ix[::-1]]) if last else ix
    else:
        return None
    
def explode(x,ef,et):
    exp_nums = x[ef:et][1:-1].split(',')
    return '0'.join([(y[:z[0]]
                      + str(eval(f'{y[z[0]:z[1]]}+{exp_nums[0]}'))
                      + y[z[1]:]) if (z:=get_numix((y:=x[:ef]),True)) else y,
                     (y[:z[0]]
                      + str(eval(f'{y[z[0]:z[1]]}+{exp_nums[1]}'))
                      + y[z[1]:]) if (z:=get_numix(y:=x[et:])) else y])
    
def tenplus(x):
    mask = num_mask(x)
    if (tp:=[i 
             for i,j in enumerate(zip(mask[1:],mask)) 
             if {*j}=={'X'}]):
        return tp[0], mask.index(' ',tp[0])
    
def split(x,sf,st):
    return x[:sf]+f'[{int(x[sf:st])//2},{-(-int(x[sf:st])//2)}]'+x[st:]

def evolve(x):
    if 5 in (nd:=nest_depth(x)):
        et = nd.index(4,(ef:=nd.index(5)))
        x = explode(x,ef,et)
    elif (tp:=tenplus(x)):
        x = split(x,*tp)
    return x

def add_and_reduce(x,y):
    xy = f'[{x},{y}]'
    while xy != (xy:=evolve(xy)):  pass
    return xy

def get_magnitude(x):
    while '[' in x:
        for mf,mt in [[i,z] 
                      for i,j in enumerate(x) 
                      if j=='[' 
                      and '[' not in x[i+1:(z:=x.index(']',i)+1)]][::-1]:
            x = (x[:mf] 
                 + str(sum([i*j for i,j in zip(eval(x[mf:mt]),[3,2])])) 
                 + x[mt:])
    return int(x)

numbers = [i for i in rawinput.split('\n')]
snail_sum = [(z:=add_and_reduce(z,j) if i else j) for i,j in enumerate(numbers)][-1]
get_magnitude(snail_sum)
```
