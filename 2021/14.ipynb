```python
# Get raw data
with open('input/14.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
template, raw_rules = [i for i in rawinput.split('\n\n')]
rules = {j[0]:j[1] for i in raw_rules.split('\n') for j in [i.split(' -> ')]}

known = dict()
def elements_from_pair(pair, n):
    dkey = f'{pair}-{n}'
    if dkey not in known:
        if n:
            known[dkey] = {k: x.get(k,0)+y.get(k,0)+(k==z)
                           for z in [rules[pair]]
                           for x,y in [[elements_from_pair(pair[0]+z, n-1),
                                        elements_from_pair(z+pair[1], n-1)]]
                           for k in {*x,*y,z}}
        else:
            known[dkey] = dict()
    return known[dkey]

element_counts = {i: template.count(i) for i in {*template}}
for i in range(len(template)-1):
    for k,v in elements_from_pair(template[i:i+2], 10).items():
        element_counts[k] = element_counts.get(k,0) + v

(z:=sorted(element_counts.values()))[-1]-z[0]
```

```python
# Part 2
element_counts = {i: template.count(i) for i in {*template}}
for i in range(len(template)-1):
    for k,v in elements_from_pair(template[i:i+2], 40).items():
        element_counts[k] = element_counts.get(k,0) + v

(z:=sorted(element_counts.values()))[-1]-z[0]
```
