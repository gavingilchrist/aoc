```python
# Get raw data
with open('input/02.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
pos = depth = 0
for i in rawinput.split('\n'):
    if i.startswith('f'):
        pos += int(i.split()[1])
    else:
        depth += int(i.split()[1]) * [-1,1][i.startswith('d')]
pos*depth
```

```python
# Part 2
pos = depth = aim = 0
for i in rawinput.split('\n'):
    if i.startswith('f'):
        pos += (z:=int(i.split()[1]))
        depth += z * aim
    else:
        aim += int(i.split()[1]) * [-1,1][i.startswith('d')]
pos*depth
```
