```python
# Get raw data
with open('input/15.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
evolve = lambda x,y: np.minimum(x,
                                np.min(np.stack(((z:=np.pad(x+y,
                                                            1,
                                                            constant_values=np.sum(y)))[:-2,1:-1], 
                                                 z[2:, 1:-1], z[1:-1,:-2], z[1:-1,2:])), 
                                       axis=0))

def get_min_risk(riskmap):
    rmtot = np.sum(riskmap)
    minrisk = np.full((riskmap.shape), rmtot)
    minrisk[-1,-1] = 0
    while np.any(minrisk != (minrisk:=evolve(minrisk,riskmap))):  pass
    return minrisk[0,0].item()

riskmap = np.array([[int(j) for j in i] for i in rawinput.split('\n')])

get_min_risk(riskmap)
```

```python
# Part 2
for i in [0,1]:
    riskmap = np.concatenate([np.mod(riskmap+j-1,9)+1 for j in range(5)], axis=i)
    
get_min_risk(riskmap)
```
