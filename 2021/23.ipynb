```python
# Get raw data
with open('input/23.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
def get_options(burrows, hallway):
    for i,j in hallway.items():
        if (j!='.' 
            and len({*burrows[(y:=ord(j)*2-127)]}-{j})==0
            and all([hallway[k]=='.' 
                     for k in hallway 
                     if min([i,y])<k<max([i,y])])):
            return [(j,'H',i,'B',y,
                     (10**(ord(j)-65))*(abs(y-i)+N_PER_TYPE-len(burrows[y])))]
    for i,j in burrows.items():
        if (j
            and (y:=ord(j[-1])*2-127)!=i
            and len({*burrows[y]}-{j[-1]})==0
            and all([hallway[k]=='.' 
                     for k in hallway 
                     if min([i,y])<k<max([i,y])])):
            return [(j[-1],'B',i,'B',y,
                     (10**(ord(j[-1])-65))
                     *(abs(y-i)+2*N_PER_TYPE+1-len(burrows[i])-len(burrows[y])))]
    return [([j[-1],'B',i,'H',h,
              (10**(ord(j[-1])-65))*(abs(h-i)+N_PER_TYPE+1-len(burrows[i]))])
            for i,j in burrows.items()
            for h in hallway
            if (j
                and hallway[h]=='.'
                and ((y:=ord(j[-1])*2-127)!=i or len({*burrows[i]}-{j[-1]})!=0)
                and all([hallway[k]=='.' 
                         for k in hallway 
                         if min(i,h)<k<max(i,h)]))]

def apply_option(burrows, hallway, movedet):
    atype,ft,fc,tt,tc=movedet
    return ({i:[[j,j+[atype]][(tt=='B')&(tc==i)],j[:-1]][(ft=='B')&(fc==i)]
             for i,j in burrows.items()},
            {i:[[j,atype][(tt=='H')&(tc==i)],'.'][(ft=='H')&(fc==i)]
             for i,j in hallway.items()})

def get_dkey(burrows, hallway):
    return '|'.join([''.join(burrows[i]) for i in sorted(burrows)]
                    +[''.join([hallway[i] for i in sorted(hallway)])])

def get_min_energy(burrows, hallway):
    dkey = get_dkey(burrows, hallway)
    if dkey not in known:
        if all([j==[chr((i+127)//2)]*N_PER_TYPE for i,j in burrows.items()]):
            known[dkey] = 0
        else:
            known[dkey] = (min(z) 
                           if (z:=[cost+y 
                                   for *movedet,cost in get_options(burrows, hallway)
                                   if (y:=get_min_energy(*apply_option(burrows, 
                                                                       hallway, 
                                                                       movedet))) is not None]) 
                           else None)
    return known[dkey]

hallway = {i:(z:=rawinput.split('\n'))[1][i] 
           for i in {*range(1,12)}-{*range(3,10,2)}}
burrows = {i:[k for k in j[-2:1:-1] if k!='.'] 
           for i,j in enumerate(zip(*z)) 
           if j[2]!='#'}
N_PER_TYPE = len(z)-3

known = dict()
get_min_energy(burrows, hallway)
```
