```python
# Get raw data
with open('input/20.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
raw_algo, raw_image = [i for i in rawinput.split('\n\n')]
algorithm = [int(i=='#') for i in raw_algo]
pixels = {(i,j): int(c=='#') 
          for i,r in enumerate(raw_image.split('\n')) 
          for j,c in enumerate(r)}

def get_lit_pixels(pixels, nsteps, inf_fill=0):
    if nsteps:
        ylims,xlims = [[min(i)-1, max(i)+2] for i in zip(*pixels)]
        newstate = {(y,x): algorithm[int(''.join([str(pixels.get((y+yy,x+xx), inf_fill)) 
                                                  for yy in [-1,0,1] 
                                                  for xx in [-1,0,1]]),
                                         2)]
                    for y in range(*ylims)
                    for x in range(*xlims)}
        return get_lit_pixels({**pixels, **newstate},
                              nsteps-1, 
                              algorithm[int(str(inf_fill)*9,2)])
    else:
        return sum(pixels.values())
    
get_lit_pixels(pixels, 2)
```
