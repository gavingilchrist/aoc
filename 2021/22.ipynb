```python
# Get raw data
with open('input/22.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
from math import prod
```

```python
# Part 1
steps = [(int(j[0]=='on'), 
          [tuple([m+int(n) 
                  for m,n in enumerate(l[2:].split('..'))]) 
           for l in k]) 
         for i in rawinput.split('\n') 
         for j in [i.split()] 
         for k in [j[1].split(',')]]

def get_n_on_cubes(steps):
    on_cubes = {}
    for s,t in steps:
        on_cubes_add = []
        for j,i in on_cubes.items():
            if all([t[k][0]<j[k][1] and j[k][0]<t[k][1] 
                    for k in [0,1,2]]):
                on_cubes_add += [([(max(t[k][0],j[k][0]), min(t[k][1],j[k][1])) 
                                   for k in [0,1,2]],
                                  -i)]
        if s:
            on_cubes_add += [(t, 1)]
            
        for i,j in on_cubes_add:
            if (z:=on_cubes.get((y:=tuple(i)),0) + j):
                on_cubes[y] = z
            else:
                del(on_cubes[y])
                
    return sum([prod([j-i for i,j in k]) * on_cubes[k] 
                for k in on_cubes])
    
get_n_on_cubes([i for i in steps 
                if all([-50<=j<k<=50 for j,k in i[1]])])
```

```python
# Part 2
get_n_on_cubes(steps)
```
