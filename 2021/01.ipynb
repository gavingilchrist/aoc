```python
# Get raw data
with open('input/01.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum([i>j 
     for z in[[int(i) for i in rawinput.split('\n')]]
     for i,j in zip(z[1:], z[:-1])])
```

```python
# Part 2
sum([sum([i,j,k])>sum([j,k,l])
     for z in[[int(i) for i in rawinput.split('\n')]]
     for i,j,k,l in zip(z[3:], z[2:-1], z[1:-2], z[:-3])])
```
