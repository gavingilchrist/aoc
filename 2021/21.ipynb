```python
# Get raw data
with open('input/21.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
def make_die(n):
    while True:
        for i in range(n):
            yield i+1

startpos = [int(i.split(': ')[1]) for i in rawinput.split('\n')]

pos = [*startpos]
score = [0]*2
pturn = 0
nrolls = 0
die_roll = make_die(100)

while max(score) < 1000:
    pos[pturn] = (pos[pturn]+sum([next(die_roll) for i in '123'])-1) % 10 + 1
    score[pturn] += pos[pturn]
    pturn = 1-pturn
    nrolls += 3

nrolls * min(score)
```

```python
# Part 2
roll_freqs = [[l, z.count(l)]
              for z in [[sum([i,j,k]) 
                         for i in [1,2,3] for j in [1,2,3] for k in [1,2,3]]]
              for l in {*z}]

known = dict()
def get_win_counts(gamestate=None):
    pos, score, pturn = gamestate or ([*startpos], [0]*2, 0)
    dkey = '-'.join([str(i) for i in [*pos, *score, pturn]])
    if dkey not in known:
        if max(score)>=21:
            known[dkey] = [int(i>=21) for i in score]
        else:
            known[dkey] = [sum(l) 
                           for l in zip(*[[k*rf
                                           for z in [[[j,(j+r-1)%10+1][i==pturn] 
                                                      for i,j in enumerate(pos)]]
                                           for k in get_win_counts([z,
                                                                    [[j,j+z[i]][i==pturn] 
                                                                     for i,j in enumerate(score)],
                                                                    1-pturn])]
                                          for r,rf in roll_freqs])]
    return known[dkey]

max(get_win_counts())
```
