```python
# Get raw data
with open('input/04.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
is_bingo = lambda x,y: any([all([c in y for c in r]) 
                            for i in [x, zip(*x)] 
                            for r in i])

[draw], *boards = [[[int(k) 
                     for k in j.replace(',',' ').split()] 
                    for j in i.split('\n')]
                   for i in rawinput.split('\n\n')]

d=0
while not any(z:=[is_bingo(b,(y:=draw[:d])) for b in boards]):  d+=1

sum([i for i in sum(boards[z.index(1)],[]) if i not in y]) * y[-1]
```

```python
# Part 2
for d2 in range(d+1,len(draw)):
    if any(z:=[not is_bingo(b,(y:=draw[:d2])) for b in boards]):
        ix = z.index(1)
    elif is_bingo(boards[ix],y):
        break
sum([i for i in sum(boards[ix],[]) if i not in y]) * y[-1]
```
