```python
# Get raw data
with open('input/11.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
octos = {(i,j): int(c) 
         for i,r in enumerate(rawinput.split('\n')) 
         for j,c in enumerate(r)}
n_flashes = 0

def do_step(octos):
    for i in octos:
        octos[i] += 1
    
    flashed = []
    while (f:=[k for k,v in octos.items() 
            if k not in flashed and v>9]):
        flashed += f
        for i,j in f:
            for k in [-1,0,1]:
                for l in [-1,0,1]:
                    if ((k or l) 
                        and (z:=(i+k,j+l)) in octos 
                        and z not in flashed):
                        octos[z] += 1
    for i in flashed:
        octos[i] = 0
        
    return len(flashed)

for h in range(1,101):
    n_flashes += do_step(octos)
        
n_flashes
```

```python
# Part 2
while {*octos.values()} != {0}:
    do_step(octos)
    h += 1
h
```
