```python
# Get raw data
with open('input/17.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
def get_ypos(vy):
    ypos = [0]
    while ypos[-1] >= yl:
        ypos += [ypos[-1]+vy]
        vy -= 1
    return ([i 
             for i,j in enumerate(ypos) 
             if yl<=j<=yh], 
            max(ypos))

def get_xpos(vx):
    xpos = [0]
    while True:
        xpos += [xpos[-1]+vx]
        vx = max(vx-1,0)
        if (z:=len({*xpos[-2:]})==1) or xpos[-1]>xh:
            break
    return [i 
            for i,j in enumerate(zip(xpos,[0]+xpos[:-1])) 
            if sum([xl<=k<=xh for k in j])==1]
    
(xl,xh),(yl,yh) = [[int(j) for j in i[2:].split('..')]
                   for i in rawinput.replace(',',' ').split()[2:]]

ypos = [[vy,*z] for vy in range(yl-1,1-yl) if (z:=get_ypos(vy))[0]]
xpos = [[vx,z] for vx in range(xh+1) if (z:=get_xpos(vx))]

max([maxy
     for vy,nposs,maxy in ypos
     for n in nposs
     for vx,nrg in xpos
     if n>=nrg[0] and (len(nrg)==1 or n<nrg[1])])
```
