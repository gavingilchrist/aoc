```python
# Get raw data
with open('input/03.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
gamma, epsilon = [int(''.join([(z:={i.count(j):j for j in {*i}})[f(z)]
                               for i in zip(*[list(i) for i in rawinput.split('\n')])]), 2)
                  for f in [max,min]]
gamma * epsilon
```

```python
# Part 2
def get_rating(numbers, func):
    for i in range(len(numbers[0])):
        numbers = [numbers[l] 
                   for j in [[*zip(*numbers)][i]]
                   for l,m in enumerate(j) 
                   if m==(z:={j.count(k)+int(k)/10:k for k in {*j}})[func(z)]]
        if len(numbers) <= 1:
            break
    return int(numbers[0], 2)

get_rating(z:=[i for i in rawinput.split('\n')], max) * get_rating(z, min)
```
