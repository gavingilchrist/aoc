```python
# Get raw data
with open('input/05.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
vent_lines = [sorted([tuple([int(l) for l in k.split(',')])
                      for k in j])
              for i in rawinput.split('\n') 
              for j in [i.split(' -> ')]]
vln = len(vent_lines)
vltype = [['H','V','R','L'][[i==k, j==l, j<l, j>l].index(1)]
          for (i,j),(k,l) in vent_lines] 
         
grid = np.zeros((1000,1000), dtype=int)
for i in range(vln):
    if vltype[i] in ['H','V']:
        (t,u),(v,w)=vent_lines[i]
        grid[t:v+1,u:w+1] += 1
        
np.sum(grid > 1).item()
```

```python
# Part 2
for i in range(vln):
    if vltype[i] not in ['H','V']:
        (t,u),(v,w)=vent_lines[i]
        grid[np.arange(t,v+1),
             np.arange(min(u,w),max(u,w)+1)[::2*(w>=u)-1]] += 1
np.sum(grid > 1).item()
```
