```python
# Get raw data
with open('input/10.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
def checkline(x,y=None):
    if x:
        if x[0] in('([{<'):
            return checkline(x[1:], ([] if y is None else y)+['([{<'.index(x[0])])
        elif x[0] == ')]}>'[y.pop()]:
            return checkline(x[1:], y)
        else:
            return 'Invalid', {')': 3, ']': 57, '}': 1197, '>': 25137}[x[0]]
    else:
        return 'Valid', [(z:=5*(z if i else 0)+j+1) for i,j in enumerate(y[::-1])][-1]

line_summ = [checkline(i) for i in rawinput.split('\n')]
sum([j for i,j in line_summ if i == 'Invalid'])
```

```python
# Part 2
(z:=sorted([j for i,j in line_summ if i == 'Valid']))[len(z)//2]
```
