```python
# Get raw data
with open('input/08.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
display_data = [[[''.join(sorted(l)) for l in k.split()] 
                 for k in j] 
                for i in rawinput.split('\n') 
                for j in [i.split(' | ')]]
sum([len(j) in[2,3,4,7] for i in display_data for j in i[1]])
```

```python
# Part 2
true_digits = ['abcefg', 'cf', 'acdeg', 'acdfg', 'bcdf', 
               'abdfg', 'abdefg', 'acf', 'abcdefg', 'abcdfg']
td_seg_cts = {(i,l): {m[0] for m in z if m[1]==l}
              for i in range(2,8)
              for y in [sum([list(j) for j in true_digits if len(j)==i],[])]
              for z in [[[k,y.count(k)] for k in {*y}]]
              for l in {*[*zip(*z)][1]}}

def get_seg_conv(patterns):
    seg_conv = {i: {*'abcdefg'} for i in 'abcdefg'}
    for i in range(2,8):
        seg_cts = [[k,z.count(k)] 
                   for z in[sum([list(j) for j in patterns if len(j)==i],[])]
                   for k in {*z}]
        for l,m in seg_cts:
            seg_conv[l] &= td_seg_cts[(i,m)]
    return {k:[*(v-s if len(v)>1 else v)][0]
            for s in[{j for i in seg_conv.values() if len(i)==1 for j in i}]
            for k,v in seg_conv.items()}

sum([sum([(10**j)*true_digits.index(''.join(sorted([seg_conv[l] for l in k])))
          for seg_conv in [get_seg_conv(i[0])] 
          for j,k in enumerate(i[1][::-1])])
     for i in display_data])
```
