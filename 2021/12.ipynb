```python
# Get raw data
with open('input/12.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
connections = dict()
for i in rawinput.split('\n'):
    for j,k in [i.split('-'), i.split('-')[::-1]]:
        if j != 'end' and k != 'start':
            connections[j] = connections.get(j,[])+[k]

def get_n_paths(cave, prev=None):
    if prev is None:
        prev = []
    if cave == 'end':
        return 1
    else:
        return sum([get_n_paths(i, prev+[cave])
                    for i in connections[cave]
                    if (65 <= ord(i[0]) <= 90) or (i not in prev)])

get_n_paths('start')
```

```python
# Part 2
def get_n_paths(cave, prev=None, twice=False):
    if prev is None:
        prev = []
    if cave == 'end':
        return 1
    else:
        return sum([get_n_paths(i, 
                                prev+[cave], 
                                twice or (97 <= ord(i[0]) <= 122 and i in prev))
                    for i in connections[cave]
                    if (65 <= ord(i[0]) <= 90) 
                       or (i not in prev)
                       or (not twice and i != 'start')])

get_n_paths('start')
```
