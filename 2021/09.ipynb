```python
# Get raw data
with open('input/09.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
heightmap = np.array([[int(j) for j in i] for i in rawinput.split('\n')])
np.sum(np.where(heightmap 
                < np.min(np.stack([z[i:,j:][:heightmap.shape[0], :heightmap.shape[1]]
                                   for z in [np.pad(heightmap, 1, constant_values=9)]
                                   for i,j in [[0,1],[2,1],[1,0],[1,2]]]), 
                         axis=0),
                heightmap + 1,
                0)).item()
```

```python
# Part 2
ncells = heightmap.size
evolve = lambda x: np.where(x==ncells, ncells,
                            np.min(np.stack([z[i:,j:][:heightmap.shape[0], :heightmap.shape[1]]
                                             for z in [np.pad(x, 1, constant_values=ncells)]
                                             for i,j in [[1,1],[0,1],[2,1],[1,0],[1,2]]]),
                                   axis=0))

basin_id = np.where(heightmap == 9, ncells, np.arange(ncells).reshape(heightmap.shape))
while np.any(basin_id != (basin_id:=evolve(basin_id))):  pass

np.prod(np.sort(np.unique(basin_id, return_counts=True)[1][:-1])[-3:]).item()
```
