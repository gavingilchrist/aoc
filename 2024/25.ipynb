```python
# Get raw data
with open('input/25.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
locks,keys = [[[k.count('#') for k in zip(*j)]
               for i in rawinput.split('\n\n')
               for j in [i.split('\n')]
               if j[0][0]==h]
              for h in '#.']
sum([all([sum(k)<=7 for k in zip(i,j)])
     for i in locks
     for j in keys])
```
