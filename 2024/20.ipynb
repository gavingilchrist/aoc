```python
# Part 1
def evolve(reach):
    rp = np.minimum(np.pad(reach,1,constant_values=rtsize)+1, rtsize)
    return np.where((reach==rtsize) & (racetrack!='#'),
                    np.min(np.stack((rp[:-2,1:-1], 
                                     rp[2:,1:-1], 
                                     rp[1:-1,:-2], 
                                     rp[1:-1,2:])), 
                           axis=0),
                    reach)
    
def get_reach(startpt):
    reach = np.full_like(racetrack, rtsize, dtype=int)
    reach[startpt] = 0
    while np.any(reach != (reach:=evolve(reach))):  pass
    return reach

racetrack = np.array([list(i) for i in rawinput.split('\n')])
rtsize = racetrack.size
startpos = ((y:=np.arange(rtsize)[racetrack.reshape(-1)=='S'][0].item()) 
            // (z:=racetrack.shape[1]), 
            y%z)

reach = get_reach(startpos)
rp = np.pad(reach,2,constant_values=rtsize)
saved = np.where(((x:=np.expand_dims(reach,0)) != rtsize)
                 & ((y:=np.stack((rp[:-4,2:-2],rp[2:-2,:-4]))) != rtsize),
                 np.abs(x-y)-2, 
                 0)
np.sum(saved >= 100).item()
```
