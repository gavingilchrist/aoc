```python
# Get raw data
with open('input/06.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
rotate = lambda x,y,z: (np.flip(x.T,axis=0), (x.shape[1]-y[1]-1,y[0]), (z+1)%4)

labmap_orig = np.flip(np.array([list(i) for i in rawinput.split('\n')]).T, axis=1)
start_pos = ((x:=np.argmax(labmap_orig.reshape(-1)=='^').item())//(y:=labmap_orig.shape[1]), x%y)

labmap = np.copy(labmap_orig)
pos = start_pos
orient = 0
while True:
    if np.any((y:=labmap[pos[0],pos[1]:])=='#'):
        y[:(z:=np.argmax(y=='#').item())] = 'X'
        labmap,pos,orient = rotate(labmap, 
                                   [sum(i) for i in zip(pos,[0,z-1])], 
                                   orient)
    else:
        y[:] = 'X'
        break
np.sum(labmap == 'X').item()
```

```python
# Part 2
def isloop(labmap,pos):
    labmap = np.copy(labmap)
    orient = 0
    hist = []
    while True:
        if (pos,orient) in hist:
            return True
        else:
            hist += [(pos,orient)]
        if np.any((y:=labmap[pos[0],pos[1]:])=='#'):
            labmap,pos,orient = rotate(labmap, 
                                       (pos[0],pos[1]+np.argmax(y=='#').item()-1), 
                                       orient)
        else:
            return False

for i in range(4-orient):
    labmap,pos,orient = rotate(labmap,pos,orient)
block_poss = [tuple(i)
              for x,y in [[np.arange(labmap.size)[labmap.reshape(-1)=='X'],
                           labmap.shape[1]]]
              for i in np.column_stack((x//y,x%y)).tolist()]

nloops = 0
for p in block_poss:
    labmap = np.copy(labmap_orig)
    labmap[*p] = '#'
    if isloop(labmap,start_pos):
        nloops += 1
nloops
```
