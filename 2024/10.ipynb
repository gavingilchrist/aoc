```python
# Get raw data
with open('input/10.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
topmap = np.array([list(i) for i in rawinput.split('\n')], dtype=int)

trailheads = np.arange(topmap.size)[topmap.reshape(-1)==0]
tstep = np.column_stack((trailheads//(z:=topmap.shape[1]), trailheads%z))

for s in range(1,10):
    trailheads = np.repeat(trailheads,4)
    tstep = (np.repeat(tstep, 4 ,axis=0) 
             + np.tile([[-1,0],[1,0],[0,-1],[0,1]], (tstep.shape[0],1)))
    f1 = np.all((tstep>=0) & (tstep<topmap.shape), axis=1)
    f2 = topmap[tstep[f1,0],tstep[f1,1]]==s
    trailheads, tstep = [i[f1][f2] for i in [trailheads, tstep]]

len({*zip(trailheads.tolist(), [tuple(i) for i in tstep.tolist()])})
```
