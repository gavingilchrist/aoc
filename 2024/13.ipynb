```python
# Get raw data
with open('input/13.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum([int(3*a_push + b_push)
     for i in rawinput.split('\n\n')
     for a,b,targ in [[[int(k[2:]) 
                        for k in j.split(': ')[1].split(', ')] 
                       for j in i.split('\n')]]
     for a_push in [(targ[0]*b[1]-targ[1]*b[0]) / (a[0]*b[1]-a[1]*b[0])]
     for b_push in [(targ[0]-a_push*a[0])/b[0]]
     if all([l>=0 and l%1==0 for l in [a_push,b_push]])])
```

```python
# Part 2
sum([int(3*a_push + b_push)
     for i in rawinput.split('\n\n')
     for a,b,targ in [[[int(k[2:]) + [0,10000000000000][nj==2]
                        for k in j.split(': ')[1].split(', ')] 
                       for nj,j in enumerate(i.split('\n'))]]
     for a_push in [(targ[0]*b[1]-targ[1]*b[0]) / (a[0]*b[1]-a[1]*b[0])]
     for b_push in [(targ[0]-a_push*a[0])/b[0]]
     if all([l>=0 and l%1==0 for l in [a_push,b_push]])])
```
