```python
# Get raw data
with open('input/09.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
diskmap = np.array([*f'{rawinput}0'], dtype=int).reshape(-1,2)

w = np.max(rtot:=np.sum(diskmap, axis=1, keepdims=True))
disk_src = np.where((z:=np.where((y:=np.expand_dims(np.arange(w),0)) < diskmap[:,:1], 1,
                                 np.where(y < rtot, 0, -1)).reshape(-1))[z != -1] == 1,
                    np.repeat(np.arange(diskmap.shape[0]), rtot[:,0]),
                    -1)

disk = np.copy(disk_src)
gapix, fillix = (rix:=np.arange(disk.shape[0]))[disk==-1], rix[disk!=-1][::-1]
n_to_move = sum([i<j for i,j in zip(gapix,fillix)])
disk[gapix[:n_to_move]] = disk[fillix[:n_to_move]]
disk[fillix[:n_to_move]] = -1

np.sum((rix*disk)[disk != -1]).item()
```

```python
# Part 2
disk = np.copy(disk_src)
for bl_id in range(len(diskmap)-1,-1,-1):
    gap_sz = ((z:=np.pad(rix[disk==-1],1,constant_values=-1))[1:-1][z[2:]-z[1:-1]!=1] 
              - (gap_ix:=z[1:-1][z[1:-1]-z[:-2]!=1]) + 1)
    bl_sz = diskmap[bl_id][0]
    src_s = np.argmax(disk==bl_id)
    if np.any(fit:=gap_sz[gap_ix<src_s] >= bl_sz):
        disk[src_s:src_s+bl_sz] = -1
        disk[(dest_s:=gap_ix[np.argmax(fit)]):dest_s+bl_sz] = bl_id

np.sum((rix*disk)[disk != -1]).item()
```
