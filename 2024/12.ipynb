```python
# Get raw data
with open('input/12.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
farm_map = np.array([list(i) for i in rawinput.split()])

region = np.pad(np.arange(1,farm_map.size+1).reshape(farm_map.shape), 1)
hreg = np.max(region)+1
fmpad = np.pad(farm_map,1)

def evolve(region):
    region = np.copy(region)
    region[1:-1,1:-1] = np.minimum(region[1:-1,1:-1],
                                   np.min(np.stack((np.where(fmpad[:-2,1:-1]==fmpad[1:-1,1:-1], 
                                                             region[:-2,1:-1], hreg),
                                                    np.where(fmpad[2:,1:-1]==fmpad[1:-1,1:-1], 
                                                             region[2:,1:-1], hreg),
                                                    np.where(fmpad[1:-1,:-2]==fmpad[1:-1,1:-1], 
                                                             region[1:-1,:-2], hreg),
                                                    np.where(fmpad[1:-1,2:]==fmpad[1:-1,1:-1], 
                                                             region[1:-1,2:], hreg))), 
                                          axis=0))
    return region

while np.any(region != (region:=evolve(region))):  pass

plot_reg = region[1:-1,1:-1].reshape(-1,1)
plot_circ = np.sum((region != np.stack(((z:=np.pad(region,1))[:-2,1:-1],
                                        z[2:,1:-1],z[1:-1,:-2],z[1:-1,2:]))), 
                   axis=0)[1:-1,1:-1].reshape(-1,1)

reg, area = np.unique(plot_reg, return_counts=True)
circ = np.sum(np.where(plot_reg==[reg], plot_circ, 0), axis=0)
np.column_stack([i.astype(object) for i in [reg,area,circ]])

np.sum(area * circ).item()
```

```python
# Part 2
reg_4ways = [(r:=np.flip(r.T,1) if o else region) for o in range(4)]
fences = np.unique(np.concatenate([(z:=reg[1:-1,1:-1]).reshape(-1)
                                       [((z!=reg[:-2,1:-1]) 
                                         & ((z!=reg[1:-1,:-2])
                                            | (reg[1:-1,:-2]==reg[:-2,:-2])))
                                        .reshape(-1)]
                                   for reg in reg_4ways]),
                   return_counts=True)[1]

np.sum(area * fences).item()
```
