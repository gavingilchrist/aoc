```python
# Get raw data
with open('input/01.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum([max(l)-min(l) 
     for l in zip(*[sorted(k) 
                    for k in zip(*[[int(j) for j in i.split()] 
                                   for i in rawinput.split('\n')])])])
```

```python
# Part 2
sum([l*k[1].count(l)
     for k in [[*zip(*[[int(j) for j in i.split()] 
                       for i in rawinput.split('\n')])]]
     for l in {*k[0]}])
```
