```python
# Get raw data
with open('input/17.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
class Program(object):
    def __init__(self, instr, registers):
        self.ops = [
            self.do_adv,
            self.do_bxl,
            self.do_bst,
            self.do_jnz,
            self.do_bxc,
            self.do_out,
            self.do_bdv,
            self.do_cdv
        ]
        self.instr = instr
        self.registers = {**registers}
        self.output = []
        self.ptr = 0
        
    def combo(self, operand):
        return self.registers.get(chr(operand+61),operand)
        
    def do_adv(self, operand):
        self.registers['A'] = int(self.registers['A'] / (2**self.combo(operand)))

    def do_bxl(self, operand):
        self.registers['B'] = self.registers['B'] ^ operand

    def do_bst(self, operand):
        self.registers['B'] = self.combo(operand) % 8
    
    def do_jnz(self, operand):
        if self.registers['A']:
            self.ptr = operand
    
    def do_bxc(self, operand):
        self.registers['B'] = self.registers['B'] ^ self.registers['C']

    def do_out(self, operand):
        self.output += [self.combo(operand) % 8]

    def do_bdv(self, operand):
        self.registers['B'] = int(self.registers['A'] / (2**self.combo(operand)))

    def do_cdv(self, operand):
        self.registers['C'] = int(self.registers['A'] / (2**self.combo(operand)))

    def do_exec(self):
        while 0 <= (p:=self.ptr) < len(self.instr)-1:
            opcode,operand = self.instr[self.ptr:self.ptr+2]
            self.ops[opcode](operand)
            if self.ptr==p:
                self.ptr += 2
        return self.output
    
    @classmethod
    def run(cls, instr, registers):
        obj = cls(instr, registers)
        return obj.do_exec()
    
raw_reg, raw_instr = [i for i in rawinput.split('\n\n')]
registers = {j[1][0]: int(j[2]) 
             for i in raw_reg.split('\n') 
             for j in [i.split()]}
instr = [int(i) for i in raw_instr.split(': ')[1].split(',')]    

print(','.join([str(i) for i in Program.run(instr, registers)]))
```

```python
# Part 2
def get_rega_value(test_a=0,codeix=-1):
    if codeix < -len(instr):
        return test_a
    else:
        next_ta = [i
                   for i in range(test_a*8,(test_a+1)*8)
                   if Program.run(instr,{**registers, **{'A': i}})[codeix]==instr[codeix]]
        for nx in next_ta:
            if (z:=get_rega_value(nx,codeix-1)):
                return z
        return None

get_rega_value()
```
