```python
# Get raw data
with open('input/11.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
from math import log10
```

```python
# Part 1
nextlist = lambda x: ([1] if x==0 
                      else [x//(z:=10**(-(-y//2))), x%z] if (y:=int(log10(x)))%2 
                      else [x*2024])
evolve = lambda x: [j for i in x for j in nextlist(i)]
stones = [int(i) for i in rawinput.split()]
for i in range(25):
    stones = evolve(stones)
len(stones)
```

```python
# Part 2
def get_n_stones(stone, blinks_left):
    dkey = (stone,blinks_left)
    if dkey not in known:
        if blinks_left:
            known[dkey] = sum([get_n_stones(i,blinks_left-1)
                               for i in nextlist(stone)])
        else:
            known[dkey] = 1
    return known[dkey]
known = {}
sum([get_n_stones(int(i),75) for i in rawinput.split()])
```
