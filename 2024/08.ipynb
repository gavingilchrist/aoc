```python
# Get raw data
with open('input/08.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
freqnodes = {}
for i,r in enumerate(rawinput.split('\n')):
    for j,c in enumerate(r):
        if c != '.':
            freqnodes[c] = freqnodes.get(c,[])+[(i,j)]
grid_max = [i,j]
antinodes = set()
for i,j in freqnodes.items():
    for k in range(len(j)-1):
        for l in range(k+1,len(j)):
            antinodes.update({z
                              for m in [1,-1]
                              for z in [tuple([2*o[0]-o[1]
                                               for n in zip(j[k],j[l])
                                               for o in [n[::m]]])]
                              if all([0<=i<=j for i,j in zip(z,grid_max)])})
len(antinodes)
```

```python
# Part 2
# (Check that no node pairs are vertically-aligned)
assert all([len(z:=[*zip(*i)][1])==len({*z}) for i in freqnodes.values()])

for i,j in freqnodes.items():
    for k in range(len(j)-1):
        for l in range(k+1,len(j)):
            m = (j[k][0]-j[l][0])/(j[k][1]-j[l][1])
            c = j[k][0]-m*j[k][1]
            antinodes.update({(int(y),x) 
                              for x in range(grid_max[1]+1) 
                              if (y:=round(m*x+c,10))%1==0
                                 and 0<=y<=grid_max[0]})
len(antinodes)
```
