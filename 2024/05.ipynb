```python
# Get raw data
with open('input/05.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
rawrules, rawupdates = [i for i in rawinput.split('\n\n')]
rules = [[int(j) for j in i.split('|')] for i in rawrules.split('\n')]
updates = [[int(j) for j in i.split(',')] for i in rawupdates.split('\n')]

sum([update[len(update)//2]
     for update in updates
     if all([any([i not in update for i in rule])
             or (z:=[update.index(i) for i in rule])==sorted(z)
             for rule in rules])])
```

```python
# Part 2
sum([sorted([[-f.count(i),i]
             for f in [[*zip(*rules_sub)][0]]
             for i in {*sum(rules_sub,[])}])[len(update)//2][1]
     for update in updates
     if not all([any([i not in update for i in rule])
                 or (z:=[update.index(i) for i in rule])==sorted(z)
                 for rule in rules])
     for rules_sub in [[rule
                        for rule in rules
                        if all([i in update for i in rule])]]])
```
