```python
# Get raw data
with open('input/02.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
def is_safe(report):
    diffs = [j-i for i,j in zip(report[:-1],report[1:])]
    l,h = [*zip(*sorted([[abs(i),i] for i in sorted(diffs)[::len(diffs)-1]]))][1]
    return h*l>0 and abs(l)>=1 and abs(h)<=3

reports = [[int(j) for j in i.split()]
           for i in rawinput.split('\n')]
sum([is_safe(report) for report in reports])
```

```python
# Part 2
sum([any([is_safe(report),
          *[is_safe(report[:i]+report[i+1:])
            for i in range(len(report))]])
     for report in reports])
```
