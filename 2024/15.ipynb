```python
# Get raw data
with open('input/15.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
def push(path):
    path_cp = np.copy(path[::-1] if (fl:=path[-1]=='@') else path)
    ix = sorted([[np.argmax(path_cp==j)+i,j] 
                 for i,j in enumerate('#.')])[0][0]
    path_cp[:ix] = np.sort(path_cp[:ix])
    path[:] = path_cp[::-1] if fl else path_cp[:]
    return path_cp[0] != '@'
    
raw_whmap, raw_moves = rawinput.split('\n\n')
whmap = np.array([[*i] for i in raw_whmap.split('\n')])
moves = raw_moves.replace('\n','')

pos = np.array([(y:=np.arange(whmap.size)
                 [whmap.reshape(-1)=='@'][0].item())//(z:=whmap.shape[1]), 
                y%z])
for m in moves:
    if push({'^':whmap[:pos[0]+1,pos[1]],
             'v':whmap[pos[0]:,pos[1]],
             '<':whmap[pos[0],:pos[1]+1],
             '>':whmap[pos[0],pos[1]:]}[m]):
        pos += {'^':[-1,0],'v':[1,0],'<':[0,-1],'>':[0,1]}[m]
        
np.sum(np.column_stack(((y:=np.arange(whmap.size)[whmap.reshape(-1)=='O'])//z, 
                        y%z)) 
       * [[100,1]]).item()
```

```python
# Part 2
def push(whmap, pos, m):
    if m in '<>':
        pdir = [-1,1][m=='>']
        tomv = (y:=whmap[pos[0]][::pdir])[np.argmax(y=='@'):]
        if ((z:=[tomv.shape[0],np.argmax(tomv=='.')][int(np.any(tomv=='.'))]) 
            < np.argmax(tomv=='#')):
            tomv[:z+1] = np.roll(tomv[:z+1],1)
            pos[1] += pdir
    else:
        pdir = [-1,1][m=='v']
        tomv = [[pos[0],{pos[1]}]]
        while tomv[-1][1] and -1 not in tomv[-1][1]:
            tomv += [[tomv[-1][0]+pdir, 
                      {j 
                       for i in tomv[-1][1] 
                       for j in {'@':[i],
                                 '[':[i,i+1],
                                 ']':[i-1,i],
                                 '#':[-1],
                                 '.':[]}[whmap[tomv[-1][0]+pdir,i]]}]]
        if -1 not in tomv[-1][1]:
            for i,j in tomv[:-1][::-1]:
                whmap[i+pdir,[*j]] = whmap[i,[*j]]
                whmap[i,[*j]] = '.'
            pos[0] += pdir
    return whmap, pos

whmap = np.array([[k
                   for j in i 
                   for k in {'#':'##', '.':'..', '@':'@.', 'O':'[]'}[j]] 
                  for i in raw_whmap.split('\n')])
pos = np.array([(y:=np.arange(whmap.size)
                 [whmap.reshape(-1)=='@'][0].item())//(z:=whmap.shape[1]), 
                y%z])

for m in moves:
    whmap, pos = push(whmap, pos, m)
    
np.sum(np.column_stack(((y:=np.arange(whmap.size)[whmap.reshape(-1)=='['])//z, 
                        y%z)) 
       * [[100,1]]).item()
```
