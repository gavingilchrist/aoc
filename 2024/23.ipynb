```python
# Get raw data
with open('input/23.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
comp_links = {}
for i in rawinput.split('\n'):
    pair = i.split('-')
    for j in [1,-1]:
        f,t = pair[::j]
        if f in comp_links:
            comp_links[f] += [t]
        else:
            comp_links[f] = [t]
            
len({tuple(sorted([i,j,k]))
     for i in comp_links
     for jn in range(len(comp_links[i])-1)
     for kn in range(jn+1,len(comp_links[i]))
     if (j:=comp_links[i][kn]) in comp_links[(k:=comp_links[i][jn])]
     and 't' in [*zip(*[i,j,k])][0]})
```

```python
# Part 2
comp_sets = [[i] for i in comp_links]
while len(comp_sets)>1:
    comp_sets = [i+[j]
                 for i in comp_sets
                 for j in comp_links[i[-1]]
                 if j>i[-1] and {*i}.issubset({*comp_links[j]})]
print(','.join(comp_sets[0]))
```
