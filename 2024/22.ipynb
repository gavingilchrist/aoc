```python
# Get raw data
with open('input/22.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
from collections import deque
```

```python
# Parts 1 and 2
class Buyer(object):
    def __init__(self,init_secret):
        self.init_secret = init_secret
        self.reset()
        
    def reset(self):
        self.secret = self.init_secret
        self.pch_hist = deque(maxlen=4)
        self.seq_price = dict()
        
    def mix(self,num):
        self.secret ^= num
        
    def prune(self):
        self.secret %= 16777216
        
    def evolve(self):
        pr_prev = self.secret % 10
        self.mix(self.secret*64)
        self.prune()
        self.mix(self.secret//32)
        self.prune()
        self.mix(self.secret*2048)
        self.prune()
        pr_now = self.secret % 10
        self.pch_hist += [pr_now-pr_prev]
        if len(z:=tuple(self.pch_hist))==4 and z not in self.seq_price:
            self.seq_price[z] = pr_now
        
    def get_secret_n(self,n,each=False):
        self.reset()
        if each: secrets = []
        for i in range(n):
            self.evolve()
            if each: secrets += [self.secret]
        return secrets if each else self.secret

init_sn = [int(i) for i in rawinput.split('\n')]

sum_s2k = 0
seq_total = {}
for s in init_sn:
    sum_s2k += (b:=Buyer(s)).get_secret_n(2000)
    for k,v in b.seq_price.items():
        if k in seq_total:
            seq_total[k] += v
        else:
            seq_total[k] = v
            
print('Part 1:', sum_s2k)
print('Part 2:', max(seq_total.values()))
```
