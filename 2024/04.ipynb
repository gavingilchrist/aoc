```python
# Get raw data
with open('input/04.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
linecount = lambda x: sum([''.join(j)=='XMAS' for j in zip(*[x[i:] for i in range(4)])])

grid = [list(i) for i in rawinput.split('\n')]
n_xmas = 0
for h in range(4):
    diag = dict()
    for i,r in enumerate(grid):
        n_xmas += linecount(r)
        for j,c in enumerate(r):
            if i+j in diag:
                diag[i+j] += c
            else:
                diag[i+j] = c
    n_xmas += sum([linecount(r) for r in diag.values()])
    grid = [[*i][::-1] for i in zip(*grid)]
    
n_xmas
```

```python
# Part 2
diag = [dict() for i in '01']
for i,r in enumerate(grid):
    for j,c in enumerate(r):
        for k in [0,1]:
            if (ix:=[i+j,i-j][k]) in diag[k]:
                diag[k][ix] += c
            else:
                diag[k][ix] = c

len({((r:=max(i-len(grid[0])+1,0)+k)+1, i-r-1)
     for i in diag[0]
     for k,l in enumerate(zip(*[diag[0][i][j:] for j in range(3)])) 
     if ''.join(l) in ['MAS','SAM']}
    & {((r:=max(i,0)+k)+1,r-i+1)
       for i in diag[1]
       for k,l in enumerate(zip(*[diag[1][i][j:] for j in range(3)])) 
       if ''.join(l) in ['MAS','SAM']})
```
