```python
# Get raw data
with open('input/19.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
def is_possible(pattern):
    if pattern not in known:
        if pattern == '':
            known[pattern] = True
        else:
            known[pattern] = any([(pattern[:i+1] in avail)
                                  and is_possible(pattern[i+1:])
                                  for i in range(min(maxlen,len(pattern)))])
    return known[pattern]

raw_avail, raw_patterns = [i for i in rawinput.split('\n\n')]
avail = [i for i in raw_avail.split(', ')]
maxlen = max([len(i) for i in avail])
patterns = [i for i in raw_patterns.split('\n')]

known = dict()
sum([is_possible(pattern) for pattern in patterns])
```

```python
# Part 2
def n_ways_to_make(pattern):
    if pattern not in known2:
        if pattern == '':
            known2[pattern] = 1
        else:
            known2[pattern] = sum([n_ways_to_make(pattern[i+1:]) if (pattern[:i+1] in avail) else 0
                                  for i in range(min(maxlen,len(pattern)))])
    return known2[pattern]

known2 = dict()
sum([n_ways_to_make(pattern) for pattern in patterns])
```
