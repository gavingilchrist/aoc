```python
# Get raw data
with open('input/03.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
sum_products = 0
ix0 = 0
while 'mul(' in rawinput[ix0:]:
    ix1 = rawinput.index('mul(',ix0)
    if ',' in rawinput[ix1:]:
        ix2 = rawinput.index(',',ix1)
        if (y:=rawinput[ix1+4:ix2]).isnumeric() and ')' in rawinput[ix2:]:
            ix3 = rawinput.index(')',ix2)
            if (z:=rawinput[ix2+1:ix3]).isnumeric():
                sum_products += eval(f'{y}*{z}')
    ix0 = ix1+1
sum_products
```

```python
# Part 2
sum_products = 0
ix0 = 0
do = True
while 'mul(' in rawinput[ix0:]:
    ix1 = rawinput.index('mul(',ix0)
    if (z:=sorted([[rawinput[:ix1].index(i,ix0), i=="do()"] 
                   for i in ["do()", "don't()"] 
                   if i in rawinput[ix0:ix1]])):
        do = z[0][1]
        ix0 = z[0][0]+1
    else:
        if do and ',' in rawinput[ix1:]:
            ix2 = rawinput.index(',',ix1)
            if (y:=rawinput[ix1+4:ix2]).isnumeric() and ')' in rawinput[ix2:]:
                ix3 = rawinput.index(')',ix2)
                if (z:=rawinput[ix2+1:ix3]).isnumeric():
                    sum_products += eval(f'{y}*{z}')
        ix0 = ix1+1
sum_products
```
