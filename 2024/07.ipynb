```python
# Get raw data
with open('input/07.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
def is_poss(lhs,rhs):
    if len(lhs)==1:
        return lhs[0]==rhs
    else:
        return any([is_poss([lhs[0]+lhs[1], *lhs[2:]], rhs),
                    is_poss([lhs[0]*lhs[1], *lhs[2:]], rhs)])

equations = [[[int(k) for k in j[1].split()], int(j[0])] 
             for i in rawinput.split('\n') 
             for j in [i.split(': ')]]

sum([rhs for lhs,rhs in equations if is_poss(lhs,rhs)])
```

```python
# Part 2
def is_poss(lhs,rhs):
    if len(lhs)==1:
        return lhs[0]==rhs
    else:
        return any([is_poss([lhs[0]+lhs[1], *lhs[2:]], rhs),
                    is_poss([lhs[0]*lhs[1], *lhs[2:]], rhs),
                    is_poss([int(f'{lhs[0]}{lhs[1]}'), *lhs[2:]], rhs)])

equations = [[[int(k) for k in j[1].split()], int(j[0])] 
             for i in rawinput.split('\n') 
             for j in [i.split(': ')]]

sum([rhs for lhs,rhs in equations if is_poss(lhs,rhs)])
```
