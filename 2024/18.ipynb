```python
# Part 1
def evolve(reach):
    rp = np.minimum(np.pad(reach,1,constant_values=memarea)+1, memarea)
    return np.where((reach==memarea) & (memspace=='.'),
                    np.min(np.stack((rp[:-2,1:-1], 
                                     rp[2:,1:-1], 
                                     rp[1:-1,:-2], 
                                     rp[1:-1,2:])), 
                           axis=0),
                    reach)
    
def get_reach(startpt):
    reach = np.full_like(memspace, memarea, dtype=int)
    reach[startpt] = 0
    while np.any(reach != (reach:=evolve(reach))):  pass
    return reach

fbytes = np.array([[int(j) for j in i.split(',')[::-1]] 
                   for i in rawinput.split('\n')])
memsize = 71
memarea = memsize**2
memspace = np.full([memsize]*2,'.')
memspace[tuple(fbytes[:1024].T)] = '#'

get_reach((0,0))[-1,-1].item()
```
