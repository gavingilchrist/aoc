```python
# Get raw data
rawinput = 'ckczppom'
```

```python
import hashlib
```

```python
# Part 1
key = 0
while hashlib.md5(f'{rawinput}{key}'.encode('utf-8')).hexdigest()[:5] != '00000':
    key += 1
key
```
