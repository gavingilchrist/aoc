```python
# Get raw data
rawinput = 'hepxcrrq'
```

```python
# Part 1
req1 = lambda x: any([ord(k)-ord(j) == ord(l)-ord(k) == 1 
                      for j,k,l in zip(*[list(x[i:][:6]) for i in[0,1,2]])])
req2 = lambda x: all([i not in x for i in 'iol'])
req3 = lambda x, n=2: n<=0 if len(x)<2 else req3(x[2:], n-1) if x[0]==x[1] else req3(x[1:], n)
nextpw = lambda x: x[:(z:=[i for i in range(7, -1, -1) if ord(x[i])<ord('z')][0])] + chr(ord(x[z])+1) + 'a'*(7-z)

pw = rawinput
while not (req1(pw) & req2(pw) & req3(pw)):
    pw = nextpw(pw)
pw
```

```python
# Part 2
pw = nextpw(pw)
while not (req1(pw) & req2(pw) & req3(pw)):
    pw = nextpw(pw)
pw
```
