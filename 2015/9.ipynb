```python
# Get raw data
with open('input/9.txt', 'r') as f:
    rawinput = f.read()
```

```python
# Part 1
raw_distances = [[j[0], j[2], int(j[4])] for i in rawinput.split('\n') if i for j in [i.split()]]
distances = {j: {[*{*k}-{j}][0]: l for *k, l in raw_distances if j in k} 
             for j in {*sum([[*i] for i in zip(*raw_distances)][:2], [])}}
visited, distances['START'] = {i: False for i in distances}, {i: 0 for i in distances}

def getdist(visited, loc='START'):
    if (ytv:=[k for k,v in visited.items() if not v]):
        return min([distances[loc][i] + getdist({**visited, **{i: True}}, i) for i in ytv])
    else:
        return 0
        
getdist(visited)
```

```python
# Part 2
def getdist(visited, loc='START'):
    if (ytv:=[k for k,v in visited.items() if not v]):
        return max([distances[loc][i] + getdist({**visited, **{i: True}}, i) for i in ytv])
    else:
        return 0
        
getdist(visited)
```
