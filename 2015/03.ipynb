```python
# Get raw data
with open('input/03.txt', 'r') as f:
    rawinput = f.read()
```

```python
# Part 1
lochist = []
lochist += [(loc:=(0, 0))]
for m in rawinput:
    lochist += [(loc:=tuple([sum(i) for i in zip(loc, {'^': (-1, 0), 'v': (1, 0), '<': (0, -1), '>': (0, 1)}[m])]))]
len({*lochist})
```

```python
# Part 2
lochist = []
for s in [0,1]:
    lochist += [(loc:=(0, 0))]
    for m in rawinput[s::2]:
        lochist += [(loc:=tuple([sum(i) for i in zip(loc, {'^': (-1, 0), 'v': (1, 0), '<': (0, -1), '>': (0, 1)}[m])]))]
len({*lochist})
```
