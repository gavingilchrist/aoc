```python
# Get raw data
with open('input/22.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
known = dict()
def do_turn(player_hit, player_mana, boss_hit, shield_timer=0, poison_timer=0, recharge_timer=0, mana_spent=0, turn=0):
    dkey = f'{player_hit}|{player_mana}|{boss_hit}|{shield_timer}|{poison_timer}|{recharge_timer}|{mana_spent}|{turn}'
    if dkey not in known:
        if shield_timer:
            shield_timer -= 1
        if poison_timer:
            boss_hit -= 3
            poison_timer -= 1
        if recharge_timer:
            player_mana += 101
            recharge_timer -= 1
        if boss_hit <= 0:
            known[dkey] = [mana_spent]
        elif turn:
            player_hit -= (boss_damage - 7*(shield_timer>0))
            known[dkey] = do_turn(player_hit, player_mana, boss_hit, shield_timer, poison_timer, recharge_timer, mana_spent, 1-turn) if player_hit > 0 else []
        else:
            branch_poss = [*([] if player_mana<53 else do_turn(player_hit, player_mana-53, boss_hit-4, shield_timer, poison_timer, recharge_timer, mana_spent+53, 1-turn) if boss_hit>4 else [mana_spent+53]),
                           *([] if player_mana<73 else do_turn(player_hit+2, player_mana-73, boss_hit-2, shield_timer, poison_timer, recharge_timer, mana_spent+73, 1-turn) if boss_hit>2 else [mana_spent+73]),
                           *([] if player_mana<113 or shield_timer>0 else do_turn(player_hit, player_mana-113, boss_hit, 6, poison_timer, recharge_timer, mana_spent+113, 1-turn)),
                           *([] if player_mana<173 or poison_timer>0 else do_turn(player_hit, player_mana-173, boss_hit, shield_timer, 6, recharge_timer, mana_spent+173, 1-turn)),
                           *([] if player_mana<229 or recharge_timer>0 else do_turn(player_hit, player_mana-229, boss_hit, shield_timer, poison_timer, 5, mana_spent+229, 1-turn))]
            known[dkey] = [min(branch_poss)] if len(branch_poss) else []
    return known[dkey]

boss_hit, boss_damage = [int(i.split()[-1]) for i in rawinput.split('\n')]

do_turn(50, 500, boss_hit)[0]
```
