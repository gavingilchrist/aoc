```python
# Get raw data
with open('input/5.txt', 'r') as f:
    rawinput = f.read()
```

```python
# Part 1
strings = [i for i in rawinput.split('\n') if i]
sum([sum([i in 'aeiou' for i in string]) >= 3
     and any([string[i]==j for i,j in enumerate(string[1:])])
     and all([i not in string for i in ['ab', 'cd', 'pq', 'xy']])
    for string in strings])
```

```python
# Part 2
sum([any([string[s1:s1+2]==string[s2:s2+2] 
          for s1 in range(len(string)-3) 
          for s2 in range(s1+2, len(string)-1)])
     and any([string[s1]==string[s1+2] 
              for s1 in range(len(string)-2)])
     for string in strings])
```
