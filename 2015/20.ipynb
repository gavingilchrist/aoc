```python
import numpy as np
```

```python
# Part 1
target = 29000000
n_pres = np.zeros(1000000, dtype=int)
elf = 1
while True:
    n_pres[::elf] += (elf*10)
    if n_pres[0] >= target and np.any(z:=n_pres[1:elf+1] >= target):
        break
    elf += 1
np.where(z)[0][0].item()+1
```

```python
# Part 2
target = 29000000
n_pres = np.zeros(10000000, dtype=int)
elf = 1
while True:
    n_pres[:elf*51:elf] += (elf*11)
    if n_pres[0] >= target and np.any(z:=n_pres[1:elf+1] >= target):
        break
    elf += 1
np.where(z)[0][0].item()+1
```
