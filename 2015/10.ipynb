```python
# Part 1
getnext = lambda x: ''.join([f'{bps[i+1]-bps[i]}{x[bps[i]]}' 
                             for bps in[[0]+[i for i,(j,k) in enumerate(zip(list(x[1:]), list(x[:-1])), 1) if j!=k]+[len(x)]]
                             for i in range(len(bps)-1)])
numstr = rawinput
for i in range(40):
    numstr = getnext(numstr)
len(numstr)
```
