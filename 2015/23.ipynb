```python
# Get raw data
with open('input/23.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
instr = [((z:=i.split()[0]), 
          '' if z=='jmp' else i.split()[1][:1], 
          int(i.split()[-1]) if z.startswith('j') else 0) 
         for i in rawinput.split('\n')]

def run(registers):
    ptr = 0
    while ptr < len(instr):
        i, r, p = instr[ptr]
        if i=='hlf':  
            registers[r] //= 2
            ptr += 1
        elif i=='tpl':  
            registers[r] *= 3
            ptr += 1
        elif i=='inc':  
            registers[r] += 1
            ptr += 1
        elif i=='jmp':  
            ptr += p
        elif i=='jie':
            ptr += (p if registers[r] % 2 == 0 else 1)
        elif i=='jio':
            ptr += (p if registers[r] == 1 else 1)
    return registers['b']

run({'a': 0, 'b': 0})
```

```python
# Part 2
run({'a': 1, 'b': 0})
```
