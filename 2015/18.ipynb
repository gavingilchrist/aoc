```python
# Get raw data
with open('input/18.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
lights_start = np.pad(np.array([[{'.': 0, '#': 1}[j] for j in i] 
                                for i in rawinput.split('\n')]), 
                      1, constant_values=0)

n_adj = lambda x: np.sum(np.stack([x[i:i+100, j:j+100] 
                                   for i in [0,1,2] for j in [0,1,2] 
                                   if i*j != 1]), axis=0)

lights = np.copy(lights_start)
for i in range(100):
    lights[1:-1, 1:-1] = (((z:=n_adj(lights)) == 3) | ((z == 2) & (lights[1:-1, 1:-1] == 1))).astype(int)

np.sum(lights).item()
```

```python
# Part 2
lights = np.copy(lights_start)
lights[1:102:99, 1:102:99] = 1
for i in range(100):
    lights[1:-1, 1:-1] = (((z:=n_adj(lights)) == 3) | ((z == 2) & (lights[1:-1, 1:-1] == 1))).astype(int)
    lights[1:102:99, 1:102:99] = 1

np.sum(lights).item()
```
