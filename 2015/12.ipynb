```python
# Get raw data
with open('input/12.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
def add_nums(obj):
    if isinstance(obj, int):
        return obj
    if isinstance(obj, str):
        return 0
    elif isinstance(obj, dict):
        return sum([add_nums(i) for i in obj.values()])
    elif isinstance(obj, list):
        return sum([add_nums(i) for i in obj])
add_nums(eval(rawinput))
```
