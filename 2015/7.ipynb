```python
# Get raw data
with open('input/7.txt', 'r') as f:
    rawinput = f.read()
```

```python
# Part 1
instrs = [i for i in rawinput.split('\n') if i]
f = lambda x: [f"wires['{x}']", x][x.isdigit()]
wires = {}
while wires.get('a') is None:
    for instr in instrs:
        if 'AND' in instr:
            in1, _, in2, _, out = instr.split()
            cmd = f'{f(out)} = {f(in1)} & {f(in2)}'
        elif 'OR' in instr:
            in1, _, in2, _, out = instr.split()
            cmd = f'{f(out)} = {f(in1)} | {f(in2)}'
        elif 'NOT' in instr:
            _, in1, _, out = instr.split()
            cmd = f'{f(out)} = (~{f(in1)})%65536'
        elif 'LSHIFT' in instr:
            in1, _, in2, _, out = instr.split()
            cmd = f'{f(out)} = {f(in1)} << {f(in2)}'
        elif 'RSHIFT' in instr:
            in1, _, in2, _, out = instr.split()
            cmd = f'{f(out)} = {f(in1)} >> {f(in2)}'
        else:
            in1, _, out = instr.split()
            cmd = f'{f(out)} = {f(in1)}'
        if out not in wires:
            try:
                exec(cmd)
            except KeyError:
                pass
wires['a']
```
