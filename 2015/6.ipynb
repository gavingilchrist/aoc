```python
# Get raw data
with open('input/6.txt', 'r') as f:
    rawinput = f.read()
```

```python
import numpy as np
```

```python
# Part 1
bulbs = np.zeros((1000, 1000), dtype=int)
for instr, x1, y1, _, x2, y2 in [i.replace('n o', 'n_o').replace(',', ' ').split() 
                                 for i in rawinput.split('\n') if i]:
    x1, y1, x2, y2 = [int(i) for i in [x1, y1, x2, y2]]
    if instr=='turn_on':  bulbs[y1:y2+1, x1:x2+1] = 1
    elif instr=='turn_off':  bulbs[y1:y2+1, x1:x2+1] = 0
    else:  bulbs[y1:y2+1, x1:x2+1] ^= 1
np.sum(bulbs).item()
```

```python
# Part 1
bulbs = np.zeros((1000, 1000), dtype=int)
for instr, x1, y1, _, x2, y2 in [i.replace('n o', 'n_o').replace(',', ' ').split() 
                                 for i in rawinput.split('\n') if i]:
    x1, y1, x2, y2 = [int(i) for i in [x1, y1, x2, y2]]
    if instr=='turn_on':  bulbs[y1:y2+1, x1:x2+1] += 1
    elif instr=='turn_off':  bulbs[y1:y2+1, x1:x2+1] = np.maximum(bulbs[y1:y2+1, x1:x2+1]-1, 0)
    else:  bulbs[y1:y2+1, x1:x2+1] += 2
np.sum(bulbs).item()
```
