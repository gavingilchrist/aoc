```python
# Get raw data
with open('input/14.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
descs = {j[0]: [int(j[3]), int(j[6]), int(j[13])] 
         for i in rawinput.split('\n') for j in [i.split()]}
seconds = 2503
max([((seconds//(fly+rest))*fly + min(seconds%(fly+rest), fly)) * speed
     for r, (speed, fly, rest) in descs.items()])
```

```python
# Part 2
racepos = [*zip(*[[sum(i[:j+1])
                   for i in [(([speed]*fly + [0]*rest) * ((seconds//(fly+rest))+1))[:seconds]]
                   for j in range(len(i))]
                  for r, (speed, fly, rest) in descs.items()])]
pts = [[int(j==max(i)) for j in i] for i in racepos]
max([sum(i) for i in zip(*pts)])
```
