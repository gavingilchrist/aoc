```python
# Get raw data
with open('input/17.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
containers = sorted([int(i) for i in rawinput.split('\n')])

def count_combos(ctrs_left, amt=150):
    if amt:
        return sum([count_combos(ctrs_left[i+1:], amt-j) 
                    for i,j in enumerate(ctrs_left) 
                    if j <= amt])
    else:
        return 1

count_combos(containers)
```

```python
# Part 2
def count_combos(ctrs_left, amt=150, nc=0):
    if amt:
        return [[i, sum([k for j,k in z if j==i])]
                for z in[[k for i,j in enumerate(ctrs_left) if j <= amt
                            for k in count_combos(ctrs_left[i+1:], amt-j, nc+1)]]
                for i in {*([*zip(*z)] or [[]])[0]}]
    else:
        return [[nc, 1]]

[j for z in [count_combos(containers)] for i,j in z if i==min([*zip(*z)][0])][0]
```
