```python
# Get raw data
with open('input/16.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
target = {
    'children': 3, 
    'cats': 7, 
    'samoyeds': 2, 
    'pomeranians': 3, 
    'akitas': 0, 
    'vizslas': 0, 
    'goldfish': 5, 
    'trees': 3, 
    'cars': 2, 
    'perfumes': 1
}
```

```python
# Part 1
sues = [eval("{'"+i[i.index(':')+2:].replace(':', "':").replace(', ', ", '")+'}') for i in rawinput.split('\n')]
[i for i,s in enumerate(sues, 1) 
 if all([s[j]==target[j] for j in s])][0]
```

```python
# Part 2
[i for i,s in enumerate(sues, 1) 
 if all([(j in ['cats', 'trees'] and s[j]>target[j])
         or (j in ['pomeranians', 'goldfish'] and s[j]<target[j])
         or (j not in ['cats', 'trees', 'pomeranians', 'goldfish'] and s[j]==target[j])
         for j in s])][0]
```
