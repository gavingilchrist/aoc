```python
# Get raw data
with open('input/21.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
weapon_stats = [[ 8, 4, 0], [10, 5, 0], [25, 6, 0], [40, 7, 0], [74, 8, 0]]
armor_stats = [[  0, 0, 0], [ 13, 0, 1], [ 31, 0, 2], [ 53, 0, 3], [ 75, 0, 4], [102, 0, 5]]
ring_stats = [[  0, 0, 0], [ 25, 1, 0], [ 50, 2, 0], [100, 3, 0], [ 20, 0, 1], [ 40, 0, 2], [ 80, 0, 3]]
ring_pair_stats = [[0, 0, 0],
                   *[[sum(k) for k in zip(ring_stats[i], ring_stats[j])]
                     for i in range(6) for j in range(i+1,7)]]
scenarios = sorted([[sum(k) for k in zip(w, a, r)]
                    for w in weapon_stats
                    for a in armor_stats
                    for r in ring_pair_stats])

lhc_hit = 100
boss_hit, boss_damage, boss_armor = [int(i.split()[-1]) for i in rawinput.split('\n')]

get_hit_deg = lambda i,j,k: [*range(i, (z-y if (z:=i % (y:=max(j-k, 1))) else z)-1, -y)]

for cost, lhc_damage, lhc_armor in scenarios:
    if [max(j,0)
        for i in zip(get_hit_deg(boss_hit, lhc_damage, boss_armor), 
                     get_hit_deg(lhc_hit, boss_damage, lhc_armor))
        for j in i].index(0) % 2 == 0:
        break
cost
```

```python
# Part 2
for cost, lhc_damage, lhc_armor in scenarios[::-1]:
    if [max(j,0)
        for i in zip(get_hit_deg(boss_hit, lhc_damage, boss_armor), 
                     get_hit_deg(lhc_hit, boss_damage, lhc_armor))
        for j in i].index(0) % 2 == 1:
        break
cost
```
