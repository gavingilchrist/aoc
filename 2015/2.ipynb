```python
# Get raw data
with open('input/2.txt', 'r') as f:
    rawinput = f.read()
```

```python
# Part 1
sum([3*x*y + 2*x*z + 2*y*z 
     for i in rawinput.split('\n') if i 
     for x,y,z in[sorted([int(j) for j in i.split('x')])]])
```

```python
# Part 2
sum([(x+y)*2 + x*y*z 
     for i in rawinput.split('\n') if i 
     for x,y,z in[sorted([int(j) for j in i.split('x')])]])
```
