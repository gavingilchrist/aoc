```python
# Get raw data
with open('input/24.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
packages = [int(i) for i in rawinput.split('\n')]
target = sum(packages) // 3

def get_combos(pkg, tgt):
    if tgt == 0:
        return [set()]
    else:
        return [{i}|k 
                for i in pkg if i <= tgt 
                for k in get_combos([j for j in pkg if i<j<=tgt], tgt-i)]
                
groups = get_combos(packages, target)

def chkgp(i):
    for j in groups:
        if len(i&j)==0:
            return True
    return False

for gpsize in sorted({len(i) for i in groups}):
    groups_sub = [i for i in groups if len(i)==gpsize]
    groups_sub_chk = []
    for i in groups_sub:
        if chkgp(i):
            groups_sub_chk += [i]
    if groups_sub_chk:
        break

prod = lambda ns: ns[0]*prod(ns[1:]) if ns else 1

min([prod([*i]) for i in groups_sub_chk])
```
