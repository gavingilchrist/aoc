```python
# Get raw data
with open('input/08.txt', 'r') as f:
    rawinput = f.read()
```

```python
# Part 1
strlits = [i for i in rawinput.split('\n') if i]
sum([len(s) for s in strlits]) - sum([len(eval(s)) for s in strlits])
```

```python
# Part 2
sum([len('"'+''.join(['\\"' if i=='"' else '\\\\' if i=='\\' else i for i in s])+'"') for s in strlits]) - sum([len(s) for s in strlits])
```
