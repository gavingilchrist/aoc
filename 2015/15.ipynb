```python
# Get raw data
with open('input/15.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
import numpy as np
```

```python
# Part 1
igprop = np.array([[[int(j[(k+1)*2].replace(',', '')) for k in range(5)]
                    for i in rawinput.split('\n') for j in [i.split()]]])
mixcombos = np.array([[[i1], [i2], [i3], [100-i1-i2-i3]] 
                      for i1 in range(101)
                      for i2 in range(101-i1)
                      for i3 in range(101-i1-i2)])
score = np.prod(np.maximum(np.sum(mixcombos * igprop[:, :, :-1], axis=1), 0), axis=1)
np.max(score)
```

```python
# Part 2
filt = np.sum(mixcombos[:, :, 0] * igprop[:, :, -1], axis=1) == 500
score = np.prod(np.maximum(np.sum(mixcombos[filt] * igprop[:, :, :-1], axis=1), 0), axis=1)
np.max(score)
```
