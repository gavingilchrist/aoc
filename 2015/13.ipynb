```python
# Get raw data
with open('input/13.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
raw_scores_1way = [[tuple(sorted([j[0], j[-1][:-1]])), [-1, 1][j[2]=='gain']*int(j[3])] 
                   for i in rawinput.split('\n') for j in [i.split()]]
raw_scores_2way = [[*j, sum([l for k,l in raw_scores_1way if k==j])] for j in {i[0] for i in raw_scores_1way}]
scores = {j: {[*{*k}-{j}][0]: l for *k, l in raw_scores_2way if j in k} 
          for j in {*sum([[*i] for i in zip(*raw_scores_2way)][:2], [])}}
placed = {i: i=='Alice' for i in scores}

def getbest(placed, lp='Alice'):
    if (ytp:=[k for k,v in placed.items() if not v]):
        return max([scores[lp][i] + getbest({**placed, **{i: True}}, i) for i in ytp])
    else:
        return scores[lp]['Alice']

getbest(placed)
```

```python
# Part 2
raw_scores_2way += [[i, 'ME', 0] for i in {*sum([[*i] for i in zip(*raw_scores_2way)][:2], [])}]
scores = {j: {[*{*k}-{j}][0]: l for *k, l in raw_scores_2way if j in k} 
          for j in {*sum([[*i] for i in zip(*raw_scores_2way)][:2], [])}}
placed = {i: i=='Alice' for i in scores}

def getbest(placed, lp='Alice'):
    if (ytp:=[k for k,v in placed.items() if not v]):
        return max([scores[lp][i] + getbest({**placed, **{i: True}}, i) for i in ytp])
    else:
        return scores[lp]['Alice']

getbest(placed)
```
