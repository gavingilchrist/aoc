```python
# Get raw data
with open('input/19.txt', 'r') as f:
    rawinput = f.read().strip()
```

```python
# Part 1
repl = {i: [k for j,k in z if j==i] 
        for z in [[i.split()[::2] for i in rawinput.split('\n') if i][:-1]]
        for i in sorted({*[*zip(*z)][0]})}
molecule = rawinput.split('\n')[-1]

len({molecule[:j]+k+molecule[j+len(i):]
     for i in repl
     for j in range(len(molecule)) if molecule[j:].startswith(i)
     for k in repl[i]})
```

```python
# Part 2
get_comp = lambda x: [(x+'X')[i:i+2-int(64<ord((x+'X')[i+1])<91)] for i in range(len(x)) if 64<ord(x[i])<91]

def get_sep_ix(mol):
    if '|' in mol:
        nb = [0]
        for i in mol:
            nb += [nb[-1]+{'(': 1, ')': -1}.get(i, 0)]
        sepix = [i for i,(j,k) in enumerate(zip(list(mol), nb[1:])) if j=='|' and k==0]
        return [-1]+sepix+[len(mol)] if sepix else []
    else:
        return []

def get_parent_ix(mol):
    smix = []
    p = 0
    while '(' in mol[p:]:
        p += mol[p:].index('(')
        q = p+1
        while mol[p:q].count('(') - mol[p:q].count(')') > 0:
            q += 1
        smix += [[p, q]]
        p = q
    return smix
    
def get_min_steps(mol):
    if (sepix:=get_sep_ix(mol)):
        return sum([get_min_steps(mol[sepix[i-1]+1:sepix[i]]) for i in range(1, len(sepix))])
    elif '(' in mol:
        bl = 0
        nsteps = 0
        for bs, be in get_parent_ix(mol):
            nsteps += len(get_comp(mol[bl:bs]))+get_min_steps(mol[bs:be][1:-1])+1
            bl = be
        nsteps += len(get_comp(mol[be:]))-1
        return nsteps
    else:
        return len(get_comp(mol))-1

get_min_steps(molecule.replace('Rn', '(').replace('Ar', ')').replace('Y', '|'))
```
